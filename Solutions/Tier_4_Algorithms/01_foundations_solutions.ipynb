## Problem 1: Complexity Analysis (1 star)

```python
# (a) Single loop over n elements
# Time complexity: O(n)
# Reason: The loop executes exactly n times, one step per element.

# (b) Nested loop n*n
# Time complexity: O(n^2)
# Reason: The outer loop runs n times; for each iteration the inner loop runs n times, giving n*n total.

# (c) Binary search
# Time complexity: O(log n)
# Reason: Each iteration halves the search space, so at most log2(n) iterations are needed.

# (d) Three nested loops
# Time complexity: O(n^3)
# Reason: Each of the three loops independently runs n times, giving n*n*n total iterations.

# (e) Loop that halves n each iteration
# Time complexity: O(log n)
# Reason: i is halved each step, so after k steps i = n/2^k; the loop stops when i <= 1, i.e. k = log2(n).

print("Answers:")
print("(a) O(n)")
print("(b) O(n^2)")
print("(c) O(log n)")
print("(d) O(n^3)")
print("(e) O(log n)")
```

## Problem 2: Sorting Comparison (1 star)

```python
import random
import time


def selection_sort(arr: list) -> list:
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr


def merge_sort(arr: list) -> list:
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return _merge(left, right)


def _merge(left: list, right: list) -> list:
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result


random.seed(42)
sizes = [1000, 5000, 10000]
for n in sizes:
    data = [random.randint(0, 100000) for _ in range(n)]

    t0 = time.perf_counter()
    selection_sort(data[:])
    t_sel = time.perf_counter() - t0

    t0 = time.perf_counter()
    merge_sort(data[:])
    t_merge = time.perf_counter() - t0

    print(f"n={n:<6}  selection={t_sel:.3f}s   merge={t_merge:.4f}s")
```

## Problem 3: Binary Search Variants (2 stars)

```python
def find_leftmost(arr: list, target: float) -> int:
    lo, hi = 0, len(arr) - 1
    result = -1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            result = mid
            hi = mid - 1   # keep searching left
        elif arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return result


def find_rightmost(arr: list, target: float) -> int:
    lo, hi = 0, len(arr) - 1
    result = -1
    while lo <= hi:
        mid = (lo + hi) // 2
        if arr[mid] == target:
            result = mid
            lo = mid + 1   # keep searching right
        elif arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
    return result


def find_insertion_point(arr: list, target: float) -> int:
    lo, hi = 0, len(arr)
    while lo < hi:
        mid = (lo + hi) // 2
        if arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid
    return lo


expression = [0.3, 1.1, 1.7, 1.7, 1.7, 2.4, 3.0, 3.0, 4.5, 6.2]

print(f"Array: {expression}")
print(f"find_leftmost(1.7)         -> {find_leftmost(expression, 1.7)}")   # 2
print(f"find_rightmost(1.7)        -> {find_rightmost(expression, 1.7)}")  # 4
print(f"find_insertion_point(2.0)  -> {find_insertion_point(expression, 2.0)}")  # 5
print(f"find_leftmost(5.0)         -> {find_leftmost(expression, 5.0)}")   # -1
```

## Problem 4: Stack-Based Expression Evaluator (2 stars)

```python
def evaluate_postfix(expression: str) -> float:
    stack = []
    ops = {
        "+": lambda a, b: a + b,
        "-": lambda a, b: a - b,
        "*": lambda a, b: a * b,
        "/": lambda a, b: a / b,
    }
    for token in expression.split():
        if token in ops:
            b = stack.pop()
            a = stack.pop()
            stack.append(ops[token](a, b))
        else:
            stack.append(float(token))
    return stack[0]


print(evaluate_postfix("3 4 + 2 *"))          # 14.0
print(evaluate_postfix("5 1 2 + 4 * + 3 -"))  # 14.0
print(evaluate_postfix("100 25 /"))            # 4.0
print(evaluate_postfix("6.022 10 23 * *"))     # 6.022e+24
```

## Problem 5: Priority Queue (2 stars)

```python
import heapq


class MinHeapPQ:
    """Min-heap priority queue backed by the standard heapq module."""

    def __init__(self):
        self._heap: list = []   # (priority, item)

    def insert(self, item, priority: int | float) -> None:
        heapq.heappush(self._heap, (priority, item))

    def extract_min(self) -> tuple:
        priority, item = heapq.heappop(self._heap)
        return item, priority

    def peek(self) -> tuple:
        priority, item = self._heap[0]
        return item, priority

    def __len__(self) -> int:
        return len(self._heap)


jobs = [
    ("sample_001", 2),
    ("sample_002", 7),
    ("sample_003", 1),
    ("sample_004", 9),
    ("sample_005", 3),
]

pq = MinHeapPQ()
for name, priority in jobs:
    pq.insert(name, priority)

print(f"Next job (peek): {pq.peek()}")
print()
while len(pq) > 0:
    item, priority = pq.extract_min()
    print(f"Processing: {item} (priority {priority})")
```

## Problem 6: BST Operations (2 stars)

```python
import random


class BSTNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class BST:
    def __init__(self):
        self.root = None

    def insert(self, value) -> None:
        self.root = self._insert(self.root, value)

    def _insert(self, node, value):
        if node is None:
            return BSTNode(value)
        if value < node.value:
            node.left = self._insert(node.left, value)
        elif value > node.value:
            node.right = self._insert(node.right, value)
        return node  # duplicate: ignore

    def search(self, value) -> bool:
        node = self.root
        while node:
            if value == node.value:
                return True
            node = node.left if value < node.value else node.right
        return False

    def delete(self, value) -> None:
        self.root = self._delete(self.root, value)

    def _delete(self, node, value):
        if node is None:
            return None
        if value < node.value:
            node.left = self._delete(node.left, value)
        elif value > node.value:
            node.right = self._delete(node.right, value)
        else:
            # Node to delete found
            if node.left is None:
                return node.right
            if node.right is None:
                return node.left
            # Two children: replace with in-order successor (min of right subtree)
            successor = self._min_node(node.right)
            node.value = successor.value
            node.right = self._delete(node.right, successor.value)
        return node

    def _min_node(self, node):
        while node.left:
            node = node.left
        return node

    def inorder(self) -> list:
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node is None:
            return
        self._inorder(node.left, result)
        result.append(node.value)
        self._inorder(node.right, result)


random.seed(42)
expression_values = [round(random.uniform(0, 10), 2) for _ in range(20)]

bst = BST()
for v in expression_values:
    bst.insert(v)
print(f"Inserted {len(expression_values)} values.")

sorted_vals = bst.inorder()
lowest_five = sorted_vals[:5]
print(f"Deleting 5 lowest: {lowest_five}")

for v in lowest_five:
    bst.delete(v)

print(f"Remaining (sorted): {bst.inorder()}")
```

## Problem 7: Hash Table with Chaining (3 stars)

```python
class HashTableChaining:
    def __init__(self, size: int = 64):
        self.size = size
        self._buckets: list[list] = [[] for _ in range(size)]
        self.n_items = 0
        self.collision_count = 0

    def _hash(self, key: str) -> int:
        # Polynomial rolling hash
        h = 0
        for ch in key:
            h = (h * 31 + ord(ch)) % self.size
        return h

    def insert(self, key: str, value) -> None:
        idx = self._hash(key)
        bucket = self._buckets[idx]
        for i, (k, _) in enumerate(bucket):
            if k == key:
                bucket[i] = (key, value)  # update existing
                return
        if bucket:  # non-empty bucket means a collision
            self.collision_count += 1
        bucket.append((key, value))
        self.n_items += 1

    def search(self, key: str):
        idx = self._hash(key)
        for k, v in self._buckets[idx]:
            if k == key:
                return v
        return None

    def delete(self, key: str) -> None:
        idx = self._hash(key)
        bucket = self._buckets[idx]
        for i, (k, _) in enumerate(bucket):
            if k == key:
                del bucket[i]
                self.n_items -= 1
                return

    @property
    def load_factor(self) -> float:
        return self.n_items / self.size


dna = (
    "ATCGATCGATCGTAGCTAGCTAGCATGCATGCATGCGATCGATCG"
    "TAGCTAGCTAGCATGCATGCATGCGATCGATCGTAGCTAGCTAGC"
)
k = 4
ht = HashTableChaining(size=64)

for i in range(len(dna) - k + 1):
    kmer = dna[i:i + k]
    current = ht.search(kmer)
    ht.insert(kmer, (current or 0) + 1)

print(f"Table size: {ht.size} buckets")
print(f"Items stored: {ht.n_items} unique {k}-mers")
print(f"Load factor: {ht.load_factor:.3f}")
print(f"Collisions during inserts: {ht.collision_count}")
print(f"Count of 'ATCG': {ht.search('ATCG')}")
```

## Problem 8: Bloom Filter Implementation (3 stars)

```python
import math
import random


class BloomFilter:
    def __init__(self, m: int, k: int):
        self.m = m
        self.k = k
        self._bits = bytearray(m)   # 0 or 1 per position

    def _hashes(self, item: str) -> list[int]:
        # Double hashing: h_i(x) = (h1(x) + i * h2(x)) % m
        h1 = hash(item) % self.m
        h2 = hash(item + "\x00") % self.m
        return [(h1 + i * h2) % self.m for i in range(self.k)]

    def add(self, item: str) -> None:
        for pos in self._hashes(item):
            self._bits[pos] = 1

    def __contains__(self, item: str) -> bool:
        return all(self._bits[pos] for pos in self._hashes(item))


random.seed(0)

def random_dna(length: int) -> str:
    return "".join(random.choice("ACGT") for _ in range(length))

n_insert = 10_000
m = 100_000
k = 7

genome = random_dna(10_100)
insert_kmers = [genome[i:i+6] for i in range(n_insert)]

bf = BloomFilter(m=m, k=k)
for kmer in insert_kmers:
    bf.add(kmer)

other = random_dna(2_000)
inserted_set = set(insert_kmers)
test_kmers = [other[i:i+6] for i in range(len(other) - 5) if other[i:i+6] not in inserted_set]
test_kmers = test_kmers[:1_000]

fp = sum(1 for km in test_kmers if km in bf)
actual_rate = fp / len(test_kmers)
theoretical_rate = (1 - math.exp(-k * n_insert / m)) ** k

print(f"Inserted {n_insert} k-mers  (m={m}, k={k})")
print(f"Actual false positive rate:      {actual_rate:.3f}")
print(f"Theoretical false positive rate: {theoretical_rate:.3f}")
```
