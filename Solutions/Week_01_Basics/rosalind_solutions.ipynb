## Problem 1: Counting DNA Nucleotides

Count occurrences of A, C, G, T in a DNA string.

```python
def count_nucleotides(dna: str) -> dict:
    """Count each nucleotide in a DNA string."""
    return {nuc: dna.count(nuc) for nuc in 'ACGT'}

# Test
test_dna = "AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC"
result = count_nucleotides(test_dna)
print(f"A: {result['A']}, C: {result['C']}, G: {result['G']}, T: {result['T']}")
# Expected: A: 20, C: 12, G: 17, T: 21
```

## Problem 2: Transcribing DNA into RNA

Replace all T's with U's to convert DNA to RNA.

```python
def transcribe(dna: str) -> str:
    """Transcribe DNA to RNA by replacing T with U."""
    return dna.replace('T', 'U')

# Test
test_dna = "GATGGAACTTGACTACGTAAATT"
print(transcribe(test_dna))
# Expected: GAUGGAACUUGACUACGUAAAUU
```

## Problem 3: Complementing a Strand of DNA

Return the reverse complement of a DNA string.

```python
def reverse_complement(dna: str) -> str:
    """Return the reverse complement of a DNA string."""
    complement = str.maketrans('ATGC', 'TACG')
    return dna.translate(complement)[::-1]

# Alternative using dictionary
def reverse_complement_v2(dna: str) -> str:
    comp = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    return ''.join(comp[nuc] for nuc in reversed(dna))

# Test
test_dna = "AAAACCCGGT"
print(reverse_complement(test_dna))
# Expected: ACCGGGTTTT
```

## Problem 4: Computing GC Content

Find the sequence with highest GC content from a FASTA file.

```python
def parse_fasta(fasta_text: str) -> dict:
    """Parse FASTA format text into a dictionary."""
    sequences = {}
    current_id = None
    current_seq = []
    
    for line in fasta_text.strip().split('\n'):
        if line.startswith('>'):
            if current_id:
                sequences[current_id] = ''.join(current_seq)
            current_id = line[1:].split()[0]
            current_seq = []
        else:
            current_seq.append(line.strip())
    
    if current_id:
        sequences[current_id] = ''.join(current_seq)
    
    return sequences

def gc_content(seq: str) -> float:
    """Calculate GC content as percentage."""
    if len(seq) == 0:
        return 0.0
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

def highest_gc_content(fasta_text: str) -> tuple:
    """Find sequence with highest GC content."""
    sequences = parse_fasta(fasta_text)
    best_id = None
    best_gc = -1
    
    for seq_id, seq in sequences.items():
        gc = gc_content(seq)
        if gc > best_gc:
            best_gc = gc
            best_id = seq_id
    
    return best_id, best_gc

# Test
test_fasta = """>Rosalind_6404
CCTGCGGAAGATCGGCACTAGAATAGCCAGAACCGTTTCTCTGAGGCTTCCGGCCTTCCC
TCCCACTAATAATTCTGAGG
>Rosalind_5959
CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCCCTATCCAGGCGCTCCGCCGAAGGTCT
ATATCCATTTGTCAGCAGACACGC
>Rosalind_0808
CCACCCTCGTGGTATGGCTAGGCATTCAGGAACCGGAGAACGCTTCAGACCAGCCCGGAC
TGGGAACCTGCGGGCAGTAGGTGGAAT"""

seq_id, gc = highest_gc_content(test_fasta)
print(f"{seq_id}\n{gc:.6f}")
```

## Problem 5: Counting Point Mutations (Hamming Distance)

Count the number of positions where two strings differ.

```python
def hamming_distance(s1: str, s2: str) -> int:
    """Calculate the Hamming distance between two strings."""
    return sum(c1 != c2 for c1, c2 in zip(s1, s2))

# Test
s1 = "GAGCCTACTAACGGGAT"
s2 = "CATCGTAATGACGGCCT"
print(hamming_distance(s1, s2))
# Expected: 7
```

## Problem 6: Finding a Motif in DNA

Find all starting positions of a substring (1-indexed).

```python
def find_motif(dna: str, motif: str) -> list:
    """Find all starting positions of motif in DNA (1-indexed)."""
    positions = []
    start = 0
    while True:
        pos = dna.find(motif, start)
        if pos == -1:
            break
        positions.append(pos + 1)  # Convert to 1-indexed
        start = pos + 1
    return positions

# Alternative using list comprehension
def find_motif_v2(dna: str, motif: str) -> list:
    return [i + 1 for i in range(len(dna) - len(motif) + 1) 
            if dna[i:i+len(motif)] == motif]

# Test
dna = "GATATATGCATATACTT"
motif = "ATAT"
print(' '.join(map(str, find_motif(dna, motif))))
# Expected: 2 4 10
```

## Problem 7: Translating RNA into Protein

Translate an RNA string into a protein string using the genetic code.

```python
CODON_TABLE = {
    'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
    'UAU': 'Y', 'UAC': 'Y', 'UAA': '*', 'UAG': '*',
    'UGU': 'C', 'UGC': 'C', 'UGA': '*', 'UGG': 'W',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'AUG': 'M',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
    'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}

def translate_rna(rna: str) -> str:
    """Translate RNA to protein, stopping at stop codon."""
    protein = []
    for i in range(0, len(rna) - 2, 3):
        codon = rna[i:i+3]
        amino_acid = CODON_TABLE.get(codon, '')
        if amino_acid == '*':
            break
        protein.append(amino_acid)
    return ''.join(protein)

# Test
rna = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
print(translate_rna(rna))
# Expected: MAMAPRTEINSTRING
```

## Problem 8: Calculating Protein Mass

Calculate the total mass of a protein string.

```python
MONOISOTOPIC_MASS = {
    'A': 71.03711, 'C': 103.00919, 'D': 115.02694, 'E': 129.04259,
    'F': 147.06841, 'G': 57.02146, 'H': 137.05891, 'I': 113.08406,
    'K': 128.09496, 'L': 113.08406, 'M': 131.04049, 'N': 114.04293,
    'P': 97.05276, 'Q': 128.05858, 'R': 156.10111, 'S': 87.03203,
    'T': 101.04768, 'V': 99.06841, 'W': 186.07931, 'Y': 163.06333
}

def protein_mass(protein: str) -> float:
    """Calculate total monoisotopic mass of a protein."""
    return sum(MONOISOTOPIC_MASS[aa] for aa in protein)

# Test
protein = "SKADYEK"
print(f"{protein_mass(protein):.3f}")
# Expected: 821.392
```

## Problem 9: Finding a Shared Subsequence

Find the longest common subsequence of two sequences.

```python
def longest_common_substring(s1: str, s2: str) -> str:
    """Find the longest common substring of two strings."""
    # Ensure s1 is shorter for efficiency
    if len(s1) > len(s2):
        s1, s2 = s2, s1
    
    # Try all substrings of s1 from longest to shortest
    for length in range(len(s1), 0, -1):
        for start in range(len(s1) - length + 1):
            substring = s1[start:start + length]
            if substring in s2:
                return substring
    return ""

# Dynamic programming version for efficiency
def lcs_dp(s1: str, s2: str) -> str:
    """Find longest common substring using dynamic programming."""
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    max_len = 0
    end_pos = 0
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
                if dp[i][j] > max_len:
                    max_len = dp[i][j]
                    end_pos = i
    
    return s1[end_pos - max_len:end_pos]

# Test
fasta = """>Rosalind_1
GATTACA
>Rosalind_2
TAGACCA
>Rosalind_3
ATACA"""

seqs = list(parse_fasta(fasta).values())
result = seqs[0]
for seq in seqs[1:]:
    result = longest_common_substring(result, seq)
print(result)
# Expected: TA or AC (any valid common substring)
```

## Problem 10: Consensus and Profile

Build a profile matrix and consensus string from aligned sequences.

```python
def build_profile(sequences: list) -> dict:
    """Build a profile matrix from aligned sequences."""
    if not sequences:
        return {}
    
    seq_len = len(sequences[0])
    profile = {nuc: [0] * seq_len for nuc in 'ACGT'}
    
    for seq in sequences:
        for i, nuc in enumerate(seq):
            if nuc in profile:
                profile[nuc][i] += 1
    
    return profile

def get_consensus(profile: dict) -> str:
    """Get consensus sequence from profile matrix."""
    if not profile or not profile.get('A'):
        return ""
    
    seq_len = len(profile['A'])
    consensus = []
    
    for i in range(seq_len):
        max_count = -1
        max_nuc = 'A'
        for nuc in 'ACGT':
            if profile[nuc][i] > max_count:
                max_count = profile[nuc][i]
                max_nuc = nuc
        consensus.append(max_nuc)
    
    return ''.join(consensus)

def consensus_and_profile(fasta_text: str) -> tuple:
    """Build consensus and profile from FASTA input."""
    sequences = list(parse_fasta(fasta_text).values())
    profile = build_profile(sequences)
    consensus = get_consensus(profile)
    return consensus, profile

# Test
test_fasta = """>Rosalind_1
ATCCAGCT
>Rosalind_2
GGGCAACT
>Rosalind_3
ATGGATCT
>Rosalind_4
AAGCAACC
>Rosalind_5
TTGGAACT
>Rosalind_6
ATGCCATT
>Rosalind_7
ATGGCACT"""

consensus, profile = consensus_and_profile(test_fasta)
print(consensus)
for nuc in 'ACGT':
    print(f"{nuc}: {' '.join(map(str, profile[nuc]))}")
```

## Summary

Key patterns used in these solutions:

1. **String operations**: `count()`, `replace()`, `translate()`, `find()`
2. **List comprehensions**: Efficient sequence transformations
3. **Dictionary lookups**: Codon table, mass table
4. **Slicing**: `[::-1]` for reversal, `[i:i+3]` for codons
5. **FASTA parsing**: Line-by-line with header detection
6. **Dynamic programming**: For efficient string matching
