## Problem 1: Codon Extraction

```python
def extract_codons_all_frames(sequence):
    """
    Extract codons from all three reading frames.
    """
    result = {}
    
    for frame in range(3):
        codons = []
        # Start from frame position, step by 3
        for i in range(frame, len(sequence) - 2, 3):
            codons.append(sequence[i:i+3])
        result[frame] = codons
    
    return result

# Alternative with comprehension:
def extract_codons_all_frames_v2(sequence):
    return {
        frame: [sequence[i:i+3] for i in range(frame, len(sequence) - 2, 3)]
        for frame in range(3)
    }

# Test
test_seq = "ATGGCCGATCGA"
print(extract_codons_all_frames(test_seq))
print(extract_codons_all_frames_v2(test_seq))
```

## Problem 2: Genetic Code Dictionary

```python
CODON_TABLE = {
    'TTT': 'F', 'TTC': 'F', 'TTA': 'L', 'TTG': 'L',
    'TCT': 'S', 'TCC': 'S', 'TCA': 'S', 'TCG': 'S',
    'TAT': 'Y', 'TAC': 'Y', 'TAA': '*', 'TAG': '*',
    'TGT': 'C', 'TGC': 'C', 'TGA': '*', 'TGG': 'W',
    'CTT': 'L', 'CTC': 'L', 'CTA': 'L', 'CTG': 'L',
    'CCT': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'CAT': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'CGT': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'ATT': 'I', 'ATC': 'I', 'ATA': 'I', 'ATG': 'M',
    'ACT': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'AAT': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
    'AGT': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
    'GTT': 'V', 'GTC': 'V', 'GTA': 'V', 'GTG': 'V',
    'GCT': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'GAT': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGT': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}

def translate_dna(sequence):
    """
    Translate DNA sequence to protein.
    Stop translation at stop codon.
    """
    protein = []
    
    for i in range(0, len(sequence) - 2, 3):
        codon = sequence[i:i+3]
        amino_acid = CODON_TABLE.get(codon, 'X')  # X for unknown
        
        if amino_acid == '*':  # Stop codon
            break
        protein.append(amino_acid)
    
    return ''.join(protein)

# Test
print(translate_dna("ATGGCCGATTAA"))  # MAD
print(translate_dna("ATGGCCGATTGA"))  # MAD (TGA is stop)
```

## Problem 3: K-mer Counter

```python
from collections import Counter

def count_kmers(sequence, k):
    """
    Count all k-mers in sequence.
    Returns list of tuples sorted by count descending.
    """
    # Extract all k-mers
    kmers = [sequence[i:i+k] for i in range(len(sequence) - k + 1)]
    
    # Count using Counter (or dictionary)
    kmer_counts = Counter(kmers)
    
    # Sort by count descending
    sorted_kmers = sorted(kmer_counts.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_kmers

# Manual version without Counter:
def count_kmers_manual(sequence, k):
    counts = {}
    for i in range(len(sequence) - k + 1):
        kmer = sequence[i:i+k]
        counts[kmer] = counts.get(kmer, 0) + 1
    
    return sorted(counts.items(), key=lambda x: -x[1])

# Test
print(count_kmers("ATGATGATG", 3))
```

## Problem 4: Set Operations for Sequences

```python
def compare_protein_sets(sample1, sample2):
    """
    Compare two sets of proteins.
    """
    return {
        'shared': sample1 & sample2,           # Intersection
        'unique_to_1': sample1 - sample2,      # Difference
        'unique_to_2': sample2 - sample1,      # Difference
        'all': sample1 | sample2               # Union
    }

# Test
s1 = {"ProteinA", "ProteinB", "ProteinC"}
s2 = {"ProteinB", "ProteinC", "ProteinD"}
result = compare_protein_sets(s1, s2)
print(result)
```

## Problem 5: Comprehension Challenge

```python
# 5a: Find all ATG positions
def find_atg_positions(sequence):
    """Return list of positions where 'ATG' starts"""
    return [i for i in range(len(sequence) - 2) if sequence[i:i+3] == 'ATG']

# Test
print(find_atg_positions("ATGCCATGGATG"))  # [0, 5, 9]
```

```python
# 5b: Reverse codon table
def reverse_codon_table(codon_table):
    """
    Create reverse mapping: amino acid -> list of codons.
    """
    reverse = {}
    for codon, aa in codon_table.items():
        if aa not in reverse:
            reverse[aa] = []
        reverse[aa].append(codon)
    return reverse

# Using defaultdict:
from collections import defaultdict
def reverse_codon_table_v2(codon_table):
    reverse = defaultdict(list)
    for codon, aa in codon_table.items():
        reverse[aa].append(codon)
    return dict(reverse)

# Test
reverse_table = reverse_codon_table(CODON_TABLE)
print(f"Leucine (L): {reverse_table.get('L')}")
```

```python
# 5c: Filter by GC content
def gc_content(seq):
    """Calculate GC content as percentage"""
    gc = seq.count('G') + seq.count('C')
    return (gc / len(seq)) * 100

def filter_high_gc(sequences, threshold=50):
    """Filter sequences with GC content above threshold."""
    return [seq for seq in sequences if gc_content(seq) > threshold]

# Test
seqs = ["ATATATAT", "GCGCGCGC", "ATGCATGC", "AAAAAAA"]
high_gc = filter_high_gc(seqs, 50)
print(high_gc)  # ['GCGCGCGC', 'ATGCATGC'] - ATGCATGC is exactly 50%, so depends on > vs >=
```

## Problem 6: Nucleotide Pair Counter

```python
def count_dinucleotides(sequence):
    """
    Count all dinucleotide pairs.
    """
    nucleotides = ['A', 'T', 'G', 'C']
    
    # Initialize counts
    counts = {n1: {n2: 0 for n2 in nucleotides} for n1 in nucleotides}
    
    # Count pairs
    for i in range(len(sequence) - 1):
        n1, n2 = sequence[i], sequence[i+1]
        if n1 in counts and n2 in counts[n1]:
            counts[n1][n2] += 1
    
    return counts

def print_dinucleotide_table(counts):
    """Print the dinucleotide counts as a formatted table"""
    nucleotides = ['A', 'T', 'G', 'C']
    
    # Header
    print("   " + "  ".join(nucleotides))
    
    # Rows
    for n1 in nucleotides:
        row = f"{n1} "
        row += "  ".join(str(counts[n1][n2]) for n2 in nucleotides)
        print(row)

# Test
seq = "ATGCATGC"
counts = count_dinucleotides(seq)
print_dinucleotide_table(counts)
```

## Problem 7: Codon Usage Bias

```python
def calculate_codon_usage(sequence, codon_table):
    """
    Calculate codon usage statistics.
    """
    usage = defaultdict(lambda: defaultdict(int))
    
    # Count codons
    for i in range(0, len(sequence) - 2, 3):
        codon = sequence[i:i+3]
        if codon in codon_table:
            aa = codon_table[codon]
            usage[aa][codon] += 1
    
    # Convert to regular dict
    return {aa: dict(codons) for aa, codons in usage.items()}

# Test
test_seq = "ATGTTTTCCCTCGCGAAATGA"
usage = calculate_codon_usage(test_seq, CODON_TABLE)
print(usage)
```

## Problem 8: FASTA Parser with Dictionary

```python
def parse_fasta_string(fasta_text):
    """
    Parse FASTA-formatted string to dictionary.
    """
    sequences = {}
    current_id = None
    current_seq = []
    
    for line in fasta_text.strip().split('\n'):
        line = line.strip()
        if line.startswith('>'):
            # Save previous sequence if exists
            if current_id is not None:
                sequences[current_id] = ''.join(current_seq)
            # Start new sequence
            current_id = line[1:].split()[0]  # Get ID (first word after >)
            current_seq = []
        else:
            current_seq.append(line)
    
    # Don't forget last sequence
    if current_id is not None:
        sequences[current_id] = ''.join(current_seq)
    
    return sequences

# Test
fasta = """>gene1 Human insulin
ATGGCCCTG
TGGATGCGA
>gene2 Mouse hemoglobin
ATGGTGCAC
"""

sequences = parse_fasta_string(fasta)
print(sequences)
```

## Problem 9: Find Shared K-mers

```python
def find_shared_kmers(seq1, seq2, k):
    """
    Find k-mers shared between two sequences.
    """
    def get_kmer_positions(seq, k):
        """Return dictionary of kmer -> list of positions"""
        positions = defaultdict(list)
        for i in range(len(seq) - k + 1):
            kmer = seq[i:i+k]
            positions[kmer].append(i)
        return positions
    
    # Get positions for both sequences
    pos1 = get_kmer_positions(seq1, k)
    pos2 = get_kmer_positions(seq2, k)
    
    # Find shared k-mers
    shared = {}
    for kmer in pos1:
        if kmer in pos2:
            shared[kmer] = {
                'seq1': pos1[kmer],
                'seq2': pos2[kmer]
            }
    
    return shared

# Test
result = find_shared_kmers("ATGCATGC", "GCATGCAT", 4)
for kmer, positions in result.items():
    print(f"{kmer}: {positions}")
```

## Problem 10: Gene Expression Matrix

```python
def highly_expressed_genes(expression_data, sample, threshold):
    """
    Find genes with expression above threshold in a specific sample.
    """
    return [(gene, expression_data[gene][sample]) 
            for gene in expression_data 
            if expression_data[gene].get(sample, 0) > threshold]

def differential_genes(expression_data, sample1, sample2, fold_change=2):
    """
    Find genes with >= fold_change difference between samples.
    """
    results = {}
    
    for gene, samples in expression_data.items():
        expr1 = samples.get(sample1, 1)  # Use 1 to avoid division by zero
        expr2 = samples.get(sample2, 1)
        
        # Calculate ratio
        if expr1 > expr2:
            ratio = expr1 / expr2 if expr2 > 0 else float('inf')
            higher_in = sample1
        else:
            ratio = expr2 / expr1 if expr1 > 0 else float('inf')
            higher_in = sample2
        
        if ratio >= fold_change:
            results[gene] = {'ratio': round(ratio, 2), 'higher_in': higher_in}
    
    return results

# Test
expr = {
    'GeneA': {'Sample1': 100, 'Sample2': 50},
    'GeneB': {'Sample1': 20, 'Sample2': 200},
    'GeneC': {'Sample1': 5, 'Sample2': 5}
}

print("Highly expressed in Sample1:")
print(highly_expressed_genes(expr, 'Sample1', 50))

print("\nDifferentially expressed (2-fold):")
print(differential_genes(expr, 'Sample1', 'Sample2', 2))
```
