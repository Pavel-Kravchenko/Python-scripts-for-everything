## Exercise 1: Extract FASTA Names

```python
def extract_fasta_names(filename):
    """Extract sequence names from FASTA file."""
    names = []
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith('>'):
                names.append(line[1:].strip())
    return names

# Alternative using list comprehension
def extract_fasta_names_v2(filename):
    with open(filename, 'r') as f:
        return [line[1:].strip() for line in f if line.startswith('>')]
```

## Exercise 2: GC Content Calculator

```python
def calculate_gc_content(sequence):
    """Calculate GC content percentage."""
    seq = sequence.upper()
    gc_count = seq.count('G') + seq.count('C')
    total = len(seq)
    return (gc_count / total * 100) if total > 0 else 0

def gc_from_fasta(filename):
    """Calculate GC content from FASTA file."""
    sequence = ''
    with open(filename, 'r') as f:
        for line in f:
            if not line.startswith('>'):
                sequence += line.strip()
    return calculate_gc_content(sequence)

# Test
print(f"GC content of ATGCGATCGA: {calculate_gc_content('ATGCGATCGA'):.1f}%")
```

## Exercise 3: Count CDS Features

```python
from Bio import SeqIO

def count_cds_features(genbank_file):
    """Count CDS features in GenBank file."""
    cds_count = 0
    for record in SeqIO.parse(genbank_file, 'genbank'):
        for feature in record.features:
            if feature.type == 'CDS':
                cds_count += 1
    return cds_count

# Without BioPython (manual parsing)
def count_cds_manual(genbank_file):
    count = 0
    with open(genbank_file, 'r') as f:
        for line in f:
            if line.strip().startswith('CDS'):
                count += 1
    return count
```

## Exercise 4: Purine/Pyrimidine Classification

```python
def classify_nucleotides(sequence):
    """
    Classify nucleotides as purines (A, G) or pyrimidines (C, T).
    """
    classification = {'A': 'R', 'G': 'R', 'C': 'Y', 'T': 'Y', 'U': 'Y'}
    return ''.join(classification.get(n.upper(), '?') for n in sequence)

def purine_pyrimidine_ratio(sequence):
    """Calculate purine to pyrimidine ratio."""
    seq = sequence.upper()
    purines = seq.count('A') + seq.count('G')
    pyrimidines = seq.count('C') + seq.count('T') + seq.count('U')
    return purines / pyrimidines if pyrimidines > 0 else float('inf')

# Test
test_seq = "ATGCGATCGA"
print(f"Sequence: {test_seq}")
print(f"Classification: {classify_nucleotides(test_seq)}")
print(f"Pu/Py ratio: {purine_pyrimidine_ratio(test_seq):.2f}")
```

## Exercise 5: FASTA Parser (Complete)

```python
def parse_fasta(filename):
    """Complete FASTA parser."""
    sequences = []
    current_name = None
    current_seq = []
    
    with open(filename, 'r') as f:
        for line in f:
            line = line.strip()
            if line.startswith('>'):
                if current_name is not None:
                    sequences.append((current_name, ''.join(current_seq)))
                current_name = line[1:]
                current_seq = []
            else:
                current_seq.append(line)
        
        if current_name is not None:
            sequences.append((current_name, ''.join(current_seq)))
    
    return sequences
```
