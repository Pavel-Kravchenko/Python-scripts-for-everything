## Exercise 3: Circle Calculations

```python
def circle_area(radius):
    """Calculate area of circle."""
    return math.pi * radius ** 2

def circle_circumference(radius):
    """Calculate circumference of circle."""
    return 2 * math.pi * radius

# Test
r = 5
print(f"Circle with radius {r}:")
print(f"  Area = {circle_area(r):.2f}")
print(f"  Circumference = {circle_circumference(r):.2f}")
```

## Exercise 4: Temperature Conversion

```python
def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit."""
    return celsius * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit):
    """Convert Fahrenheit to Celsius."""
    return (fahrenheit - 32) * 5/9

def celsius_to_kelvin(celsius):
    """Convert Celsius to Kelvin."""
    return celsius + 273.15

# Test - relevant for molecular biology (PCR temperatures)
temps_c = [4, 25, 37, 72, 95]  # Common lab temperatures
for c in temps_c:
    print(f"{c}°C = {celsius_to_fahrenheit(c):.1f}°F = {celsius_to_kelvin(c):.2f}K")
```

## Exercise 5: Loops and Ranges

```python
# Print numbers 1-10
for i in range(1, 11):
    print(i, end=' ')
print()

# Sum of first n numbers
def sum_to_n(n):
    return sum(range(1, n + 1))

print(f"Sum 1-100 = {sum_to_n(100)}")

# Factorial
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(f"5! = {factorial(5)}")
print(f"10! = {factorial(10)}")
```

## Exercise 6: Conditionals

```python
def check_nucleotide(base):
    """Check if character is a valid nucleotide."""
    base = base.upper()
    if base in 'ATGC':
        return f"{base} is a valid DNA nucleotide"
    elif base == 'U':
        return f"{base} is an RNA nucleotide"
    else:
        return f"{base} is not a nucleotide"

def classify_sequence(seq):
    """Classify sequence as DNA, RNA, or protein."""
    seq = seq.upper()
    if all(c in 'ATGC' for c in seq):
        return 'DNA'
    elif all(c in 'AUGC' for c in seq) and 'U' in seq:
        return 'RNA'
    elif all(c in 'ACDEFGHIKLMNPQRSTVWY' for c in seq):
        return 'Protein'
    else:
        return 'Unknown'

# Test
test_seqs = ['ATGCGATCGA', 'AUGCGAUCGA', 'MVLSPADKTN']
for seq in test_seqs:
    print(f"{seq[:10]}... is {classify_sequence(seq)}")
```
