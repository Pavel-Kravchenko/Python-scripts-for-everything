## Exercise 1: Fibonacci Sequence

```python
def fib_recursive(n):
    """Basic recursive Fibonacci (slow for large n)."""
    if n <= 1:
        return n
    return fib_recursive(n - 1) + fib_recursive(n - 2)

@lru_cache(maxsize=None)
def fib_memoized(n):
    """Efficient Fibonacci with memoization."""
    if n <= 1:
        return n
    return fib_memoized(n - 1) + fib_memoized(n - 2)

def fib_iterative(n):
    """Iterative Fibonacci (most efficient)."""
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Test
print("Fibonacci sequence (first 15):")
print([fib_iterative(i) for i in range(15)])
```

## Exercise 2: Random DNA Generator

```python
def random_dna(length, gc_content=0.5):
    """Generate random DNA sequence with specified GC content."""
    nucleotides = []
    for _ in range(length):
        if random.random() < gc_content:
            nucleotides.append(random.choice('GC'))
        else:
            nucleotides.append(random.choice('AT'))
    return ''.join(nucleotides)

def random_dna_uniform(length):
    """Generate random DNA with uniform nucleotide distribution."""
    return ''.join(random.choice('ATGC') for _ in range(length))

# Test
print(f"Random DNA (50bp, 50% GC): {random_dna(50)}")
print(f"Random DNA (50bp, 70% GC): {random_dna(50, gc_content=0.7)}")
```

## Exercise 3: DNA Palindrome Finder

```python
def reverse_complement(seq):
    """Calculate reverse complement of DNA sequence."""
    complement = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    return ''.join(complement[n] for n in reversed(seq.upper()))

def is_palindrome(seq):
    """Check if DNA sequence is palindromic (equals its reverse complement)."""
    return seq.upper() == reverse_complement(seq)

def find_palindromes(seq, min_len=4, max_len=12):
    """Find all palindromic sequences (restriction sites)."""
    palindromes = []
    seq = seq.upper()
    
    for length in range(min_len, max_len + 1, 2):  # Even lengths only
        for i in range(len(seq) - length + 1):
            subseq = seq[i:i + length]
            if is_palindrome(subseq):
                palindromes.append((i, subseq))
    
    return palindromes

# Test
test_seq = "ATGAATTCGATGCATGC"
print(f"Sequence: {test_seq}")
print(f"Palindromes found: {find_palindromes(test_seq)}")
```

## Exercise 4: Trigonometry Module

```python
def degrees_to_radians(degrees):
    """Convert degrees to radians."""
    return degrees * math.pi / 180

def radians_to_degrees(radians):
    """Convert radians to degrees."""
    return radians * 180 / math.pi

def sine_table(start=0, end=360, step=30):
    """Generate sine table for angles."""
    return {
        angle: math.sin(degrees_to_radians(angle))
        for angle in range(start, end + 1, step)
    }

# Test
print("Sine table:")
for angle, value in sine_table().items():
    print(f"  sin({angle}°) = {value:.4f}")
```

## Exercise 5: Prime Numbers

```python
def is_prime(n):
    """Check if number is prime."""
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

def primes_up_to(n):
    """Generate all primes up to n using Sieve of Eratosthenes."""
    sieve = [True] * (n + 1)
    sieve[0] = sieve[1] = False
    
    for i in range(2, int(math.sqrt(n)) + 1):
        if sieve[i]:
            for j in range(i*i, n + 1, i):
                sieve[j] = False
    
    return [i for i, is_p in enumerate(sieve) if is_p]

# Test
print(f"Primes up to 50: {primes_up_to(50)}")
```
