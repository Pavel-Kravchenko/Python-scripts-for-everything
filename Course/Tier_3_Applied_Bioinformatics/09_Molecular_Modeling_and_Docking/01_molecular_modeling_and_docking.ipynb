# Molecular Modeling and Docking: Module Overview

**Tier 3 — Module 09**

This module covers computational structural biology: force fields, molecular mechanics, homology modeling, AlphaFold, molecular docking, molecular dynamics, and chemoinformatics.

## Notebooks in This Module

1. **[01_molecular_modeling.ipynb](./01_molecular_modeling.ipynb)** — Complete notebook covering all topics
   - Force fields, energy minimization, homology modeling, docking (AutoDock Vina), MD (GROMACS), chemoinformatics (RDKit)
   - This is the primary reference notebook for the module

2. **[02_docking.ipynb](./02_docking.ipynb)** — Same curriculum with emphasis on docking and drug discovery
   - Contains the same content as 01_molecular_modeling.ipynb
   - Use for focused study of docking workflows (Sections 4–6)
