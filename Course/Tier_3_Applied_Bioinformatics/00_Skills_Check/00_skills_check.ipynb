### Question 2: FASTQ Format
- **a)** Phred = 73 - 33 = **40** (Q40 = 99.99% accuracy)
- **b)** P(error) = 10^(-20/10) = **0.01 (1%)**
- **c)** Positions 10-12 (characters `!`, `#`, `#`) have the lowest quality. `!` = Q0, `#` = Q2.

---

### Question 3: SAM/BAM Format
- **a)** FLAG 99 = 64 + 32 + 2 + 1 = paired (1) + properly paired (2) + mate reverse strand (32) + first in pair (64). This is a properly-paired forward read, first in pair, with mate on reverse strand.
- **b)** 75M2I23M = 75 bases match/mismatch, then 2-base insertion in the read, then 23 bases match/mismatch. Total read length = 100 bp, alignment spans 98 reference bases.
- **c)** MAPQ 60 indicates very high mapping confidence. The read maps uniquely with high certainty.

---

### Question 4: Quality Control
- **a)** Use **Trimmomatic** or **fastp** to trim low-quality bases from read ends and remove adapter sequences.
- **b)** Bimodal GC distribution suggests **contamination** from another organism (e.g., adapter dimers, host contamination in metagenomics, or a second species mixed in).

---

### Question 5: VCF Format
- **a)** **SNP** (single nucleotide polymorphism): single-base change G to A.
- **b)** Genotype `0/1` = **heterozygous** (one reference allele G, one alternate allele A).
- **c)** INFO DP=120 is the **total read depth at this site across all samples**. FORMAT DP=55 is the **read depth for SAMPLE1 specifically**. They differ because INFO DP sums all samples (or includes filtered reads), while FORMAT DP is sample-specific.

---

### Question 6: Variant Types
- A->G: **SNP** (single nucleotide polymorphism)
- ACG->A: **Deletion** (2 bp deleted)
- T->TGCA: **Insertion** (3 bp inserted)
- Large region duplicated: **CNV** (copy number variation) or structural variant
- 500 bp deleted: **Structural variant** (large deletion / SV)

### Question 9: RNA-seq Pipeline

- **c)** **Upper right** corner of the volcano plot (high positive log2FC on x-axis, high -log10(p-value) on y-axis).

### Question 12: Population Genetics
- **a)** Under HWE (p=0.7, q=0.3): AA = p^2 = **0.49**, AG = 2pq = **0.42**, GG = q^2 = **0.09**
- **b)** Expected in 100: 49 AA, 42 AG, 9 GG. Observed: 50 AA, 40 AG, 10 GG. These are **close to HWE** -- the small deviations could be due to sampling. A chi-squared test would confirm.
- **c)** Causes of HWE deviation: **natural selection**, non-random mating (inbreeding/assortative mating), genetic drift (small population), mutation, or migration/gene flow.

````python
# Optional: quick HWE calculation to verify Question 12
p = 0.7  # frequency of A allele
q = 0.3  # frequency of G allele
n = 100  # sample size

# Expected under HWE
exp_AA = p**2 * n
exp_AG = 2 * p * q * n
exp_GG = q**2 * n

print("Hardy-Weinberg Expected Genotype Frequencies")
print(f"  p(A) = {p}, q(G) = {q}")
print(f"  AA: p^2 = {p**2:.2f} -> {exp_AA:.1f} individuals")
print(f"  AG: 2pq = {2*p*q:.2f} -> {exp_AG:.1f} individuals")
print(f"  GG: q^2 = {q**2:.2f} -> {exp_GG:.1f} individuals")

# Chi-squared test
obs = [50, 40, 10]
exp = [exp_AA, exp_AG, exp_GG]
chi2 = sum((o - e)**2 / e for o, e in zip(obs, exp))
print(f"\nObserved: {obs}")
print(f"Expected: [{exp_AA:.1f}, {exp_AG:.1f}, {exp_GG:.1f}]")
print(f"Chi-squared = {chi2:.3f} (df=1, critical value at p=0.05 is 3.841)")
print(f"Result: {'Reject HWE' if chi2 > 3.841 else 'Cannot reject HWE (consistent with equilibrium)'}")
````
