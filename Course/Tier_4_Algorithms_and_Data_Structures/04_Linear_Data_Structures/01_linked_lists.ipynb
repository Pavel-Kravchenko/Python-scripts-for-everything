## 1. Introduction to Linked Lists

### What is a Linked List?

A **linked list** is a linear data structure where elements are stored in nodes, and each node points to the next node in the sequence. Unlike arrays, linked list elements are not stored in contiguous memory locations.

### Why Use Linked Lists?

| Advantage | Description |
|-----------|-------------|
| **Dynamic Size** | Can grow or shrink during runtime |
| **Efficient Insertion/Deletion** | O(1) at known positions (no shifting required) |
| **Memory Efficiency** | Allocates memory as needed |
| **No Wasted Space** | No pre-allocated unused slots |

### Comparison: Linked Lists vs Arrays

| Feature | Array | Linked List |
|---------|-------|-------------|
| **Memory Layout** | Contiguous | Non-contiguous |
| **Size** | Fixed (static) or costly resize | Dynamic |
| **Random Access** | O(1) | O(n) |
| **Insertion at Beginning** | O(n) | O(1) |
| **Memory Overhead** | None | Extra pointer(s) per node |
| **Cache Performance** | Excellent (locality) | Poor (scattered) |

### Memory Representation

```
Array (contiguous memory):
┌─────┬─────┬─────┬─────┬─────┐
│  5  │  3  │  8  │  1  │  9  │   Memory addresses: 100, 104, 108, 112, 116
└─────┴─────┴─────┴─────┴─────┘

Linked List (scattered memory):
┌─────┐     ┌─────┐     ┌─────┐     ┌─────┐
│  5  │────→│  3  │────→│  8  │────→│  1  │────→ null
└─────┘     └─────┘     └─────┘     └─────┘
addr:200    addr:508    addr:124    addr:336
```

## 2. Singly Linked List

### Theory

A **singly linked list** is the simplest form of linked list where each node contains:
- **Data**: The value stored in the node
- **Next**: A pointer/reference to the next node

The list maintains a **head** pointer to the first node. The last node's `next` pointer is `null`.

### Visual Structure

```
head
  ↓
┌───┬───┐   ┌───┬───┐   ┌───┬───┐   ┌───┬───┐
│ 5 │ ●─┼──→│ 3 │ ●─┼──→│ 8 │ ●─┼──→│ 1 │ ○ │
└───┴───┘   └───┴───┘   └───┴───┘   └───┴───┘
 data next   data next   data next   data null
```

### Time Complexity

| Operation | Time Complexity | Notes |
|-----------|-----------------|-------|
| Access by index | O(n) | Must traverse from head |
| Insert at head | O(1) | Direct pointer update |
| Insert at tail | O(n) or O(1)* | *O(1) if tail pointer maintained |
| Insert at position | O(n) | Must find position first |
| Delete at head | O(1) | Direct pointer update |
| Delete at tail | O(n) | Must find second-to-last |
| Delete by value | O(n) | Must search for value |
| Search | O(n) | Linear traversal |
| Get length | O(1) or O(n) | O(1) if length cached |

### Implementation

````python
from typing import Optional, Any, Iterator


class Node:
    """
    A node in a singly linked list.
    
    Attributes:
        data: The value stored in this node
        next: Reference to the next node (None if this is the last node)
    """
    
    def __init__(self, data: Any, next_node: Optional['Node'] = None) -> None:
        self.data = data
        self.next = next_node
    
    def __repr__(self) -> str:
        return f"Node({self.data})"


class SinglyLinkedList:
    """
    A singly linked list implementation with head and tail pointers.
    
    Maintains both head and tail pointers for O(1) insertion at both ends.
    Also caches length for O(1) size queries.
    
    Attributes:
        head: Reference to the first node
        tail: Reference to the last node
        _length: Number of nodes in the list
    """
    
    def __init__(self) -> None:
        """Initialize an empty linked list."""
        self.head: Optional[Node] = None
        self.tail: Optional[Node] = None
        self._length: int = 0
    
    def __len__(self) -> int:
        """Return the number of nodes in the list. O(1)"""
        return self._length
    
    def __iter__(self) -> Iterator[Any]:
        """Iterate over values in the list."""
        current = self.head
        while current is not None:
            yield current.data
            current = current.next
    
    def __str__(self) -> str:
        """Return string representation: [5] -> [3] -> [8] -> null"""
        if self.head is None:
            return "empty list"
        
        nodes = []
        current = self.head
        while current is not None:
            nodes.append(f"[{current.data}]")
            current = current.next
        nodes.append("null")
        return " -> ".join(nodes)
    
    def is_empty(self) -> bool:
        """Check if the list is empty. O(1)"""
        return self.head is None
    
    # ==================== INSERTION OPERATIONS ====================
    
    def add_at_head(self, data: Any) -> None:
        """
        Insert a new node at the beginning of the list. O(1)
        
        Visual:
            Before: head -> [3] -> [8] -> null
            After:  head -> [5] -> [3] -> [8] -> null
        """
        new_node = Node(data, self.head)
        self.head = new_node
        
        # If list was empty, new node is also the tail
        if self.tail is None:
            self.tail = new_node
        
        self._length += 1
    
    def add_at_tail(self, data: Any) -> None:
        """
        Insert a new node at the end of the list. O(1) with tail pointer.
        
        Visual:
            Before: head -> [5] -> [3] -> null
            After:  head -> [5] -> [3] -> [8] -> null
        """
        new_node = Node(data)
        
        if self.tail is None:
            # List is empty
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        
        self._length += 1
    
    def add(self, data: Any) -> None:
        """Alias for add_at_tail - adds element to the end. O(1)"""
        self.add_at_tail(data)
    
    def insert_at_index(self, index: int, data: Any) -> bool:
        """
        Insert a new node at the specified index. O(n)
        
        Args:
            index: Position to insert (0-based)
            data: Value to insert
            
        Returns:
            True if successful, False if index out of bounds
        """
        if index < 0 or index > self._length:
            return False
        
        if index == 0:
            self.add_at_head(data)
            return True
        
        if index == self._length:
            self.add_at_tail(data)
            return True
        
        # Find node at position (index - 1)
        current = self.head
        for _ in range(index - 1):
            current = current.next
        
        # Insert new node after current
        new_node = Node(data, current.next)
        current.next = new_node
        self._length += 1
        return True
    
    # ==================== DELETION OPERATIONS ====================
    
    def delete_at_head(self) -> Optional[Any]:
        """
        Remove and return the first node's data. O(1)
        
        Visual:
            Before: head -> [5] -> [3] -> [8] -> null
            After:  head -> [3] -> [8] -> null
            Returns: 5
        """
        if self.head is None:
            return None
        
        data = self.head.data
        self.head = self.head.next
        
        # If list becomes empty, update tail
        if self.head is None:
            self.tail = None
        
        self._length -= 1
        return data
    
    def delete_at_tail(self) -> Optional[Any]:
        """
        Remove and return the last node's data. O(n)
        
        Note: O(n) because we need to find the second-to-last node
        to update its next pointer.
        """
        if self.head is None:
            return None
        
        # Single node case
        if self.head == self.tail:
            data = self.head.data
            self.head = self.tail = None
            self._length -= 1
            return data
        
        # Find second-to-last node
        current = self.head
        while current.next != self.tail:
            current = current.next
        
        data = self.tail.data
        current.next = None
        self.tail = current
        self._length -= 1
        return data
    
    def delete_by_value(self, data: Any) -> bool:
        """
        Remove the first node with the specified value. O(n)
        
        Visual:
            Delete 3:
            Before: [5] -> [3] -> [8] -> null
            After:  [5] ---------> [8] -> null
        
        Returns:
            True if node was found and deleted, False otherwise
        """
        if self.head is None:
            return False
        
        # Special case: deleting head
        if self.head.data == data:
            self.delete_at_head()
            return True
        
        # Find the node before the one to delete
        current = self.head
        while current.next is not None and current.next.data != data:
            current = current.next
        
        # Node not found
        if current.next is None:
            return False
        
        # Delete the node
        if current.next == self.tail:
            self.tail = current
        
        current.next = current.next.next
        self._length -= 1
        return True
    
    def delete_at_index(self, index: int) -> Optional[Any]:
        """
        Remove and return the node at the specified index. O(n)
        
        Returns:
            The deleted value, or None if index out of bounds
        """
        if index < 0 or index >= self._length:
            return None
        
        if index == 0:
            return self.delete_at_head()
        
        # Find node at position (index - 1)
        current = self.head
        for _ in range(index - 1):
            current = current.next
        
        data = current.next.data
        
        # Update tail if needed
        if current.next == self.tail:
            self.tail = current
        
        current.next = current.next.next
        self._length -= 1
        return data
    
    # ==================== SEARCH OPERATIONS ====================
    
    def search(self, data: Any) -> int:
        """
        Find the index of the first node with the specified value. O(n)
        
        Returns:
            Index of the node (0-based), or -1 if not found
        """
        current = self.head
        index = 0
        
        while current is not None:
            if current.data == data:
                return index
            current = current.next
            index += 1
        
        return -1
    
    def contains(self, data: Any) -> bool:
        """Check if the list contains the specified value. O(n)"""
        return self.search(data) != -1
    
    def get_at_index(self, index: int) -> Optional[Any]:
        """
        Get the value at the specified index. O(n)
        
        Returns:
            The value at the index, or None if index out of bounds
        """
        if index < 0 or index >= self._length:
            return None
        
        current = self.head
        for _ in range(index):
            current = current.next
        
        return current.data
    
    # ==================== UTILITY OPERATIONS ====================
    
    def reverse(self) -> None:
        """
        Reverse the linked list in place. O(n)
        
        Algorithm:
            Before: [1] -> [2] -> [3] -> null
            
            Step 1: null <- [1]    [2] -> [3] -> null
            Step 2: null <- [1] <- [2]    [3] -> null  
            Step 3: null <- [1] <- [2] <- [3]
            
            After:  [3] -> [2] -> [1] -> null
        """
        # Update tail to current head before reversing
        self.tail = self.head
        
        prev = None
        current = self.head
        
        while current is not None:
            next_node = current.next  # Save next
            current.next = prev       # Reverse pointer
            prev = current            # Move prev forward
            current = next_node       # Move current forward
        
        self.head = prev
    
    def to_list(self) -> list:
        """Convert linked list to Python list. O(n)"""
        return list(self)
    
    def clear(self) -> None:
        """Remove all nodes from the list. O(1)"""
        self.head = None
        self.tail = None
        self._length = 0
````

### Examples

````python
# Create and demonstrate singly linked list operations
print("=" * 60)
print("SINGLY LINKED LIST DEMONSTRATION")
print("=" * 60)

# Create a new list
sll = SinglyLinkedList()
print(f"\n1. Created empty list: {sll}")
print(f"   Length: {len(sll)}")
````

````python
# Add elements at tail
print("\n2. Adding elements at tail:")
for value in [5, 3, 8, 1]:
    sll.add(value)
    print(f"   Added {value}: {sll}")
````

````python
# Insert at specific index
print("\n4. Inserting at index 2:")
sll.insert_at_index(2, 99)
print(f"   Inserted 99 at index 2: {sll}")
````

````python
# Search operations
print("\n5. Search operations:")
print(f"   Index of 8: {sll.search(8)}")
print(f"   Index of 100 (not present): {sll.search(100)}")
print(f"   Contains 3: {sll.contains(3)}")
print(f"   Value at index 2: {sll.get_at_index(2)}")
````

````python
# Delete operations
print("\n6. Delete operations:")
print(f"   Current list: {sll}")

deleted = sll.delete_at_head()
print(f"   Deleted head ({deleted}): {sll}")

sll.delete_by_value(99)
print(f"   Deleted value 99: {sll}")

deleted = sll.delete_at_index(2)
print(f"   Deleted at index 2 ({deleted}): {sll}")
````

````python
# Reverse the list
print("\n7. Reverse operation:")
print(f"   Before reverse: {sll}")
sll.reverse()
print(f"   After reverse:  {sll}")
````

````python
# Iteration and conversion
print("\n8. Iteration and conversion:")
print(f"   As Python list: {sll.to_list()}")
print(f"   Iterating: ", end="")
for value in sll:
    print(value, end=" ")
print()
````

## 3. Doubly Linked List

### Theory

A **doubly linked list** extends the singly linked list by adding a backward reference. Each node contains:
- **Data**: The value stored in the node
- **Next**: A pointer to the next node
- **Prev**: A pointer to the previous node

The list maintains both **head** and **tail** pointers.

### Visual Structure

```
      head                                    tail
        ↓                                       ↓
null ←──┌───┬───┬───┐   ┌───┬───┬───┐   ┌───┬───┬───┐──→ null
        │ ○ │ 5 │ ●─┼←─→│ ● │ 3 │ ●─┼←─→│ ● │ 8 │ ○ │
        └───┴───┴───┘   └───┴───┴───┘   └───┴───┴───┘
        prev data next  prev data next  prev data next
```

### Advantages Over Singly Linked List

| Feature | Singly LL | Doubly LL |
|---------|-----------|----------|
| Backward traversal | ✗ (must restart from head) | ✓ O(1) per step |
| Delete node given pointer | O(n) (need prev) | O(1) |
| Delete at tail | O(n) | O(1) |
| Memory per node | Less (1 pointer) | More (2 pointers) |

### Time Complexity

| Operation | Time Complexity | Notes |
|-----------|-----------------|-------|
| Access by index | O(n) | Can start from head or tail |
| Insert at head | O(1) | Direct pointer update |
| Insert at tail | O(1) | Direct pointer update |
| Insert at position | O(n) | Must find position first |
| Delete at head | O(1) | Direct pointer update |
| Delete at tail | O(1) | Has prev pointer |
| Delete given node | O(1) | Has prev pointer |
| Search | O(n) | Can search from both ends |
| Reverse | O(n) | Swap all prev/next pointers |

### Implementation

````python
class DoublyNode:
    """
    A node in a doubly linked list.
    
    Attributes:
        data: The value stored in this node
        next: Reference to the next node
        prev: Reference to the previous node
    """
    
    def __init__(self, data: Any, 
                 next_node: Optional['DoublyNode'] = None,
                 prev_node: Optional['DoublyNode'] = None) -> None:
        self.data = data
        self.next = next_node
        self.prev = prev_node
    
    def __repr__(self) -> str:
        return f"DoublyNode({self.data})"


class DoublyLinkedList:
    """
    A doubly linked list implementation.
    
    Supports bidirectional traversal and O(1) operations at both ends.
    
    Attributes:
        head: Reference to the first node
        tail: Reference to the last node
        _length: Number of nodes in the list
    """
    
    def __init__(self) -> None:
        """Initialize an empty doubly linked list."""
        self.head: Optional[DoublyNode] = None
        self.tail: Optional[DoublyNode] = None
        self._length: int = 0
    
    def __len__(self) -> int:
        """Return the number of nodes in the list. O(1)"""
        return self._length
    
    def __iter__(self) -> Iterator[Any]:
        """Iterate forward over values in the list."""
        current = self.head
        while current is not None:
            yield current.data
            current = current.next
    
    def __reversed__(self) -> Iterator[Any]:
        """Iterate backward over values in the list."""
        current = self.tail
        while current is not None:
            yield current.data
            current = current.prev
    
    def __str__(self) -> str:
        """Return string representation: null <-> [5] <-> [3] <-> [8] <-> null"""
        if self.head is None:
            return "empty list"
        
        nodes = ["null"]
        current = self.head
        while current is not None:
            nodes.append(f"[{current.data}]")
            current = current.next
        nodes.append("null")
        return " <-> ".join(nodes)
    
    def is_empty(self) -> bool:
        """Check if the list is empty. O(1)"""
        return self.head is None
    
    # ==================== INSERTION OPERATIONS ====================
    
    def add_at_head(self, data: Any) -> None:
        """
        Insert a new node at the beginning of the list. O(1)
        
        Visual:
            Before: null <-> [3] <-> [8] <-> null
            After:  null <-> [5] <-> [3] <-> [8] <-> null
        """
        new_node = DoublyNode(data, next_node=self.head)
        
        if self.head is not None:
            self.head.prev = new_node
        else:
            # List was empty, new node is also tail
            self.tail = new_node
        
        self.head = new_node
        self._length += 1
    
    def add_at_tail(self, data: Any) -> None:
        """
        Insert a new node at the end of the list. O(1)
        
        Visual:
            Before: null <-> [5] <-> [3] <-> null
            After:  null <-> [5] <-> [3] <-> [8] <-> null
        """
        new_node = DoublyNode(data, prev_node=self.tail)
        
        if self.tail is not None:
            self.tail.next = new_node
        else:
            # List was empty, new node is also head
            self.head = new_node
        
        self.tail = new_node
        self._length += 1
    
    def add(self, data: Any) -> None:
        """Alias for add_at_tail - adds element to the end. O(1)"""
        self.add_at_tail(data)
    
    def insert_at_index(self, index: int, data: Any) -> bool:
        """
        Insert a new node at the specified index. O(n)
        
        Optimizes by starting from head or tail depending on index.
        
        Returns:
            True if successful, False if index out of bounds
        """
        if index < 0 or index > self._length:
            return False
        
        if index == 0:
            self.add_at_head(data)
            return True
        
        if index == self._length:
            self.add_at_tail(data)
            return True
        
        # Find the node currently at the index
        # Start from head or tail depending on which is closer
        if index <= self._length // 2:
            current = self.head
            for _ in range(index):
                current = current.next
        else:
            current = self.tail
            for _ in range(self._length - 1 - index):
                current = current.prev
        
        # Insert new node before current
        new_node = DoublyNode(data, next_node=current, prev_node=current.prev)
        current.prev.next = new_node
        current.prev = new_node
        self._length += 1
        return True
    
    # ==================== DELETION OPERATIONS ====================
    
    def delete_at_head(self) -> Optional[Any]:
        """
        Remove and return the first node's data. O(1)
        """
        if self.head is None:
            return None
        
        data = self.head.data
        self.head = self.head.next
        
        if self.head is not None:
            self.head.prev = None
        else:
            # List is now empty
            self.tail = None
        
        self._length -= 1
        return data
    
    def delete_at_tail(self) -> Optional[Any]:
        """
        Remove and return the last node's data. O(1)
        
        Unlike singly linked list, this is O(1) because we have prev pointer.
        """
        if self.tail is None:
            return None
        
        data = self.tail.data
        self.tail = self.tail.prev
        
        if self.tail is not None:
            self.tail.next = None
        else:
            # List is now empty
            self.head = None
        
        self._length -= 1
        return data
    
    def delete_node(self, node: DoublyNode) -> Any:
        """
        Delete a specific node (given a reference to it). O(1)
        
        This is a key advantage of doubly linked lists!
        
        Returns:
            The data from the deleted node
        """
        data = node.data
        
        if node.prev is not None:
            node.prev.next = node.next
        else:
            # Deleting head
            self.head = node.next
        
        if node.next is not None:
            node.next.prev = node.prev
        else:
            # Deleting tail
            self.tail = node.prev
        
        self._length -= 1
        return data
    
    def delete_by_value(self, data: Any) -> bool:
        """
        Remove the first node with the specified value. O(n)
        
        Returns:
            True if node was found and deleted, False otherwise
        """
        current = self.head
        
        while current is not None:
            if current.data == data:
                self.delete_node(current)
                return True
            current = current.next
        
        return False
    
    def delete_at_index(self, index: int) -> Optional[Any]:
        """
        Remove and return the node at the specified index. O(n)
        """
        if index < 0 or index >= self._length:
            return None
        
        if index == 0:
            return self.delete_at_head()
        
        if index == self._length - 1:
            return self.delete_at_tail()
        
        # Find node at index (optimize by starting from closer end)
        if index <= self._length // 2:
            current = self.head
            for _ in range(index):
                current = current.next
        else:
            current = self.tail
            for _ in range(self._length - 1 - index):
                current = current.prev
        
        return self.delete_node(current)
    
    # ==================== SEARCH OPERATIONS ====================
    
    def search(self, data: Any) -> int:
        """
        Find the index of the first node with the specified value. O(n)
        
        Returns:
            Index of the node (0-based), or -1 if not found
        """
        current = self.head
        index = 0
        
        while current is not None:
            if current.data == data:
                return index
            current = current.next
            index += 1
        
        return -1
    
    def contains(self, data: Any) -> bool:
        """Check if the list contains the specified value. O(n)"""
        return self.search(data) != -1
    
    def get_at_index(self, index: int) -> Optional[Any]:
        """
        Get the value at the specified index. O(n)
        
        Optimizes by starting from head or tail.
        """
        if index < 0 or index >= self._length:
            return None
        
        # Start from closer end
        if index <= self._length // 2:
            current = self.head
            for _ in range(index):
                current = current.next
        else:
            current = self.tail
            for _ in range(self._length - 1 - index):
                current = current.prev
        
        return current.data
    
    # ==================== UTILITY OPERATIONS ====================
    
    def reverse(self) -> None:
        """
        Reverse the linked list in place. O(n)
        
        For each node, swap its prev and next pointers.
        Then swap head and tail.
        """
        current = self.head
        
        while current is not None:
            # Swap prev and next
            current.prev, current.next = current.next, current.prev
            # Move to "next" (which is now prev after swap)
            current = current.prev
        
        # Swap head and tail
        self.head, self.tail = self.tail, self.head
    
    def to_list(self) -> list:
        """Convert linked list to Python list. O(n)"""
        return list(self)
    
    def to_list_reversed(self) -> list:
        """Convert linked list to Python list in reverse order. O(n)"""
        return list(reversed(self))
    
    def clear(self) -> None:
        """Remove all nodes from the list. O(1)"""
        self.head = None
        self.tail = None
        self._length = 0
````

### Visualization of Operations

#### Insertion at Head

```
Before: null <-> [3] <-> [8] <-> null
                  ↑
                head

Insert 5:
Step 1: Create new node [5]
Step 2: Set new_node.next = head ([3])
Step 3: Set head.prev = new_node
Step 4: Set head = new_node

After:  null <-> [5] <-> [3] <-> [8] <-> null
                  ↑
                head
```

#### Deletion at Tail (O(1) in Doubly LL!)

```
Before: null <-> [5] <-> [3] <-> [8] <-> null
                                  ↑
                                tail

Delete tail:
Step 1: Save tail.data (8)
Step 2: Set tail = tail.prev ([3])
Step 3: Set new_tail.next = null

After:  null <-> [5] <-> [3] <-> null
                          ↑
                        tail
```

#### Delete Node Given Pointer (O(1) advantage!)

```
Delete node [3] given direct reference:

Before: null <-> [5] <-> [3] <-> [8] <-> null
                          ↑
                    node to delete

Step 1: node.prev.next = node.next
        [5].next = [8]

Step 2: node.next.prev = node.prev
        [8].prev = [5]

After:  null <-> [5] <-> [8] <-> null
```

### 4.1 Detect Cycle in Linked List

**Problem**: Determine if a linked list contains a cycle.

**Solution**: Floyd's Cycle Detection (Tortoise and Hare)
- Use two pointers: slow moves 1 step, fast moves 2 steps
- If they meet, there's a cycle; if fast reaches null, no cycle

```
Cycle visualization:

[1] -> [2] -> [3] -> [4] -> [5]
              ↑              |
              └──────────────┘
```

````python
def has_cycle(head: Optional[Node]) -> bool:
    """
    Detect if a linked list has a cycle using Floyd's algorithm.
    
    Time: O(n), Space: O(1)
    
    Args:
        head: The head node of the linked list
        
    Returns:
        True if cycle exists, False otherwise
    """
    if head is None:
        return False
    
    slow = head  # Moves 1 step at a time
    fast = head  # Moves 2 steps at a time
    
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return True
    
    return False


# Example: Create a list with a cycle
print("Cycle Detection Example:")
print("="*40)

# Create nodes
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

# Link them: 1 -> 2 -> 3 -> 4 -> None (no cycle)
node1.next = node2
node2.next = node3
node3.next = node4

print(f"List without cycle: {has_cycle(node1)}")

# Create cycle: 4 -> 2
node4.next = node2
print(f"List with cycle (4->2): {has_cycle(node1)}")

# Clean up cycle for further use
node4.next = None
````

### 4.2 Find Middle Element

**Problem**: Find the middle node of a linked list in one pass.

**Solution**: Use slow and fast pointers
- When fast reaches end, slow is at middle

```
For odd length:  [1] -> [2] -> [3] -> [4] -> [5] -> null
                              ↑
                           middle (3)

For even length: [1] -> [2] -> [3] -> [4] -> null
                              ↑
                        middle (3, second of two)
```

````python
def find_middle(head: Optional[Node]) -> Optional[Node]:
    """
    Find the middle node of a linked list.
    
    For even-length lists, returns the second middle node.
    
    Time: O(n), Space: O(1)
    """
    if head is None:
        return None
    
    slow = head
    fast = head
    
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
    
    return slow


# Example
print("Find Middle Element:")
print("="*40)

# Create list: 1 -> 2 -> 3 -> 4 -> 5
sll = SinglyLinkedList()
for val in [1, 2, 3, 4, 5]:
    sll.add(val)
print(f"List: {sll}")
middle = find_middle(sll.head)
print(f"Middle element: {middle.data}")

# Even length: 1 -> 2 -> 3 -> 4
sll2 = SinglyLinkedList()
for val in [1, 2, 3, 4]:
    sll2.add(val)
print(f"\nList: {sll2}")
middle = find_middle(sll2.head)
print(f"Middle element: {middle.data}")
````

### 4.3 Merge Two Sorted Lists

**Problem**: Merge two sorted linked lists into one sorted list.

```
List 1: [1] -> [3] -> [5] -> null
List 2: [2] -> [4] -> [6] -> null

Merged: [1] -> [2] -> [3] -> [4] -> [5] -> [6] -> null
```

````python
def merge_sorted_lists(head1: Optional[Node], head2: Optional[Node]) -> Optional[Node]:
    """
    Merge two sorted linked lists into one sorted list.
    
    Time: O(n + m), Space: O(1)
    
    Args:
        head1: Head of first sorted list
        head2: Head of second sorted list
        
    Returns:
        Head of merged sorted list
    """
    # Dummy node to simplify edge cases
    dummy = Node(0)
    current = dummy
    
    while head1 is not None and head2 is not None:
        if head1.data <= head2.data:
            current.next = head1
            head1 = head1.next
        else:
            current.next = head2
            head2 = head2.next
        current = current.next
    
    # Attach remaining nodes
    current.next = head1 if head1 is not None else head2
    
    return dummy.next


def list_to_string(head: Optional[Node]) -> str:
    """Helper to print a list from head node."""
    nodes = []
    while head:
        nodes.append(f"[{head.data}]")
        head = head.next
    nodes.append("null")
    return " -> ".join(nodes)


# Example
print("Merge Two Sorted Lists:")
print("="*40)

# Create two sorted lists
list1 = SinglyLinkedList()
for val in [1, 3, 5, 7]:
    list1.add(val)

list2 = SinglyLinkedList()
for val in [2, 4, 6, 8]:
    list2.add(val)

print(f"List 1: {list1}")
print(f"List 2: {list2}")

merged_head = merge_sorted_lists(list1.head, list2.head)
print(f"Merged: {list_to_string(merged_head)}")
````

### 4.4 Remove N-th Node from End

**Problem**: Remove the n-th node from the end in one pass.

**Solution**: Use two pointers with n-node gap

```
Remove 2nd from end:

         gap of 2
        |-------|  
[1] -> [2] -> [3] -> [4] -> [5] -> null
                      ↑
                to remove (4)
```

````python
def remove_nth_from_end(head: Optional[Node], n: int) -> Optional[Node]:
    """
    Remove the n-th node from the end of the list.
    
    Time: O(length), Space: O(1)
    
    Args:
        head: Head of the linked list
        n: Position from end (1-based)
        
    Returns:
        Head of modified list
    """
    # Dummy node to handle edge case of removing head
    dummy = Node(0)
    dummy.next = head
    
    first = dummy
    second = dummy
    
    # Move first pointer n+1 steps ahead
    for _ in range(n + 1):
        if first is None:
            return head  # n is larger than list length
        first = first.next
    
    # Move both pointers until first reaches end
    while first is not None:
        first = first.next
        second = second.next
    
    # second.next is the node to remove
    second.next = second.next.next
    
    return dummy.next


# Example
print("Remove N-th Node from End:")
print("="*40)

# Create list: 1 -> 2 -> 3 -> 4 -> 5
nodes = [Node(i) for i in range(1, 6)]
for i in range(len(nodes) - 1):
    nodes[i].next = nodes[i + 1]

print(f"Original: {list_to_string(nodes[0])}")
new_head = remove_nth_from_end(nodes[0], 2)
print(f"After removing 2nd from end: {list_to_string(new_head)}")
````

### 4.5 Palindrome Linked List

**Problem**: Check if a linked list is a palindrome.

**Solution**: Find middle, reverse second half, compare

```
Palindrome: [1] -> [2] -> [3] -> [2] -> [1] -> null
                         ↑
                      middle

First half:  [1] -> [2] -> [3]
Second half: [1] -> [2] -> [3] (reversed)
Match! ✓
```

````python
def is_palindrome(head: Optional[Node]) -> bool:
    """
    Check if a linked list is a palindrome.
    
    Time: O(n), Space: O(1)
    
    Note: This modifies the list temporarily but restores it.
    """
    if head is None or head.next is None:
        return True
    
    # Find middle
    slow = head
    fast = head
    while fast.next is not None and fast.next.next is not None:
        slow = slow.next
        fast = fast.next.next
    
    # Reverse second half
    second_half = slow.next
    slow.next = None  # Split the list
    
    prev = None
    while second_half is not None:
        next_node = second_half.next
        second_half.next = prev
        prev = second_half
        second_half = next_node
    
    # Compare halves
    first_half = head
    second_half = prev  # Head of reversed second half
    result = True
    
    while second_half is not None:
        if first_half.data != second_half.data:
            result = False
            break
        first_half = first_half.next
        second_half = second_half.next
    
    return result


# Example
print("Palindrome Check:")
print("="*40)

# Palindrome: 1 -> 2 -> 3 -> 2 -> 1
pal_list = SinglyLinkedList()
for val in [1, 2, 3, 2, 1]:
    pal_list.add(val)
print(f"List: {pal_list}")
print(f"Is palindrome: {is_palindrome(pal_list.head)}")

# Not palindrome: 1 -> 2 -> 3 -> 4 -> 5
not_pal = SinglyLinkedList()
for val in [1, 2, 3, 4, 5]:
    not_pal.add(val)
print(f"\nList: {not_pal}")
print(f"Is palindrome: {is_palindrome(not_pal.head)}")
````

## 5. Summary

### Complexity Comparison

| Operation | Singly LL | Doubly LL | Array |
|-----------|-----------|-----------|-------|
| Access by index | O(n) | O(n) | O(1) |
| Insert at head | O(1) | O(1) | O(n) |
| Insert at tail | O(1)* | O(1) | O(1)** |
| Delete at head | O(1) | O(1) | O(n) |
| Delete at tail | O(n) | O(1) | O(1) |
| Delete given node | O(n) | O(1) | O(n) |
| Search | O(n) | O(n) | O(n) |
| Memory per element | data + 1 ptr | data + 2 ptrs | data only |

*O(1) with tail pointer  
**Amortized for dynamic arrays

### When to Use Each

**Use Singly Linked List when:**
- Memory is constrained (one pointer per node)
- Operations are primarily at the head
- Forward-only traversal is sufficient
- Implementing stacks or simple queues

**Use Doubly Linked List when:**
- Need bidirectional traversal
- Need O(1) deletion at tail or given node
- Implementing deques or LRU caches
- Navigation (browser history, undo/redo)

**Use Array when:**
- Need frequent random access by index
- Cache performance is critical
- Memory overhead must be minimal
- Size is known or changes infrequently

### Key Takeaways

1. **Linked lists excel at dynamic insertion/deletion** - especially at known positions
2. **Trade-off**: Linked lists sacrifice random access for flexible memory allocation
3. **Two-pointer techniques** solve many linked list problems efficiently
4. **Dummy nodes** simplify edge cases in many algorithms
5. **Doubly linked lists** provide O(1) operations at both ends at the cost of extra memory
