## 1. Stack (LIFO)

### What is a Stack?

A **stack** is a linear data structure that follows the **Last-In-First-Out (LIFO)** principle. The last element added to the stack is the first one to be removed.

### Real-World Analogies

- **Stack of plates**: You add plates to the top and remove from the top
- **Browser history**: Back button goes to the most recently visited page
- **Undo functionality**: Most recent action is undone first
- **Function call stack**: Most recently called function returns first

### When to Use?

- **Backtracking algorithms** (maze solving, parsing)
- **Expression evaluation** (postfix, prefix notation)
- **Undo/redo mechanisms**
- **Converting recursion to iteration**
- **Depth-First Search (DFS)**
- **Bracket matching / syntax validation**

### Core Operations

| Operation | Description |
|-----------|-------------|
| `push(item)` | Add item to the top |
| `pop()` | Remove and return top item |
| `peek()` / `top()` | Return top item without removing |
| `is_empty()` | Check if stack is empty |
| `size()` | Return number of elements |

### ASCII Art Visualization

```
         push(5)     push(3)     pop()       push(7)
           ↓           ↓          ↑            ↓
         ┌───┐       ┌───┐      ┌───┐       ┌───┐
         │   │       │ 3 │      │   │       │ 7 │
         ├───┤       ├───┤      ├───┤       ├───┤
         │ 5 │       │ 5 │      │ 5 │       │ 5 │
         └───┘       └───┘      └───┘       └───┘
        top=0       top=1      top=0       top=1
                              returns 3
```

**Internal array representation:**
```
Array: [5, 3, _, _, _]     After pop(): [5, _, _, _, _]
        ↑  ↑                             ↑
      [0][1]                           [0]
        top=1                          top=0
```

### 1.1 Array-Based Stack Implementation

````python
from typing import Any, Optional, List


class ArrayStack:
    """
    Stack implementation using a dynamic array (Python list).
    
    The stack follows LIFO (Last-In-First-Out) principle.
    Elements are added and removed from the end of the array
    for O(1) amortized time complexity.
    
    Attributes:
        _items: Internal list storing stack elements
    
    Time Complexity:
        - push: O(1) amortized
        - pop: O(1)
        - peek: O(1)
        - is_empty: O(1)
    
    Space Complexity: O(n) where n is the number of elements
    """
    
    def __init__(self) -> None:
        """Initialize an empty stack."""
        self._items: List[Any] = []
    
    def push(self, item: Any) -> None:
        """
        Add an item to the top of the stack.
        
        Args:
            item: The element to be added to the stack
        """
        self._items.append(item)
    
    def pop(self) -> Any:
        """
        Remove and return the top item from the stack.
        
        Returns:
            The top element of the stack
        
        Raises:
            IndexError: If the stack is empty
        """
        if self.is_empty():
            raise IndexError("Pop from empty stack")
        return self._items.pop()
    
    def peek(self) -> Any:
        """
        Return the top item without removing it.
        
        Returns:
            The top element of the stack
        
        Raises:
            IndexError: If the stack is empty
        """
        if self.is_empty():
            raise IndexError("Peek from empty stack")
        return self._items[-1]
    
    def is_empty(self) -> bool:
        """Return True if the stack is empty."""
        return len(self._items) == 0
    
    def size(self) -> int:
        """Return the number of elements in the stack."""
        return len(self._items)
    
    def __len__(self) -> int:
        """Return the number of elements (enables len(stack))."""
        return len(self._items)
    
    def __bool__(self) -> bool:
        """Return True if stack is non-empty (enables if stack:)."""
        return len(self._items) > 0
    
    def __str__(self) -> str:
        """Return string representation of the stack."""
        if not self._items:
            return "Stack: [] (empty)"
        items_str = " ".join(str(item) for item in self._items)
        return f"Stack: [{items_str}] ← top"
    
    def __repr__(self) -> str:
        """Return detailed string representation."""
        return f"ArrayStack({self._items})"
````

````python
# Demonstration: Array-based Stack
print("=" * 50)
print("Array-Based Stack Demonstration")
print("=" * 50)

stack = ArrayStack()
print(f"\nInitial state: {stack}")
print(f"Is empty? {stack.is_empty()}")

# Push operations
for value in [10, 20, 30, 40]:
    stack.push(value)
    print(f"push({value}): {stack}")

print(f"\nPeek: {stack.peek()}")
print(f"Size: {stack.size()}")

# Pop operations
print("\nPopping elements:")
while not stack.is_empty():
    value = stack.pop()
    print(f"pop() → {value}, {stack}")
````

### 1.2 Linked-List-Based Stack Implementation

**Advantages over array-based:**
- No need to resize (truly O(1) push, not amortized)
- No wasted space from pre-allocated capacity

**Disadvantages:**
- Extra memory for node pointers
- No random access (not needed for stacks)
- Slightly worse cache locality

````python
class StackNode:
    """
    A node in the linked-list-based stack.
    
    Attributes:
        value: The data stored in this node
        next: Reference to the next node (towards bottom of stack)
    """
    
    def __init__(self, value: Any, next_node: Optional['StackNode'] = None) -> None:
        self.value = value
        self.next = next_node


class LinkedStack:
    """
    Stack implementation using a singly linked list.
    
    Elements are added and removed from the head of the list,
    which represents the top of the stack.
    
    Attributes:
        _top: Reference to the top node of the stack
        _size: Number of elements in the stack
    
    Time Complexity:
        - push: O(1)
        - pop: O(1)
        - peek: O(1)
        - is_empty: O(1)
    
    Space Complexity: O(n) where n is the number of elements
    """
    
    def __init__(self) -> None:
        """Initialize an empty stack."""
        self._top: Optional[StackNode] = None
        self._size: int = 0
    
    def push(self, item: Any) -> None:
        """
        Add an item to the top of the stack.
        
        Creates a new node and makes it the new top.
        
        Args:
            item: The element to be added
        """
        new_node = StackNode(item, self._top)
        self._top = new_node
        self._size += 1
    
    def pop(self) -> Any:
        """
        Remove and return the top item from the stack.
        
        Returns:
            The top element
        
        Raises:
            IndexError: If the stack is empty
        """
        if self.is_empty():
            raise IndexError("Pop from empty stack")
        value = self._top.value
        self._top = self._top.next
        self._size -= 1
        return value
    
    def peek(self) -> Any:
        """
        Return the top item without removing it.
        
        Returns:
            The top element
        
        Raises:
            IndexError: If the stack is empty
        """
        if self.is_empty():
            raise IndexError("Peek from empty stack")
        return self._top.value
    
    def is_empty(self) -> bool:
        """Return True if the stack is empty."""
        return self._top is None
    
    def size(self) -> int:
        """Return the number of elements in the stack."""
        return self._size
    
    def __len__(self) -> int:
        """Return the number of elements."""
        return self._size
    
    def __bool__(self) -> bool:
        """Return True if stack is non-empty."""
        return self._top is not None
    
    def __str__(self) -> str:
        """Return string representation of the stack."""
        if self.is_empty():
            return "Stack: [] (empty)"
        
        elements = []
        current = self._top
        while current:
            elements.append(str(current.value))
            current = current.next
        
        # Reverse to show bottom-to-top order
        return f"Stack: [{' '.join(reversed(elements))}] ← top"
````

````python
# Demonstration: Linked-list-based Stack
print("=" * 50)
print("Linked-List-Based Stack Demonstration")
print("=" * 50)

stack = LinkedStack()
print(f"\nInitial state: {stack}")

# Push operations
for value in ['A', 'B', 'C', 'D']:
    stack.push(value)
    print(f"push('{value}'): {stack}")

print(f"\nPeek: {stack.peek()}")
print(f"Size: {stack.size()}")

# Pop operations
print("\nPopping elements:")
while stack:
    value = stack.pop()
    print(f"pop() → '{value}', {stack}")
````

## 2. Queue (FIFO)

### What is a Queue?

A **queue** is a linear data structure that follows the **First-In-First-Out (FIFO)** principle. The first element added is the first one to be removed.

### Real-World Analogies

- **Line at a store**: First person in line gets served first
- **Print queue**: Documents print in the order they were sent
- **Traffic at intersection**: Cars pass in arrival order
- **Customer service calls**: Calls answered in order received

### When to Use?

- **Breadth-First Search (BFS)**
- **Task scheduling** (process scheduling, print jobs)
- **Message buffers** (producer-consumer problems)
- **Simulation systems** (event-driven simulation)
- **Caching** (LRU cache eviction)

### Core Operations

| Operation | Description |
|-----------|-------------|
| `enqueue(item)` | Add item to the rear |
| `dequeue()` | Remove and return front item |
| `front()` / `peek()` | Return front item without removing |
| `is_empty()` | Check if queue is empty |
| `size()` | Return number of elements |

### ASCII Art Visualization

```
enqueue(5)  enqueue(3)  enqueue(8)   dequeue()    enqueue(2)
    ↓           ↓           ↓           ↑            ↓
  ┌───┐      ┌───┬───┐   ┌───┬───┬───┐  ┌───┬───┐   ┌───┬───┬───┐
  │ 5 │      │ 5 │ 3 │   │ 5 │ 3 │ 8 │  │ 3 │ 8 │   │ 3 │ 8 │ 2 │
  └───┘      └───┴───┘   └───┴───┴───┘  └───┴───┘   └───┴───┴───┘
  front       front       front         front        front
  rear        rear        rear          rear         rear
                                       returns 5
```

**Linked-list representation:**
```
front                                  rear
  ↓                                     ↓
┌───┐    ┌───┐    ┌───┐    ┌───┐
│ 5 │───►│ 3 │───►│ 8 │───►│ 2 │───► None
└───┘    └───┘    └───┘    └───┘

dequeue(): remove from front (5)
enqueue(): add at rear
```

````python
class QueueNode:
    """
    A node in the linked-list-based queue.
    
    Attributes:
        value: The data stored in this node
        next: Reference to the next node (towards rear)
    """
    
    def __init__(self, value: Any, next_node: Optional['QueueNode'] = None) -> None:
        self.value = value
        self.next = next_node


class LinkedQueue:
    """
    Queue implementation using a singly linked list.
    
    Maintains front and rear pointers for O(1) operations.
    Elements are added at the rear and removed from the front.
    
    Attributes:
        _front: Reference to the front node (for dequeue)
        _rear: Reference to the rear node (for enqueue)
        _size: Number of elements in the queue
    
    Time Complexity:
        - enqueue: O(1)
        - dequeue: O(1)
        - peek: O(1)
        - is_empty: O(1)
    
    Space Complexity: O(n) where n is the number of elements
    """
    
    def __init__(self) -> None:
        """Initialize an empty queue."""
        self._front: Optional[QueueNode] = None
        self._rear: Optional[QueueNode] = None
        self._size: int = 0
    
    def enqueue(self, item: Any) -> None:
        """
        Add an item to the rear of the queue.
        
        Args:
            item: The element to be added
        """
        new_node = QueueNode(item)
        
        if self.is_empty():
            self._front = self._rear = new_node
        else:
            self._rear.next = new_node
            self._rear = new_node
        
        self._size += 1
    
    def dequeue(self) -> Any:
        """
        Remove and return the front item from the queue.
        
        Returns:
            The front element
        
        Raises:
            IndexError: If the queue is empty
        """
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")
        
        value = self._front.value
        self._front = self._front.next
        self._size -= 1
        
        # If queue becomes empty, update rear as well
        if self._front is None:
            self._rear = None
        
        return value
    
    def peek(self) -> Any:
        """
        Return the front item without removing it.
        
        Returns:
            The front element
        
        Raises:
            IndexError: If the queue is empty
        """
        if self.is_empty():
            raise IndexError("Peek from empty queue")
        return self._front.value
    
    def is_empty(self) -> bool:
        """Return True if the queue is empty."""
        return self._front is None
    
    def size(self) -> int:
        """Return the number of elements in the queue."""
        return self._size
    
    def __len__(self) -> int:
        """Return the number of elements."""
        return self._size
    
    def __bool__(self) -> bool:
        """Return True if queue is non-empty."""
        return self._front is not None
    
    def __str__(self) -> str:
        """Return string representation of the queue."""
        if self.is_empty():
            return "Queue: [] (empty)"
        
        elements = []
        current = self._front
        while current:
            elements.append(str(current.value))
            current = current.next
        
        return f"Queue: front → [{' '.join(elements)}] ← rear"
````

````python
# Demonstration: Linked-list-based Queue
print("=" * 50)
print("Queue (FIFO) Demonstration")
print("=" * 50)

queue = LinkedQueue()
print(f"\nInitial state: {queue}")

# Enqueue operations
for value in [10, 20, 30, 40]:
    queue.enqueue(value)
    print(f"enqueue({value}): {queue}")

print(f"\nPeek (front): {queue.peek()}")
print(f"Size: {queue.size()}")

# Dequeue operations
print("\nDequeue elements (FIFO order):")
while queue:
    value = queue.dequeue()
    print(f"dequeue() → {value}, {queue}")
````

## 3. Queue Using Two Stacks

### Concept

A queue can be implemented using two stacks. This is a classic interview question that demonstrates how to simulate one data structure using another.

### How It Works

- **Stack 1 (inbox)**: Used for enqueue operations
- **Stack 2 (outbox)**: Used for dequeue operations

When dequeuing:
1. If outbox is empty, transfer ALL elements from inbox to outbox
2. Pop from outbox

The transfer reverses the order, achieving FIFO behavior!

### ASCII Art Visualization

```
Initial: enqueue(1), enqueue(2), enqueue(3)

Stack1 (inbox)    Stack2 (outbox)
    ┌───┐             ┌───┐
    │ 3 │ ← top       │   │
    ├───┤             ├───┤
    │ 2 │             │   │
    ├───┤             ├───┤
    │ 1 │             │   │
    └───┘             └───┘

dequeue() called → transfer all to Stack2:

Stack1 (inbox)    Stack2 (outbox)
    ┌───┐             ┌───┐
    │   │             │ 1 │ ← top (first in!)
    ├───┤             ├───┤
    │   │             │ 2 │
    ├───┤             ├───┤
    │   │             │ 3 │
    └───┘             └───┘

pop() from Stack2 → returns 1 (FIFO achieved!)
```

### Amortized Analysis

Each element is:
1. Pushed once to inbox
2. Popped once from inbox (during transfer)
3. Pushed once to outbox (during transfer)
4. Popped once from outbox

Total: 4 operations per element → **O(1) amortized** per operation!

````python
class QueueWithTwoStacks:
    """
    Queue implementation using two stacks.
    
    This implementation demonstrates how to achieve FIFO behavior
    using two LIFO structures (stacks).
    
    Algorithm:
        - enqueue: Push to inbox stack (O(1))
        - dequeue: Pop from outbox; if empty, transfer all from inbox
        - transfer: Reverses order, converting LIFO to FIFO
    
    Attributes:
        _inbox: Stack for incoming elements
        _outbox: Stack for outgoing elements
    
    Time Complexity:
        - enqueue: O(1)
        - dequeue: O(1) amortized (O(n) worst case when transfer needed)
        - peek: O(1) amortized
    
    Space Complexity: O(n) where n is the number of elements
    """
    
    def __init__(self) -> None:
        """Initialize an empty queue with two stacks."""
        self._inbox: ArrayStack = ArrayStack()   # For enqueue
        self._outbox: ArrayStack = ArrayStack()  # For dequeue
    
    def _transfer(self) -> None:
        """
        Transfer all elements from inbox to outbox.
        
        This reverses the order of elements, enabling FIFO behavior.
        Only called when outbox is empty and dequeue is needed.
        """
        while self._inbox:
            self._outbox.push(self._inbox.pop())
    
    def enqueue(self, item: Any) -> None:
        """
        Add an item to the rear of the queue.
        
        Args:
            item: The element to be added
        """
        self._inbox.push(item)
    
    def dequeue(self) -> Any:
        """
        Remove and return the front item from the queue.
        
        Returns:
            The front element
        
        Raises:
            IndexError: If the queue is empty
        """
        if self.is_empty():
            raise IndexError("Dequeue from empty queue")
        
        if self._outbox.is_empty():
            self._transfer()
        
        return self._outbox.pop()
    
    def peek(self) -> Any:
        """
        Return the front item without removing it.
        
        Returns:
            The front element
        
        Raises:
            IndexError: If the queue is empty
        """
        if self.is_empty():
            raise IndexError("Peek from empty queue")
        
        if self._outbox.is_empty():
            self._transfer()
        
        return self._outbox.peek()
    
    def is_empty(self) -> bool:
        """Return True if the queue is empty."""
        return self._inbox.is_empty() and self._outbox.is_empty()
    
    def size(self) -> int:
        """Return the number of elements in the queue."""
        return len(self._inbox) + len(self._outbox)
    
    def __len__(self) -> int:
        """Return the number of elements."""
        return self.size()
    
    def __bool__(self) -> bool:
        """Return True if queue is non-empty."""
        return not self.is_empty()
    
    def __str__(self) -> str:
        """Return string representation showing both stacks."""
        return f"QueueTwoStacks(inbox={self._inbox._items}, outbox={self._outbox._items})"
````

````python
# Demonstration: Queue with Two Stacks
print("=" * 50)
print("Queue Using Two Stacks Demonstration")
print("=" * 50)

queue = QueueWithTwoStacks()
print(f"\nInitial state: {queue}")

# Enqueue several elements
print("\n--- Enqueue Phase ---")
for value in [1, 2, 3, 4, 5]:
    queue.enqueue(value)
    print(f"enqueue({value}): {queue}")

# Dequeue some elements (triggers transfer)
print("\n--- First Dequeue (triggers transfer) ---")
value = queue.dequeue()
print(f"dequeue() → {value}: {queue}")

print("\n--- More dequeues (from outbox) ---")
for _ in range(2):
    value = queue.dequeue()
    print(f"dequeue() → {value}: {queue}")

# Enqueue more while outbox has elements
print("\n--- Enqueue while outbox has elements ---")
for value in [6, 7]:
    queue.enqueue(value)
    print(f"enqueue({value}): {queue}")

# Dequeue remaining (will need another transfer)
print("\n--- Dequeue all remaining ---")
while queue:
    value = queue.dequeue()
    print(f"dequeue() → {value}: {queue}")
````

## 4. Applications

### 4.1 Bracket Matching

One of the most common applications of stacks is validating balanced brackets in expressions.

**Algorithm:**
1. For each opening bracket, push it onto the stack
2. For each closing bracket, pop from stack and check if they match
3. At the end, stack should be empty

```
Input: "({[]})"

Read '(' → push → Stack: [(]
Read '{' → push → Stack: [(, {]
Read '[' → push → Stack: [(, {, []
Read ']' → pop '[', match! → Stack: [(, {]
Read '}' → pop '{', match! → Stack: [(]
Read ')' → pop '(', match! → Stack: []

Stack empty at end → VALID!
```

````python
def is_balanced(expression: str) -> bool:
    """
    Check if brackets in the expression are balanced.
    
    Supports: (), [], {}
    
    Args:
        expression: String containing brackets to validate
    
    Returns:
        True if all brackets are properly matched and nested
    
    Examples:
        >>> is_balanced("({[]})")
        True
        >>> is_balanced("([)]")
        False
        >>> is_balanced("((())")
        False
    """
    stack = ArrayStack()
    matching = {')': '(', ']': '[', '}': '{'}
    opening = set('([{')
    closing = set(')]}')
    
    for char in expression:
        if char in opening:
            stack.push(char)
        elif char in closing:
            if stack.is_empty():
                return False  # No matching opening bracket
            if stack.pop() != matching[char]:
                return False  # Mismatched bracket type
    
    return stack.is_empty()  # All brackets should be matched


def is_balanced_verbose(expression: str) -> tuple[bool, str]:
    """
    Check if brackets are balanced with detailed trace.
    
    Args:
        expression: String to validate
    
    Returns:
        Tuple of (is_valid, trace_string)
    """
    stack = ArrayStack()
    matching = {')': '(', ']': '[', '}': '{'}
    opening = set('([{')
    closing = set(')]}')
    trace = []
    
    for i, char in enumerate(expression):
        if char in opening:
            stack.push(char)
            trace.append(f"[{i}] '{char}' → push → Stack: {list(stack._items)}")
        elif char in closing:
            if stack.is_empty():
                trace.append(f"[{i}] '{char}' → ERROR: no opening bracket!")
                return False, '\n'.join(trace)
            popped = stack.pop()
            if popped != matching[char]:
                trace.append(f"[{i}] '{char}' → pop '{popped}' → MISMATCH!")
                return False, '\n'.join(trace)
            trace.append(f"[{i}] '{char}' → pop '{popped}', match! → Stack: {list(stack._items)}")
    
    if stack.is_empty():
        trace.append("\nStack empty → VALID!")
        return True, '\n'.join(trace)
    else:
        trace.append(f"\nStack not empty: {list(stack._items)} → INVALID!")
        return False, '\n'.join(trace)
````

````python
# Demonstration: Bracket Matching
print("=" * 50)
print("Bracket Matching Application")
print("=" * 50)

test_cases = [
    ("({[]})", True),
    ("(){}[]", True),
    ("((()))", True),
    ("([)]", False),
    ("((())", False),
    ("())", False),
    ("", True),
    ("a + (b * [c - d]) / {e}", True),
]

print("\n--- Quick Tests ---")
for expr, expected in test_cases:
    result = is_balanced(expr)
    status = "✓" if result == expected else "✗"
    print(f"{status} is_balanced('{expr}') = {result}")

# Detailed trace for one example
print("\n--- Detailed Trace ---")
expr = "({[]})"
print(f"\nInput: '{expr}'\n")
valid, trace = is_balanced_verbose(expr)
print(trace)

print("\n--- Trace for Invalid Input ---")
expr = "([)]"
print(f"\nInput: '{expr}'\n")
valid, trace = is_balanced_verbose(expr)
print(trace)
````

### 4.2 Postfix Expression Evaluation

**Postfix notation** (Reverse Polish Notation) places operators after operands.

| Infix | Postfix |
|-------|--------|
| `3 + 4` | `3 4 +` |
| `3 + 4 * 2` | `3 4 2 * +` |
| `(3 + 4) * 2` | `3 4 + 2 *` |

**Algorithm:**
1. For each operand, push onto stack
2. For each operator, pop two operands, apply operation, push result
3. Final result is the last element on stack

```
Evaluate: "3 4 2 * +"

Read '3' → push → Stack: [3]
Read '4' → push → Stack: [3, 4]
Read '2' → push → Stack: [3, 4, 2]
Read '*' → pop 2,4 → 4*2=8 → push → Stack: [3, 8]
Read '+' → pop 8,3 → 3+8=11 → push → Stack: [11]

Result: 11
```

````python
def evaluate_postfix(expression: str) -> float:
    """
    Evaluate a postfix (Reverse Polish Notation) expression.
    
    Supported operators: +, -, *, /
    Tokens must be space-separated.
    
    Args:
        expression: Space-separated postfix expression
    
    Returns:
        The result of the expression evaluation
    
    Examples:
        >>> evaluate_postfix("3 4 +")
        7.0
        >>> evaluate_postfix("3 4 2 * +")
        11.0
        >>> evaluate_postfix("5 1 2 + 4 * + 3 -")
        14.0
    """
    stack = ArrayStack()
    operators = {'+', '-', '*', '/'}
    
    for token in expression.split():
        if token in operators:
            # Pop two operands (note: order matters for - and /)
            b = stack.pop()  # Second operand
            a = stack.pop()  # First operand
            
            if token == '+':
                result = a + b
            elif token == '-':
                result = a - b
            elif token == '*':
                result = a * b
            elif token == '/':
                result = a / b
            
            stack.push(result)
        else:
            # Token is a number
            stack.push(float(token))
    
    return stack.pop()


def evaluate_postfix_verbose(expression: str) -> tuple[float, str]:
    """
    Evaluate postfix expression with detailed trace.
    
    Args:
        expression: Space-separated postfix expression
    
    Returns:
        Tuple of (result, trace_string)
    """
    stack = ArrayStack()
    operators = {'+', '-', '*', '/'}
    trace = []
    
    for token in expression.split():
        if token in operators:
            b = stack.pop()
            a = stack.pop()
            
            if token == '+':
                result = a + b
            elif token == '-':
                result = a - b
            elif token == '*':
                result = a * b
            elif token == '/':
                result = a / b
            
            stack.push(result)
            trace.append(f"'{token}' → pop {b},{a} → {a}{token}{b}={result} → push → Stack: {list(stack._items)}")
        else:
            stack.push(float(token))
            trace.append(f"'{token}' → push → Stack: {list(stack._items)}")
    
    result = stack.pop()
    trace.append(f"\nResult: {result}")
    return result, '\n'.join(trace)
````

### 4.3 Iterative QuickSort Using Stack

Recursion uses the call stack internally. We can convert recursive algorithms to iterative ones using an explicit stack.

**Why use iterative QuickSort?**
- Avoids stack overflow for large arrays
- More control over memory usage
- Can be faster due to less function call overhead

**Algorithm:**
1. Push initial range `(0, n-1)` onto stack
2. While stack is not empty:
   - Pop a range `(low, high)`
   - Partition and get pivot index
   - Push sub-ranges that need sorting

````python
def partition(arr: List[int], low: int, high: int) -> int:
    """
    Partition the array around a pivot element.
    
    Uses the first element as pivot and rearranges so that:
    - Elements < pivot are to the left
    - Elements >= pivot are to the right
    
    Args:
        arr: The array to partition
        low: Starting index of the range
        high: Ending index of the range
    
    Returns:
        The final position of the pivot element
    """
    pivot_value = arr[low]
    pivot_index = low
    
    for i in range(low + 1, high + 1):
        if arr[i] <= pivot_value:
            pivot_index += 1
            arr[i], arr[pivot_index] = arr[pivot_index], arr[i]
    
    # Place pivot in its final position
    arr[low], arr[pivot_index] = arr[pivot_index], arr[low]
    return pivot_index


def quicksort_iterative(arr: List[int]) -> None:
    """
    Sort an array in-place using iterative QuickSort.
    
    Uses an explicit stack instead of recursion to manage
    the subranges that need to be sorted.
    
    Args:
        arr: The array to sort (modified in-place)
    
    Time Complexity:
        - Average: O(n log n)
        - Worst: O(n²) (already sorted array)
    
    Space Complexity: O(log n) for the stack
    """
    if len(arr) <= 1:
        return
    
    stack = ArrayStack()
    stack.push((0, len(arr) - 1))
    
    while stack:
        low, high = stack.pop()
        
        if low < high:
            pivot_index = partition(arr, low, high)
            
            # Push larger subarray first (optimization: smaller stack depth)
            if pivot_index - low > high - pivot_index:
                stack.push((low, pivot_index - 1))
                stack.push((pivot_index + 1, high))
            else:
                stack.push((pivot_index + 1, high))
                stack.push((low, pivot_index - 1))


def quicksort_iterative_verbose(arr: List[int]) -> List[str]:
    """
    Sort array with detailed trace of stack operations.
    
    Args:
        arr: The array to sort
    
    Returns:
        List of trace messages
    """
    trace = [f"Initial array: {arr}"]
    
    if len(arr) <= 1:
        trace.append("Array too small, already sorted.")
        return trace
    
    stack = ArrayStack()
    stack.push((0, len(arr) - 1))
    trace.append(f"Push (0, {len(arr)-1}) → Stack: {list(stack._items)}")
    
    step = 1
    while stack:
        low, high = stack.pop()
        trace.append(f"\n--- Step {step} ---")
        trace.append(f"Pop ({low}, {high})")
        
        if low < high:
            pivot_index = partition(arr, low, high)
            trace.append(f"Partition around index {pivot_index} (value={arr[pivot_index]})")
            trace.append(f"Array now: {arr}")
            
            # Push sub-ranges
            if pivot_index - 1 > low:
                stack.push((low, pivot_index - 1))
                trace.append(f"Push left ({low}, {pivot_index-1})")
            if pivot_index + 1 < high:
                stack.push((pivot_index + 1, high))
                trace.append(f"Push right ({pivot_index+1}, {high})")
            
            trace.append(f"Stack: {list(stack._items)}")
        else:
            trace.append(f"Range too small (low >= high), skip")
        
        step += 1
    
    trace.append(f"\nFinal sorted array: {arr}")
    return trace
````

````python
# Demonstration: Iterative QuickSort
print("=" * 50)
print("Iterative QuickSort Using Stack")
print("=" * 50)

# Test with various arrays
test_arrays = [
    [64, 34, 25, 12, 22, 11, 90],
    [5, 4, 3, 2, 1],
    [1, 2, 3, 4, 5],
    [3, 1, 4, 1, 5, 9, 2, 6],
]

print("\n--- Quick Tests ---")
for arr in test_arrays:
    original = arr.copy()
    quicksort_iterative(arr)
    status = "✓" if arr == sorted(original) else "✗"
    print(f"{status} {original} → {arr}")

# Detailed trace
print("\n" + "=" * 50)
print("Detailed Trace")
print("=" * 50)

arr = [7, 2, 9, 4, 1, 5]
trace = quicksort_iterative_verbose(arr)
print('\n'.join(trace))
````

````python
# Performance comparison: Recursive vs Iterative QuickSort
import random
import time

def quicksort_recursive(arr: List[int], low: int = None, high: int = None) -> None:
    """Traditional recursive QuickSort for comparison."""
    if low is None:
        low, high = 0, len(arr) - 1
    
    if low < high:
        pivot_index = partition(arr, low, high)
        quicksort_recursive(arr, low, pivot_index - 1)
        quicksort_recursive(arr, pivot_index + 1, high)


print("=" * 50)
print("Performance Comparison")
print("=" * 50)

sizes = [100, 1000, 5000]

for n in sizes:
    arr1 = [random.randint(0, 10000) for _ in range(n)]
    arr2 = arr1.copy()
    
    # Recursive
    start = time.perf_counter()
    quicksort_recursive(arr1)
    recursive_time = time.perf_counter() - start
    
    # Iterative
    start = time.perf_counter()
    quicksort_iterative(arr2)
    iterative_time = time.perf_counter() - start
    
    print(f"\nn = {n:,}:")
    print(f"  Recursive: {recursive_time*1000:.3f} ms")
    print(f"  Iterative: {iterative_time*1000:.3f} ms")
    print(f"  Correctly sorted: {arr1 == arr2 == sorted(arr1)}")
````

## 5. Complexity Summary

### Time Complexity

| Operation | Array Stack | Linked Stack | Queue | Queue (2 Stacks) |
|-----------|-------------|--------------|-------|------------------|
| Push/Enqueue | O(1)* | O(1) | O(1) | O(1) |
| Pop/Dequeue | O(1) | O(1) | O(1) | O(1) amortized |
| Peek | O(1) | O(1) | O(1) | O(1) amortized |
| is_empty | O(1) | O(1) | O(1) | O(1) |
| size | O(1) | O(1) | O(1) | O(1) |

*O(1) amortized; occasionally O(n) when array needs to resize

### Space Complexity

| Structure | Space | Notes |
|-----------|-------|-------|
| Array Stack | O(n) | May have unused capacity |
| Linked Stack | O(n) | Extra memory for node pointers |
| Linked Queue | O(n) | Extra memory for node pointers |
| Queue (2 Stacks) | O(n) | Two arrays, but same total elements |

### When to Use Each

| Use Case | Data Structure | Reason |
|----------|----------------|--------|
| Undo/Redo | Stack | LIFO matches undo order |
| DFS | Stack | Explore deepest paths first |
| BFS | Queue | Explore breadth-first |
| Print queue | Queue | FIFO order |
| Expression parsing | Stack | Handle nested parentheses |
| Converting recursion | Stack | Simulates call stack |

---

## Summary

### Key Takeaways

1. **Stack (LIFO)**
   - Last element added is first removed
   - Used for backtracking, parsing, recursion conversion
   - O(1) for all core operations

2. **Queue (FIFO)**
   - First element added is first removed
   - Used for BFS, scheduling, buffering
   - O(1) for all core operations

3. **Queue with Two Stacks**
   - Demonstrates data structure transformation
   - O(1) amortized through lazy transfer
   - Classic interview problem

4. **Applications**
   - Bracket matching: validates nested structures
   - Postfix evaluation: eliminates need for precedence rules
   - Iterative algorithms: converts recursion to iteration

### Implementation Choices

| Factor | Array-based | Linked-list-based |
|--------|-------------|-------------------|
| Memory | More efficient | Extra pointer overhead |
| Cache | Better locality | Worse locality |
| Resize | Occasional O(n) | Never needed |
| Simplicity | Simpler | More complex |
