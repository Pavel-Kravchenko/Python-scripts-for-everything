# 1. Bubble Sort

## Theory

**Bubble Sort** is one of the simplest sorting algorithms. It works by repeatedly stepping through the list, comparing adjacent elements and swapping them if they are in the wrong order. This process continues until the list is sorted.

### How It Works
1. Start at the beginning of the array
2. Compare each pair of adjacent elements
3. If they're in the wrong order, swap them
4. After each pass, the largest unsorted element "bubbles up" to its correct position
5. Repeat until no swaps are needed

### When to Use
- **Educational purposes** - excellent for understanding sorting concepts
- **Small datasets** - acceptable for arrays with < 50 elements
- **Nearly sorted data** - can be efficient with optimized early termination
- **Memory constrained** - requires O(1) extra space

### Why It's Useful
- Simple to understand and implement
- Stable sorting algorithm (preserves relative order of equal elements)
- In-place algorithm (no extra memory needed)
- Can detect already-sorted arrays in O(n)

## ASCII Art Visualization

```
Initial: [5, 3, 8, 1, 2]

Pass 1:
[5, 3, 8, 1, 2]    Compare 5 and 3
 ↑  ↑              5 > 3? Yes → Swap!
[3, 5, 8, 1, 2]    Compare 5 and 8
    ↑  ↑           5 > 8? No → Keep
[3, 5, 8, 1, 2]    Compare 8 and 1
       ↑  ↑        8 > 1? Yes → Swap!
[3, 5, 1, 8, 2]    Compare 8 and 2
          ↑  ↑     8 > 2? Yes → Swap!
[3, 5, 1, 2, 8]    ← 8 bubbled to end! ✓
             ^

Pass 2:
[3, 5, 1, 2, 8]    Compare 3 and 5
 ↑  ↑              3 > 5? No → Keep
[3, 5, 1, 2, 8]    Compare 5 and 1
    ↑  ↑           5 > 1? Yes → Swap!
[3, 1, 5, 2, 8]    Compare 5 and 2
       ↑  ↑        5 > 2? Yes → Swap!
[3, 1, 2, 5, 8]    ← 5 in place! ✓
          ^  ^

Pass 3:
[3, 1, 2, 5, 8]    Compare 3 and 1
 ↑  ↑              3 > 1? Yes → Swap!
[1, 3, 2, 5, 8]    Compare 3 and 2
    ↑  ↑           3 > 2? Yes → Swap!
[1, 2, 3, 5, 8]    ← 3 in place! ✓
       ^  ^  ^

Pass 4:
[1, 2, 3, 5, 8]    Compare 1 and 2
 ↑  ↑              1 > 2? No → Keep
[1, 2, 3, 5, 8]    ← No swaps! Array is sorted! ✓
 ^  ^  ^  ^  ^

Final: [1, 2, 3, 5, 8]
```

## Complexity Analysis

| Case | Time Complexity | Space Complexity | Stable? |
|------|-----------------|------------------|--------|
| Best | O(n) | O(1) | Yes |
| Average | O(n²) | O(1) | Yes |
| Worst | O(n²) | O(1) | Yes |

- **Best case O(n)**: Array is already sorted (with optimization)
- **Worst case O(n²)**: Array is sorted in reverse order
- **Space O(1)**: Only uses a constant amount of extra memory for swapping

## Implementation

````python
def bubble_sort(arr: List[int], verbose: bool = False) -> List[int]:
    """
    Sort an array using the Bubble Sort algorithm.
    
    Bubble Sort repeatedly steps through the list, compares adjacent elements,
    and swaps them if they are in the wrong order. The pass through the list
    is repeated until the list is sorted.
    
    Args:
        arr: List of integers to be sorted
        verbose: If True, print each step of the sorting process
    
    Returns:
        The sorted list (sorted in-place)
    
    Time Complexity:
        - Best: O(n) when array is already sorted
        - Average: O(n²)
        - Worst: O(n²) when array is reverse sorted
    
    Space Complexity: O(1) - in-place sorting
    
    Example:
        >>> bubble_sort([64, 34, 25, 12, 22, 11, 90])
        [11, 12, 22, 25, 34, 64, 90]
    """
    n = len(arr)
    
    if verbose:
        print(f"Initial array: {arr}")
        print("-" * 50)
    
    # Traverse through all array elements
    for i in range(n - 1):
        swapped = False  # Optimization: track if any swaps occurred
        
        if verbose:
            print(f"\nPass {i + 1}:")
        
        # Last i elements are already in place, so we reduce the range
        for j in range(n - 1 - i):
            # Compare adjacent elements
            if arr[j] > arr[j + 1]:
                # Swap if current element is greater than next
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
                
                if verbose:
                    print(f"  Swapped {arr[j + 1]} and {arr[j]}: {arr}")
        
        # If no swapping occurred in this pass, array is sorted
        if not swapped:
            if verbose:
                print(f"  No swaps needed - array is sorted!")
            break
    
    if verbose:
        print("-" * 50)
        print(f"Final sorted array: {arr}")
    
    return arr
````

## Examples

````python
# Example 1: Basic usage with step-by-step output
test_array = [5, 3, 8, 1, 2]
print("=== Bubble Sort Example ===")
bubble_sort(test_array.copy(), verbose=True)
````

````python
# Example 2: Already sorted array (best case)
sorted_array = [1, 2, 3, 4, 5]
print("\n=== Best Case: Already Sorted ===")
bubble_sort(sorted_array.copy(), verbose=True)
````

````python
# Example 3: Reverse sorted array (worst case)
reverse_array = [5, 4, 3, 2, 1]
print("\n=== Worst Case: Reverse Sorted ===")
bubble_sort(reverse_array.copy(), verbose=True)
````

# 2. Merge Sort

## Theory

**Merge Sort** is an efficient, stable, divide-and-conquer sorting algorithm. It divides the input array into two halves, recursively sorts them, and then merges the sorted halves.

### How It Works
1. **Divide**: Split the array into two halves
2. **Conquer**: Recursively sort each half
3. **Combine**: Merge the two sorted halves into one sorted array

### When to Use
- **Large datasets** - consistent O(n log n) performance
- **Linked lists** - very efficient for linked data structures
- **External sorting** - excellent for sorting data that doesn't fit in memory
- **Stability required** - preserves relative order of equal elements
- **Parallel processing** - easily parallelizable

### Why It's Useful
- Guaranteed O(n log n) time complexity in all cases
- Stable sorting algorithm
- Well-suited for external sorting (files, databases)
- Highly parallelizable

## Complexity Analysis

| Case | Time Complexity | Space Complexity | Stable? |
|------|-----------------|------------------|--------|
| Best | O(n log n) | O(n) | Yes |
| Average | O(n log n) | O(n) | Yes |
| Worst | O(n log n) | O(n) | Yes |

- **Time O(n log n)**: Dividing takes O(log n), merging takes O(n) at each level
- **Space O(n)**: Requires auxiliary array for merging
- **Stable**: Maintains relative order of equal elements

## Implementation

````python
def merge_sort(arr: List[int], verbose: bool = False, depth: int = 0) -> List[int]:
    """
    Sort an array using the Merge Sort algorithm.
    
    Merge Sort is a divide-and-conquer algorithm that:
    1. Divides the array into two halves
    2. Recursively sorts each half
    3. Merges the sorted halves back together
    
    Args:
        arr: List of integers to be sorted
        verbose: If True, print each step of the sorting process
        depth: Current recursion depth (for visualization)
    
    Returns:
        A new sorted list
    
    Time Complexity: O(n log n) in all cases
    Space Complexity: O(n) for the auxiliary arrays
    
    Example:
        >>> merge_sort([38, 27, 43, 3, 9, 82, 10])
        [3, 9, 10, 27, 38, 43, 82]
    """
    indent = "  " * depth  # For visualization
    
    if verbose:
        print(f"{indent}Splitting: {arr}")
    
    # Base case: arrays with 0 or 1 element are already sorted
    if len(arr) <= 1:
        return arr
    
    # Divide: find the middle point
    mid = len(arr) // 2
    
    # Conquer: recursively sort both halves
    left_half = merge_sort(arr[:mid], verbose, depth + 1)
    right_half = merge_sort(arr[mid:], verbose, depth + 1)
    
    # Combine: merge the sorted halves
    merged = merge(left_half, right_half)
    
    if verbose:
        print(f"{indent}Merged: {left_half} + {right_half} → {merged}")
    
    return merged


def merge(left: List[int], right: List[int]) -> List[int]:
    """
    Merge two sorted arrays into one sorted array.
    
    Args:
        left: First sorted list
        right: Second sorted list
    
    Returns:
        A new sorted list containing all elements from both inputs
    """
    result = []
    i = j = 0
    
    # Compare elements from both arrays and add smaller one
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:  # <= ensures stability
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Add remaining elements from left array (if any)
    while i < len(left):
        result.append(left[i])
        i += 1
    
    # Add remaining elements from right array (if any)
    while j < len(right):
        result.append(right[j])
        j += 1
    
    return result
````

## Examples

````python
# Example 1: Basic usage with step-by-step output
test_array = [38, 27, 43, 3, 9, 82, 10]
print("=== Merge Sort Example ===")
print(f"Original: {test_array}")
print("\nSorting process:")
result = merge_sort(test_array.copy(), verbose=True)
print(f"\nSorted: {result}")
````

````python
# Example 2: Demonstrating stability
print("\n=== Stability Demonstration ===")
# Using tuples (value, original_index) to show stability
data = [(3, 'a'), (1, 'b'), (3, 'c'), (2, 'd'), (1, 'e')]
print(f"Original: {data}")
# Sort by first element only
sorted_data = sorted(data, key=lambda x: x[0])  # Python's sort is stable like merge sort
print(f"Stable sorted: {sorted_data}")
print("Note: (3, 'a') comes before (3, 'c'), and (1, 'b') comes before (1, 'e')")
````

# 3. Shell Sort

## Theory

**Shell Sort** is an optimization of Insertion Sort that allows exchanging items that are far apart. It starts by sorting elements far apart from each other and progressively reduces the gap between elements to be compared.

### How It Works
1. Choose a gap sequence (e.g., n/2, n/4, n/8, ..., 1)
2. For each gap, perform a gapped insertion sort
3. Reduce the gap and repeat
4. Final pass with gap=1 is a standard insertion sort

### When to Use
- **Medium-sized arrays** - good for n < 10,000
- **Memory constrained** - in-place with O(1) extra space
- **Embedded systems** - simple implementation, good cache performance
- **Nearly sorted data** - performs well on partially sorted arrays

### Why It's Useful
- Better than O(n²) in practice for most gap sequences
- In-place sorting (O(1) extra memory)
- Simple to implement
- Good for medium-sized arrays

## Complexity Analysis

| Case | Time Complexity | Space Complexity | Stable? |
|------|-----------------|------------------|--------|
| Best | O(n log n) | O(1) | No |
| Average | O(n^1.25) to O(n^1.5) | O(1) | No |
| Worst | O(n²) | O(1) | No |

- **Best case**: Depends on gap sequence, can be O(n log n)
- **Average case**: Typically O(n^1.25) with good gap sequences
- **Worst case**: O(n²) with poor gap sequence
- **Not stable**: May change relative order of equal elements

Note: Complexity depends heavily on the gap sequence used.

## Implementation

````python
def shell_sort(arr: List[int], verbose: bool = False) -> List[int]:
    """
    Sort an array using the Shell Sort algorithm.
    
    Shell Sort is an optimization of insertion sort that allows
    the exchange of items that are far apart. The idea is to arrange
    the list of elements so that, starting anywhere, taking every hth
    element produces a sorted list.
    
    Args:
        arr: List of integers to be sorted
        verbose: If True, print each step of the sorting process
    
    Returns:
        The sorted list (sorted in-place)
    
    Time Complexity:
        - Best: O(n log n)
        - Average: O(n^1.25) to O(n^1.5) depending on gap sequence
        - Worst: O(n²)
    
    Space Complexity: O(1) - in-place sorting
    
    Example:
        >>> shell_sort([35, 33, 42, 10, 14, 19, 27, 44])
        [10, 14, 19, 27, 33, 35, 42, 44]
    """
    n = len(arr)
    
    if verbose:
        print(f"Initial array: {arr}")
        print(f"Array length: {n}")
        print("-" * 50)
    
    # Start with a large gap, then reduce the gap
    # Using Shell's original gap sequence: n/2, n/4, ..., 1
    gap = n // 2
    
    while gap > 0:
        if verbose:
            print(f"\nGap = {gap}:")
        
        # Perform gapped insertion sort for this gap size
        for start_pos in range(gap):
            # Do insertion sort for elements at positions: start_pos, start_pos+gap, start_pos+2*gap, ...
            arr = gap_insertion_sort(arr, start_pos, gap, verbose)
        
        if verbose:
            print(f"After gap={gap}: {arr}")
        
        # Reduce the gap for next iteration
        gap //= 2
    
    if verbose:
        print("-" * 50)
        print(f"Final sorted array: {arr}")
    
    return arr


def gap_insertion_sort(arr: List[int], start: int, gap: int, verbose: bool = False) -> List[int]:
    """
    Perform insertion sort on elements separated by gap, starting from start.
    
    Args:
        arr: The array to sort
        start: Starting index for this sublist
        gap: The gap between elements
        verbose: If True, print operations
    
    Returns:
        The modified array
    """
    # Start from start+gap and work towards end of array
    for i in range(start + gap, len(arr), gap):
        current_value = arr[i]
        position = i
        
        # Shift elements that are greater than current_value
        while position >= gap and arr[position - gap] > current_value:
            arr[position] = arr[position - gap]
            position = position - gap
            
            if verbose and position != i - gap:  # Only print if we did a shift
                print(f"  Shifted: {arr}")
        
        # Place current_value at its correct position
        if position != i:
            arr[position] = current_value
            if verbose:
                print(f"  Inserted {current_value} at position {position}: {arr}")
    
    return arr
````

# 4. QuickSort

## Theory

**QuickSort** is a highly efficient, divide-and-conquer sorting algorithm. It works by selecting a 'pivot' element and partitioning the array around the pivot, such that elements smaller than the pivot are on the left and elements greater are on the right.

### How It Works
1. **Choose a pivot** element from the array
2. **Partition**: Reorder the array so elements less than pivot are before it, and elements greater are after it
3. **Recursively** apply the above steps to the sub-arrays

### When to Use
- **General purpose sorting** - often the fastest in practice
- **Arrays** - excellent cache performance
- **Large datasets** - average O(n log n) with low constant factors
- **In-place sorting needed** - requires minimal extra memory

### Why It's Useful
- Fastest sorting algorithm on average for random data
- In-place (O(log n) stack space for recursion)
- Cache-friendly access patterns
- Widely used in standard libraries

## ASCII Art Visualization

## Complexity Analysis

| Case | Time Complexity | Space Complexity | Stable? |
|------|-----------------|------------------|--------|
| Best | O(n log n) | O(log n) | No |
| Average | O(n log n) | O(log n) | No |
| Worst | O(n²) | O(n) | No |

- **Best/Average case**: Good pivot selection divides array roughly in half
- **Worst case O(n²)**: Poor pivot selection (already sorted, all equal)
- **Space O(log n)**: Recursion stack depth (O(n) in worst case)
- **Not stable**: Partition may change relative order of equal elements

## Implementation: Recursive QuickSort

````python
def quicksort_recursive(arr: List[int], verbose: bool = False) -> List[int]:
    """
    Sort an array using the recursive QuickSort algorithm.
    
    QuickSort is a divide-and-conquer algorithm that:
    1. Selects a pivot element
    2. Partitions the array around the pivot
    3. Recursively sorts the sub-arrays
    
    Args:
        arr: List of integers to be sorted
        verbose: If True, print each step of the sorting process
    
    Returns:
        The sorted list (sorted in-place)
    
    Time Complexity:
        - Best: O(n log n)
        - Average: O(n log n)
        - Worst: O(n²) when array is already sorted
    
    Space Complexity: O(log n) average for recursion stack
    
    Example:
        >>> quicksort_recursive([3, 7, 2, 9, 1, 5])
        [1, 2, 3, 5, 7, 9]
    """
    if verbose:
        print(f"Initial array: {arr}")
        print("-" * 50)
    
    _quicksort_helper(arr, 0, len(arr) - 1, verbose, 0)
    
    if verbose:
        print("-" * 50)
        print(f"Final sorted array: {arr}")
    
    return arr


def _quicksort_helper(arr: List[int], low: int, high: int, verbose: bool, depth: int) -> None:
    """
    Helper function for recursive quicksort.
    
    Args:
        arr: The array to sort
        low: Starting index of the subarray
        high: Ending index of the subarray
        verbose: Print steps if True
        depth: Current recursion depth for visualization
    """
    indent = "  " * depth
    
    if low < high:
        if verbose:
            print(f"{indent}Sorting subarray [{low}:{high+1}]: {arr[low:high+1]}")
        
        # Partition and get pivot position
        pivot_idx = _partition(arr, low, high, verbose, indent)
        
        if verbose:
            print(f"{indent}After partition: {arr[low:high+1]}, pivot at index {pivot_idx}")
        
        # Recursively sort elements before and after partition
        _quicksort_helper(arr, low, pivot_idx - 1, verbose, depth + 1)
        _quicksort_helper(arr, pivot_idx + 1, high, verbose, depth + 1)


def _partition(arr: List[int], low: int, high: int, verbose: bool, indent: str) -> int:
    """
    Partition the array using the last element as pivot.
    
    The Lomuto partition scheme:
    - Choose the last element as pivot
    - Maintain index i such that elements arr[low..i-1] are less than pivot
    - Elements arr[i..j-1] are greater than or equal to pivot
    
    Args:
        arr: The array to partition
        low: Starting index
        high: Ending index (pivot position)
        verbose: Print steps if True
        indent: Indentation for output
    
    Returns:
        The final position of the pivot
    """
    pivot = arr[high]  # Choose last element as pivot
    i = low  # Index of smaller element
    
    if verbose:
        print(f"{indent}Pivot = {pivot}")
    
    for j in range(low, high):
        # If current element is smaller than pivot
        if arr[j] < pivot:
            # Swap arr[i] and arr[j]
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    
    # Place pivot in its correct position
    arr[i], arr[high] = arr[high], arr[i]
    
    return i
````

## Implementation: Iterative QuickSort

````python
def quicksort_iterative(arr: List[int], verbose: bool = False) -> List[int]:
    """
    Sort an array using the iterative QuickSort algorithm with explicit stack.
    
    This version avoids recursion by using an explicit stack to manage
    the subarray boundaries that need to be sorted. This can be useful
    to avoid stack overflow for very large arrays.
    
    Args:
        arr: List of integers to be sorted
        verbose: If True, print each step of the sorting process
    
    Returns:
        The sorted list (sorted in-place)
    
    Time Complexity:
        - Best: O(n log n)
        - Average: O(n log n)
        - Worst: O(n²)
    
    Space Complexity: O(log n) for the explicit stack
    
    Example:
        >>> quicksort_iterative([3, 7, 2, 9, 1, 5])
        [1, 2, 3, 5, 7, 9]
    """
    if len(arr) <= 1:
        return arr
    
    if verbose:
        print(f"Initial array: {arr}")
        print("-" * 50)
    
    # Use explicit stack instead of recursion
    # Stack stores tuples of (low, high) indices
    stack = [(0, len(arr) - 1)]
    iteration = 0
    
    while stack:
        iteration += 1
        low, high = stack.pop()
        
        if verbose:
            print(f"\nIteration {iteration}: Processing [{low}:{high+1}]")
            print(f"  Current subarray: {arr[low:high+1]}")
            print(f"  Stack: {stack}")
        
        if low < high:
            # Partition and get pivot index
            pivot_idx = _partition_iterative(arr, low, high)
            
            if verbose:
                print(f"  Pivot index: {pivot_idx}, Pivot value: {arr[pivot_idx]}")
                print(f"  After partition: {arr}")
            
            # Push left subarray if it has more than one element
            if pivot_idx - 1 > low:
                stack.append((low, pivot_idx - 1))
                if verbose:
                    print(f"  Pushed left subarray [{low}:{pivot_idx}] to stack")
            
            # Push right subarray if it has more than one element
            if pivot_idx + 1 < high:
                stack.append((pivot_idx + 1, high))
                if verbose:
                    print(f"  Pushed right subarray [{pivot_idx+1}:{high+1}] to stack")
    
    if verbose:
        print("-" * 50)
        print(f"Final sorted array: {arr}")
    
    return arr


def _partition_iterative(arr: List[int], low: int, high: int) -> int:
    """
    Partition function for iterative quicksort.
    Uses the first element as pivot (Lomuto-like scheme).
    
    Args:
        arr: The array to partition
        low: Starting index
        high: Ending index
    
    Returns:
        The final position of the pivot
    """
    pivot_value = arr[low]  # Use first element as pivot
    pivot_idx = low
    
    for i in range(low + 1, high + 1):
        if arr[i] <= pivot_value:
            pivot_idx += 1
            arr[i], arr[pivot_idx] = arr[pivot_idx], arr[i]
    
    # Place pivot in its final position
    arr[pivot_idx], arr[low] = arr[low], arr[pivot_idx]
    
    return pivot_idx
````

# 5. Performance Comparison

Let's compare the performance of all sorting algorithms on different array sizes.

````python
def measure_time(sort_func: Callable, arr: List[int], runs: int = 3) -> float:
    """
    Measure the average execution time of a sorting function.
    
    Args:
        sort_func: The sorting function to measure
        arr: The array to sort
        runs: Number of runs to average
    
    Returns:
        Average execution time in seconds
    """
    times = []
    for _ in range(runs):
        arr_copy = arr.copy()
        start = time.time()
        sort_func(arr_copy)
        end = time.time()
        times.append(end - start)
    return np.mean(times)


def run_performance_comparison():
    """
    Run performance comparison for all sorting algorithms.
    """
    # Array sizes to test
    sizes = [100, 500, 1000, 2000, 3000, 4000, 5000]
    
    # Results storage
    results = {
        'Bubble Sort': [],
        'Merge Sort': [],
        'Shell Sort': [],
        'QuickSort (Recursive)': [],
        'QuickSort (Iterative)': []
    }
    
    print("Running performance comparison...")
    print("-" * 60)
    
    for size in sizes:
        # Generate random array
        arr = list(np.random.randint(0, 10000, size))
        print(f"\nTesting with array size: {size}")
        
        # Measure each algorithm
        results['Bubble Sort'].append(measure_time(bubble_sort, arr))
        results['Merge Sort'].append(measure_time(merge_sort, arr))
        results['Shell Sort'].append(measure_time(shell_sort, arr))
        results['QuickSort (Recursive)'].append(measure_time(quicksort_recursive, arr))
        results['QuickSort (Iterative)'].append(measure_time(quicksort_iterative, arr))
        
        # Print results for this size
        print(f"  Bubble Sort: {results['Bubble Sort'][-1]:.6f}s")
        print(f"  Merge Sort: {results['Merge Sort'][-1]:.6f}s")
        print(f"  Shell Sort: {results['Shell Sort'][-1]:.6f}s")
        print(f"  QuickSort (Recursive): {results['QuickSort (Recursive)'][-1]:.6f}s")
        print(f"  QuickSort (Iterative): {results['QuickSort (Iterative)'][-1]:.6f}s")
    
    return sizes, results
````

````python
# Run the comparison
sizes, results = run_performance_comparison()
````

# 6. Summary

## Comprehensive Comparison Table

| Algorithm | Best Time | Average Time | Worst Time | Space | Stable | In-place | Use Case |
|-----------|-----------|--------------|------------|-------|--------|----------|----------|
| **Bubble Sort** | O(n) | O(n²) | O(n²) | O(1) | ✓ Yes | ✓ Yes | Educational, small arrays, nearly sorted |
| **Merge Sort** | O(n log n) | O(n log n) | O(n log n) | O(n) | ✓ Yes | ✗ No | Large datasets, linked lists, external sorting |
| **Shell Sort** | O(n log n) | O(n^1.25) | O(n²) | O(1) | ✗ No | ✓ Yes | Medium arrays, embedded systems |
| **QuickSort** | O(n log n) | O(n log n) | O(n²) | O(log n) | ✗ No | ✓ Yes | General purpose, large datasets |

## Key Takeaways

### Bubble Sort
- **Pros**: Simple, stable, in-place, good for detecting sorted arrays
- **Cons**: Very slow O(n²), impractical for large datasets
- **Best for**: Educational purposes, tiny arrays (< 50 elements)

### Merge Sort
- **Pros**: Guaranteed O(n log n), stable, parallelizable, great for linked lists
- **Cons**: Requires O(n) extra space, not in-place
- **Best for**: Large datasets, external sorting, when stability is required

### Shell Sort
- **Pros**: In-place, good cache performance, better than O(n²) in practice
- **Cons**: Not stable, complex analysis, performance depends on gap sequence
- **Best for**: Medium-sized arrays, embedded systems with limited memory

### QuickSort
- **Pros**: Very fast in practice, in-place, cache-friendly
- **Cons**: O(n²) worst case, not stable, requires good pivot selection
- **Best for**: General-purpose sorting, when average performance matters most
