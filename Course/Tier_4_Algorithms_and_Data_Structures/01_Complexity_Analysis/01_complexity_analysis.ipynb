# 1. Theory Section

## What is Algorithmic Complexity?

**Algorithmic complexity** measures the resources (time and space) an algorithm requires as a function of the input size. It helps us:

- **Compare algorithms** independently of hardware
- **Predict performance** at scale
- **Make informed decisions** about which algorithm to use

### Why Do We Care About Efficiency?

Consider searching through a dataset:

```
┌─────────────────────────────────────────────────────────────────┐
│  Input Size (n)  │  O(n) Linear  │  O(n²) Quadratic           │
├─────────────────────────────────────────────────────────────────┤
│       10         │    10 ops     │      100 ops               │
│      100         │   100 ops     │    10,000 ops              │
│    1,000         │ 1,000 ops     │ 1,000,000 ops              │
│  1,000,000       │   1M ops      │ 1,000,000,000,000 ops  ⚠️  │
└─────────────────────────────────────────────────────────────────┘
```

**At scale, algorithm choice matters more than hardware speed!**

## Big O, Big Ω, and Big Θ Notation

These mathematical notations describe algorithm behavior:

```
╔═══════════════════════════════════════════════════════════════════════╗
║  Notation   │  Name         │  Meaning                               ║
╠═══════════════════════════════════════════════════════════════════════╣
║  O(f(n))    │  Big O        │  Upper bound (worst case)              ║
║             │               │  "grows no faster than f(n)"           ║
╠═══════════════════════════════════════════════════════════════════════╣
║  Ω(f(n))    │  Big Omega    │  Lower bound (best case)               ║
║             │               │  "grows at least as fast as f(n)"      ║
╠═══════════════════════════════════════════════════════════════════════╣
║  Θ(f(n))    │  Big Theta    │  Tight bound (average case)            ║
║             │               │  "grows exactly as fast as f(n)"       ║
╚═══════════════════════════════════════════════════════════════════════╝
```

### Visual Representation

```
    Time │
         │         ╱ O(f(n)) - Upper bound
         │        ╱
         │       ╱
         │    ══╱══ Θ(f(n)) - Tight bound (actual growth)
         │     ╱
         │    ╱
         │   ╱_____ Ω(f(n)) - Lower bound
         │  ╱
         └────────────────────── n
```

### In Practice

We typically use **Big O** because:
1. We want to know the **worst-case** scenario
2. It gives us an **upper bound guarantee**
3. It's the most commonly discussed in interviews and documentation

## Common Complexity Classes

From fastest to slowest:

```
╔════════════════════════════════════════════════════════════════════════════╗
║  Complexity   │  Name           │  Example Operations                     ║
╠════════════════════════════════════════════════════════════════════════════╣
║  O(1)         │  Constant       │  Array access, hash lookup              ║
║  O(log n)     │  Logarithmic    │  Binary search                          ║
║  O(n)         │  Linear         │  Simple loop, linear search             ║
║  O(n log n)   │  Linearithmic   │  Efficient sorting (merge, heap)        ║
║  O(n²)        │  Quadratic      │  Nested loops, bubble sort              ║
║  O(n³)        │  Cubic          │  Triple nested loops                    ║
║  O(2ⁿ)        │  Exponential    │  Recursive Fibonacci, subsets           ║
║  O(n!)        │  Factorial      │  Permutations, traveling salesman       ║
╚════════════════════════════════════════════════════════════════════════════╝
```

### Rule: Drop Constants and Lower-Order Terms

```
O(2n + 5)       →  O(n)
O(n² + n)       →  O(n²)
O(500)          →  O(1)
O(n/2)          →  O(n)
O(n² + n³)      →  O(n³)
```

**Why?** At large n, only the dominant term matters.

## O(1) - Constant Time

The operation takes the same time regardless of input size.

```
Time │
     │ ________________________________
     │
     └────────────────────────────────── n
```

````python
def get_first_element(arr):
    """
    Return the first element of an array.
    
    Time Complexity: O(1) - Direct index access is constant time
    Space Complexity: O(1) - No additional space used
    
    Args:
        arr: List of elements
    Returns:
        First element or None if empty
    """
    # Array index access is O(1) because:
    # - Arrays store elements in contiguous memory
    # - Address calculation: base_address + (index * element_size)
    # - This calculation is constant regardless of array size
    if len(arr) == 0:
        return None
    return arr[0]  # O(1) - same time for array of 10 or 10 million elements


def is_even(number):
    """
    Check if a number is even.
    
    Time Complexity: O(1) - Single arithmetic operation
    Space Complexity: O(1) - No additional space used
    """
    # Modulo operation is O(1) - it's a single CPU instruction
    # The size of the number doesn't affect the operation time
    # (for fixed-width integers)
    return number % 2 == 0


def hash_table_lookup(hash_table, key):
    """
    Look up a value in a hash table.
    
    Time Complexity: O(1) average - Hash computation + direct access
    Space Complexity: O(1) - No additional space used
    
    Note: Worst case is O(n) if all keys hash to the same bucket,
    but with good hash functions this is extremely rare.
    """
    # 1. Compute hash of key - O(1)
    # 2. Access bucket at computed index - O(1)
    # Total: O(1) average case
    return hash_table.get(key)  # Python dict uses hash table internally


# Demonstration
test_array = list(range(1000000))  # 1 million elements
print(f"First element of million-item array: {get_first_element(test_array)}")
print(f"Is 42 even? {is_even(42)}")
````

## O(log n) - Logarithmic Time

Each step eliminates half (or a constant fraction) of the remaining elements.

```
Time │
     │        ·········
     │   ·····
     │  ·
     │ ·
     │·
     └────────────────────────────────── n
```

````python
def binary_search(arr, target):
    """
    Find target in a sorted array using binary search.
    
    Time Complexity: O(log n)
        - Each iteration cuts the search space in half
        - For n elements: log₂(n) iterations maximum
        - Example: 1 million elements → ~20 comparisons max
    
    Space Complexity: O(1) - Only uses a few variables
    
    Args:
        arr: Sorted list of comparable elements
        target: Value to find
    Returns:
        Index of target, or -1 if not found
    """
    left, right = 0, len(arr) - 1
    
    # Loop runs O(log n) times
    # Why? Each iteration: search_space = search_space / 2
    # Starting with n, after k iterations: n / 2^k
    # Stops when n / 2^k = 1, solving: k = log₂(n)
    iterations = 0
    while left <= right:
        iterations += 1
        mid = (left + right) // 2  # O(1) - arithmetic
        
        if arr[mid] == target:      # O(1) - comparison
            print(f"Found in {iterations} iterations (log₂({len(arr)}) ≈ {math.log2(len(arr)):.1f})")
            return mid
        elif arr[mid] < target:
            left = mid + 1  # Eliminate left half
        else:
            right = mid - 1  # Eliminate right half
    
    print(f"Not found after {iterations} iterations")
    return -1


def count_digits(n):
    """
    Count the number of digits in a positive integer.
    
    Time Complexity: O(log n)
        - We divide by 10 each iteration
        - Number of digits ≈ log₁₀(n)
    
    Space Complexity: O(1)
    """
    if n == 0:
        return 1
    
    count = 0
    n = abs(n)
    
    # Each iteration removes one digit (divides by 10)
    # For n with d digits: d = ⌊log₁₀(n)⌋ + 1 iterations
    while n > 0:
        count += 1
        n //= 10  # Remove last digit
    
    return count


# Demonstration
import math
sorted_array = list(range(0, 1000000, 2))  # 500,000 even numbers
print("Binary search for 424242:")
binary_search(sorted_array, 424242)

print(f"\nDigits in 12345678: {count_digits(12345678)}")
````

## O(n) - Linear Time

Time grows directly proportional to input size.

```
Time │
     │                    ╱
     │                  ╱
     │                ╱
     │              ╱
     │            ╱
     │          ╱
     │        ╱
     │      ╱
     │    ╱
     │  ╱
     │╱
     └────────────────────────────────── n
```

````python
def find_maximum(arr):
    """
    Find the maximum element in an unsorted array.
    
    Time Complexity: O(n)
        - Must check every element (could be anywhere)
        - Exactly n comparisons for n elements
    
    Space Complexity: O(1) - Single variable for tracking max
    
    Args:
        arr: Non-empty list of comparable elements
    Returns:
        Maximum element
    """
    if not arr:
        return None
    
    maximum = arr[0]  # O(1)
    
    # Loop executes n-1 times (once per element after first)
    # Each iteration does O(1) work (comparison, possible assignment)
    # Total: O(n)
    for element in arr[1:]:
        if element > maximum:  # O(1) comparison
            maximum = element  # O(1) assignment
    
    return maximum


def linear_search(arr, target):
    """
    Search for target by checking each element.
    
    Time Complexity: O(n)
        - Best case: O(1) - target is first element
        - Worst case: O(n) - target is last or not present
        - Average case: O(n/2) = O(n) - target is in middle
    
    Space Complexity: O(1)
    """
    # In worst case, we examine all n elements
    for i, element in enumerate(arr):
        if element == target:  # O(1) comparison
            return i
    return -1


def sum_array(arr):
    """
    Calculate sum of all elements.
    
    Time Complexity: O(n) - Must visit each element exactly once
    Space Complexity: O(1) - Single accumulator variable
    """
    total = 0  # O(1) space
    
    # n iterations, O(1) work each
    for num in arr:
        total += num  # O(1)
    
    return total


# Demonstration
test_data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
print(f"Maximum: {find_maximum(test_data)}")
print(f"Sum: {sum_array(test_data)}")
print(f"Index of 9: {linear_search(test_data, 9)}")
````

## O(n log n) - Linearithmic Time

Slightly worse than linear, but much better than quadratic. Common in efficient sorting algorithms.

```
Time │
     │                          ╱
     │                       ╱╱
     │                    ╱╱
     │                 ╱╱
     │              ╱╱
     │           ╱╱
     │        ╱╱
     │     ╱╱     ← grows faster than O(n) but slower than O(n²)
     │  ╱╱
     │╱╱
     └────────────────────────────────── n
```

````python
def merge_sort(arr):
    """
    Sort array using merge sort algorithm.
    
    Time Complexity: O(n log n)
        - Divides array in half: log n levels of recursion
        - At each level: O(n) work to merge
        - Total: O(n) × O(log n) = O(n log n)
    
    Space Complexity: O(n) - Need auxiliary array for merging
    
    Visualization of recursion levels:
    
    Level 0 (n elements):     [████████████████]           n work
                                     ↓
    Level 1 (n/2 each):      [████████] [████████]         n work  
                                ↓           ↓
    Level 2 (n/4 each):     [████] [████] [████] [████]    n work
                               ↓     ↓      ↓      ↓
    ...continues for log n levels...
    
    Total: n work × log n levels = O(n log n)
    """
    if len(arr) <= 1:
        return arr
    
    # Divide step: O(1) to find midpoint
    mid = len(arr) // 2
    
    # Recursively sort halves
    # T(n) = 2·T(n/2) + O(n)  ← recurrence relation
    left = merge_sort(arr[:mid])    # T(n/2)
    right = merge_sort(arr[mid:])   # T(n/2)
    
    # Merge step: O(n) to combine two sorted halves
    return merge(left, right)


def merge(left, right):
    """
    Merge two sorted arrays into one sorted array.
    
    Time Complexity: O(n) where n = len(left) + len(right)
    Space Complexity: O(n) for the result array
    """
    result = []
    i = j = 0
    
    # Each iteration processes one element, total n iterations
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # Append remaining elements
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result


# Demonstration
unsorted = [38, 27, 43, 3, 9, 82, 10]
print(f"Original: {unsorted}")
print(f"Sorted:   {merge_sort(unsorted)}")
````

## O(n²) - Quadratic Time

Often results from nested loops where each element is compared with every other element.

```
Time │
     │                                    *
     │                              *
     │                        *
     │                  *
     │            *
     │       *
     │   *
     │ *
     │*
     └────────────────────────────────── n
```

````python
def bubble_sort(arr):
    """
    Sort array using bubble sort algorithm.
    
    Time Complexity: O(n²)
        - Outer loop: n iterations
        - Inner loop: n-1, n-2, ..., 1 iterations
        - Total comparisons: n(n-1)/2 = O(n²)
    
    Space Complexity: O(1) - In-place sorting
    """
    arr = arr.copy()  # Don't modify original
    n = len(arr)
    comparisons = 0
    
    # Outer loop: n passes
    for i in range(n):
        # Inner loop: compare adjacent elements
        # Each pass moves the largest unsorted element to its place
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap
    
    print(f"  Total comparisons: {comparisons} (n²/2 ≈ {n*n//2})")
    return arr


def has_duplicate_naive(arr):
    """
    Check if array has any duplicate elements using nested loops.
    
    Time Complexity: O(n²)
        - Compare every pair of elements
        - n × (n-1) / 2 comparisons
    
    Space Complexity: O(1)
    
    Note: Can be done in O(n) with a hash set!
    """
    n = len(arr)
    
    # For each element...
    for i in range(n):
        # ...compare with every element after it
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                return True
    
    return False


def print_all_pairs(arr):
    """
    Print all pairs of elements.
    
    Time Complexity: O(n²) - Nested loops, both iterate n times
    Space Complexity: O(1)
    """
    pairs = []
    # n iterations
    for i in arr:
        # n iterations for each outer iteration
        for j in arr:
            pairs.append((i, j))  # n × n = n² pairs
    
    return pairs


# Demonstration
test_arr = [64, 34, 25, 12, 22, 11, 90]
print(f"Bubble sort of {test_arr}:")
print(f"  Result: {bubble_sort(test_arr)}")
print(f"\nPairs of [1,2,3]: {print_all_pairs([1,2,3])}")
````

## O(2ⁿ) - Exponential Time

Time doubles with each additional input element. Common in brute-force algorithms.

```
Time │
     │                           ⁂
     │                       ⁂
     │                   ⁂
     │               ⁂
     │           ⁂
     │       ⁂
     │    ⁂
     │  ⁂
     │ ⁂
     │⁂
     └────────────────────────────────── n
            5   10   15   20   (gets huge fast!)
```

**Warning:** Exponential algorithms become unusable very quickly!

````python
def fibonacci_naive(n):
    """
    Calculate nth Fibonacci number using naive recursion.
    
    Time Complexity: O(2ⁿ)
        - Each call branches into TWO more calls
        - Creates a binary tree of calls
        - Tree has roughly 2ⁿ nodes
    
    Space Complexity: O(n) - Maximum recursion depth
    
    Call tree for fib(5):
    
                         fib(5)
                        /      \\
                    fib(4)      fib(3)
                   /     \\      /     \\
               fib(3)  fib(2) fib(2) fib(1)
               /    \\
           fib(2) fib(1)
    
    Notice: fib(3) and fib(2) are computed multiple times!
    """
    if n <= 1:
        return n
    
    # Two recursive calls = exponential growth
    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2)


def generate_all_subsets(arr):
    """
    Generate all possible subsets of an array.
    
    Time Complexity: O(2ⁿ)
        - Each element has 2 choices: include or exclude
        - Total subsets: 2 × 2 × 2 × ... (n times) = 2ⁿ
    
    Space Complexity: O(2ⁿ) - Storing all subsets
    
    For [1, 2, 3]:
        Element 1: include/exclude → 2 choices
        Element 2: include/exclude → 2 choices  
        Element 3: include/exclude → 2 choices
        Total: 2 × 2 × 2 = 8 subsets
    """
    result = []
    n = len(arr)
    
    # Iterate through all 2ⁿ possible combinations
    for i in range(2 ** n):  # 2ⁿ iterations
        subset = []
        for j in range(n):   # n iterations each
            # Check if jth bit is set in i
            if i & (1 << j):
                subset.append(arr[j])
        result.append(subset)
    
    return result


# Demonstration
print("Fibonacci (naive):")
for i in range(10):
    print(f"  fib({i}) = {fibonacci_naive(i)}")

print(f"\nAll subsets of [1, 2, 3]:")
for subset in generate_all_subsets([1, 2, 3]):
    print(f"  {subset}")
````

## O(n!) - Factorial Time

The most expensive common complexity. Grows incredibly fast.

```
n     n!
─────────────────
1     1
5     120
10    3,628,800
15    1,307,674,368,000
20    2,432,902,008,176,640,000  ← ~2.4 quintillion!
```

````python
def generate_permutations(arr):
    """
    Generate all permutations of an array.
    
    Time Complexity: O(n!)
        - First position: n choices
        - Second position: n-1 choices
        - Third position: n-2 choices
        - ... and so on
        - Total: n × (n-1) × (n-2) × ... × 1 = n!
    
    Space Complexity: O(n!) - Storing all permutations
                    + O(n) - Recursion depth
    
    For [1, 2, 3]:
        3 × 2 × 1 = 6 permutations
    
    For [1, 2, 3, 4]:
        4 × 3 × 2 × 1 = 24 permutations
    """
    if len(arr) <= 1:
        return [arr.copy()]
    
    result = []
    
    # Try each element as the first element
    for i in range(len(arr)):  # n choices for first position
        # Get the remaining elements
        current = arr[i]
        remaining = arr[:i] + arr[i+1:]
        
        # Recursively get permutations of remaining
        for perm in generate_permutations(remaining):  # (n-1)! permutations
            result.append([current] + perm)
    
    return result


def traveling_salesman_brute_force(distances):
    """
    Solve TSP by checking all possible routes (brute force).
    
    Time Complexity: O(n!) 
        - Check all n! permutations of cities
        - Each permutation takes O(n) to calculate distance
        - Total: O(n! × n)
    
    Space Complexity: O(n) for storing current best path
    
    This is why TSP is NP-hard!
    For 20 cities: ~2.4 quintillion routes to check
    """
    n = len(distances)
    cities = list(range(n))
    
    min_distance = float('inf')
    best_path = None
    
    # Check all n! permutations
    for perm in generate_permutations(cities):
        # Calculate total distance for this route
        total = 0
        for i in range(n - 1):
            total += distances[perm[i]][perm[i + 1]]
        total += distances[perm[-1]][perm[0]]  # Return to start
        
        if total < min_distance:
            min_distance = total
            best_path = perm
    
    return best_path, min_distance


# Demonstration
print("All permutations of [1, 2, 3]:")
for perm in generate_permutations([1, 2, 3]):
    print(f"  {perm}")

print(f"\nTotal permutations: {len(generate_permutations([1, 2, 3, 4]))} for n=4")

# Simple TSP example
distances = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
path, dist = traveling_salesman_brute_force(distances)
print(f"\nTSP best path: {path}, distance: {dist}")
````

# 4. Complexity Analysis Practice

## Analyzing Loops

### Single Loop

```python
for i in range(n):     # Runs n times
    do_something()     # O(1) operation
# Total: O(n)
```

### Nested Loops (Dependent)

```python
for i in range(n):           # n times
    for j in range(n):       # n times for each i
        do_something()       # O(1)
# Total: O(n × n) = O(n²)
```

### Nested Loops (Different Variables)

```python
for i in range(n):           # n times
    for j in range(m):       # m times for each i
        do_something()       # O(1)
# Total: O(n × m)
```

### Sequential Loops

```python
for i in range(n):     # O(n)
    do_a()

for j in range(n):     # O(n)
    do_b()
# Total: O(n) + O(n) = O(2n) = O(n)
```

### Loop with Logarithmic Iterations

```python
i = n
while i > 0:
    do_something()     # O(1)
    i = i // 2         # Halving each time
# Total: O(log n)
```

````python
# Practice: Analyze these functions

def mystery_1(n):
    """
    What is the time complexity?
    
    Analysis:
    - Outer loop: i goes 1, 2, 4, 8, ... until i >= n
    - This means i doubles each time → log n iterations
    - Inner loop: runs i times each iteration
    - Total work: 1 + 2 + 4 + 8 + ... + n/2 = n - 1 ≈ O(n)
    
    Time Complexity: O(n)  (not O(n log n)!)
    """
    count = 0
    i = 1
    while i < n:
        for j in range(i):
            count += 1
        i *= 2
    return count


def mystery_2(n):
    """
    What is the time complexity?
    
    Analysis:
    - Outer loop: n iterations
    - Inner loop: j starts at i, doubles until >= n
    - Inner iterations: log(n/i) for each i
    - Sum over all i: approximately O(n) total
    
    Time Complexity: O(n)  
    """
    count = 0
    for i in range(1, n + 1):
        j = i
        while j < n:
            count += 1
            j *= 2
    return count


def mystery_3(n):
    """
    What is the time complexity?
    
    Analysis:
    - Outer loop: n iterations
    - Inner loop: runs n-i times (decreases as i increases)
    - Total: (n-1) + (n-2) + ... + 1 + 0 = n(n-1)/2
    
    Time Complexity: O(n²)
    """
    count = 0
    for i in range(n):
        for j in range(i, n):
            count += 1
    return count


# Verify our analysis
print("Verifying complexity analysis:")
print("\nmystery_1 (should be ~O(n)):")
for n in [10, 100, 1000]:
    print(f"  n={n:4d}: count={mystery_1(n):6d}, n={n}")

print("\nmystery_2 (should be ~O(n)):")
for n in [10, 100, 1000]:
    print(f"  n={n:4d}: count={mystery_2(n):6d}, n*log(n)≈{int(n*math.log2(n))}")

print("\nmystery_3 (should be ~O(n²)):")
for n in [10, 100, 1000]:
    print(f"  n={n:4d}: count={mystery_3(n):7d}, n²/2={n*n//2}")
````

## Analyzing Recursion

### Recurrence Relations

Express the time complexity as a recurrence:

```
┌────────────────────────────────────────────────────────────────────┐
│  Pattern                 │  Recurrence       │  Solution          │
├────────────────────────────────────────────────────────────────────┤
│  Linear recursion        │  T(n) = T(n-1)+1  │  O(n)              │
│  Binary recursion        │  T(n) = 2T(n-1)+1 │  O(2ⁿ)             │
│  Divide & conquer        │  T(n) = 2T(n/2)+n │  O(n log n)        │
│  Binary search           │  T(n) = T(n/2)+1  │  O(log n)          │
│  Three-way split         │  T(n) = 3T(n/3)+n │  O(n log n)        │
└────────────────────────────────────────────────────────────────────┘
```

### The Master Theorem (Simplified)

For recurrences of form: **T(n) = aT(n/b) + f(n)**

Where:
- `a` = number of subproblems
- `n/b` = size of each subproblem  
- `f(n)` = work done outside recursion

````python
# Recursion analysis examples

def factorial_recursive(n, depth=0):
    """
    Calculate n! recursively.
    
    Recurrence: T(n) = T(n-1) + O(1)
                T(0) = O(1)
    
    Solution: T(n) = O(n)
    
    Call stack visualization for factorial(4):
    
    factorial(4)           ← waiting
      └─ factorial(3)      ← waiting
           └─ factorial(2) ← waiting
                └─ factorial(1)  ← waiting
                     └─ factorial(0) = 1  ← base case!
                        returns 1
                     returns 1 × 1 = 1
                returns 2 × 1 = 2
           returns 3 × 2 = 6
      returns 4 × 6 = 24
    
    Time Complexity: O(n) - n recursive calls
    Space Complexity: O(n) - n stack frames
    """
    indent = "  " * depth
    print(f"{indent}factorial({n})")
    
    if n <= 1:
        return 1
    
    return n * factorial_recursive(n - 1, depth + 1)


def sum_recursive(arr, depth=0):
    """
    Sum array elements using divide and conquer.
    
    Recurrence: T(n) = 2T(n/2) + O(1)
    Solution: T(n) = O(n)
    
    This divides the problem but does minimal work at each level,
    so total work is just proportional to number of elements.
    
    Time Complexity: O(n)
    Space Complexity: O(log n) - recursion depth
    """
    indent = "  " * depth
    
    if len(arr) == 0:
        return 0
    if len(arr) == 1:
        print(f"{indent}sum({arr}) = {arr[0]}")
        return arr[0]
    
    mid = len(arr) // 2
    print(f"{indent}sum({arr[:mid]}) + sum({arr[mid:]})")
    
    left_sum = sum_recursive(arr[:mid], depth + 1)
    right_sum = sum_recursive(arr[mid:], depth + 1)
    
    return left_sum + right_sum


print("Factorial recursion (O(n)):")
result = factorial_recursive(5)
print(f"Result: {result}\n")

print("Divide and conquer sum (O(n)):")
result = sum_recursive([1, 2, 3, 4, 5, 6, 7, 8])
print(f"Result: {result}")
````

# 5. Space Complexity

Space complexity measures the memory an algorithm uses relative to input size.

## Types of Space

```
┌─────────────────────────────────────────────────────────────┐
│                      TOTAL SPACE                            │
│  ┌──────────────────┐  ┌─────────────────────────────────┐ │
│  │   Input Space    │  │     Auxiliary Space             │ │
│  │                  │  │                                 │ │
│  │ (the input data) │  │  (extra space for processing)   │ │
│  │                  │  │  - Variables                    │ │
│  │  Usually O(n)    │  │  - Data structures              │ │
│  │                  │  │  - Call stack (recursion)       │ │
│  └──────────────────┘  └─────────────────────────────────┘ │
└─────────────────────────────────────────────────────────────┘
```

**Note:** We typically analyze **auxiliary space** (extra space beyond input).

````python
def constant_space_example(arr):
    """
    Find sum using O(1) auxiliary space.
    
    Time Complexity: O(n)
    Auxiliary Space: O(1) - only uses fixed number of variables
    Total Space: O(n) for input + O(1) auxiliary
    """
    total = 0           # 1 variable: O(1)
    for num in arr:     # Iterator variable: O(1)
        total += num
    return total


def linear_space_example(arr):
    """
    Create a new array with doubled values.
    
    Time Complexity: O(n)
    Auxiliary Space: O(n) - creates new array of same size
    """
    # Creating a new list of size n
    doubled = []        # Will grow to size n: O(n)
    for num in arr:
        doubled.append(num * 2)
    return doubled


def quadratic_space_example(n):
    """
    Create an n×n matrix.
    
    Time Complexity: O(n²)
    Auxiliary Space: O(n²) - matrix has n² elements
    """
    # 2D array of size n×n
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            row.append(i * n + j)
        matrix.append(row)
    return matrix


# Demonstration
test = [1, 2, 3, 4, 5]
print(f"O(1) space - Sum: {constant_space_example(test)}")
print(f"O(n) space - Doubled: {linear_space_example(test)}")
print(f"O(n²) space - Matrix (3×3):")
for row in quadratic_space_example(3):
    print(f"  {row}")
````

## Stack Space for Recursion

Each recursive call adds a frame to the call stack:

```
Call Stack for factorial(5):

┌────────────────┐
│ factorial(1)   │ ← top of stack (most recent)
├────────────────┤
│ factorial(2)   │
├────────────────┤
│ factorial(3)   │
├────────────────┤
│ factorial(4)   │
├────────────────┤
│ factorial(5)   │ ← bottom of stack (first call)
└────────────────┘

Maximum depth: n frames = O(n) space
```

````python
import sys

def recursive_linear_space(n, depth=0):
    """
    Linear recursion - O(n) stack space.
    
    Each call adds one frame, max depth is n.
    """
    if depth == 0:
        print(f"Recursion depth for n={n}:")
    
    if n <= 0:
        print(f"  Max depth reached: {depth}")
        return 0
    
    return 1 + recursive_linear_space(n - 1, depth + 1)


def recursive_log_space(arr, depth=0):
    """
    Divide and conquer - O(log n) stack space.
    
    Binary search pattern: each call halves the problem.
    Maximum stack depth is log₂(n).
    """
    if len(arr) <= 1:
        return depth
    
    mid = len(arr) // 2
    # Only ONE recursive call at a time (like binary search)
    return recursive_log_space(arr[:mid], depth + 1)


def tail_recursive_sum(arr, acc=0):
    """
    Tail recursion - can be O(1) with optimization.
    
    Tail recursion: recursive call is the LAST operation.
    Some languages optimize this to O(1) space.
    Python does NOT optimize tail recursion (still O(n) space).
    
    Time: O(n)
    Space: O(n) in Python, O(1) in optimizing languages
    """
    if not arr:
        return acc
    
    # Tail recursive: nothing happens after the recursive call
    return tail_recursive_sum(arr[1:], acc + arr[0])


# Demonstration
recursive_linear_space(5)

test_arr = list(range(1024))  # 1024 elements
max_depth = recursive_log_space(test_arr)
print(f"\nLog space recursion on 1024 elements: max depth = {max_depth} (log₂(1024) = 10)")

print(f"\nTail recursive sum of [1,2,3,4,5]: {tail_recursive_sum([1,2,3,4,5])}")
````

## Rules of Thumb

```
┌─────────────────────────────────────────────────────────────────────────────┐
│                        COMPLEXITY ANALYSIS RULES                            │
├─────────────────────────────────────────────────────────────────────────────┤
│  1. DROP CONSTANTS        │  O(2n) → O(n), O(100) → O(1)                   │
│  2. DROP LOWER TERMS      │  O(n² + n) → O(n²)                             │
│  3. DIFFERENT INPUTS      │  O(a + b), NOT O(n) if two different arrays    │
│  4. SEQUENTIAL = ADD      │  O(A) then O(B) = O(A + B)                     │

│  6. LOG BASE DOESN'T      │  log₂(n) and log₁₀(n) are both O(log n)       │

└─────────────────────────────────────────────────────────────────────────────┘
```

## Common Patterns

```
┌─────────────────────────────────────────────────────────────────────────────┐
│  Pattern                              │  Complexity                         │
├─────────────────────────────────────────────────────────────────────────────┤
│  Single loop (1 to n)                 │  O(n)                               │
│  Loop halving each iteration          │  O(log n)                           │
│  Two nested loops (both 1 to n)       │  O(n²)                              │
│  Loop inside loop (i to n, j to n)    │  O(n²)                              │
│  Loop inside loop (j from i to n)     │  O(n²) (still!)                     │
│  Two sequential loops                 │  O(n + n) = O(n)                    │
│  Recursion that halves input          │  O(log n)                           │
│  Recursion that decrements by 1       │  O(n)                               │
│  Recursion with 2 calls, decrement    │  O(2ⁿ)                              │
│  Recursion: 2 calls, halves input     │  O(n) for traversal                 │
│                                       │  O(n log n) if O(n) work per level  │
└─────────────────────────────────────────────────────────────────────────────┘
```
