## Module overview

This module covers probability theory and statistical computing with Python, split across two focused notebooks:

- **[01_probability.ipynb](./01_probability.ipynb)** — Probability distributions, the scipy.stats API, Bayesian concepts, and their applications to sequencing data (Poisson read counts, Negative Binomial RNA-seq models)
- **[02_statistics_python.ipynb](./02_statistics_python.ipynb)** — Hypothesis testing, regression, multiple testing correction, and power analysis using Python

Start with `01_probability.ipynb` if you are new to probability distributions. If you are comfortable with distributions but need statistical testing, go directly to `02_statistics_python.ipynb`.


## Prerequisites

- Python basics (variables, loops, functions)
- Basic descriptive statistics (mean, standard deviation, median) — covered in Module 0.6 Biostatistics
- `numpy` and `matplotlib` familiarity helpful but not required

Install required packages if needed:
```bash
conda install numpy scipy matplotlib pandas statsmodels
```


## What you will be able to do after this module

- Use `scipy.stats` to work with any probability distribution (compute probabilities, quantiles, generate samples)
- Apply the right statistical test to the right biological question
- Interpret p-values, confidence intervals, and effect sizes correctly
- Apply multiple testing correction to genomics-scale analyses
- Estimate required sample sizes before running an experiment
