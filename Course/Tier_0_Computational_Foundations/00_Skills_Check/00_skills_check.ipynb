### Answer 10

- **`git fetch`** downloads new commits from the remote but does NOT change your
  working files. It updates your remote-tracking branches (e.g., `origin/main`).
- **`git pull`** does `git fetch` + `git merge` -- it downloads AND integrates
  the remote changes into your current branch.

**When to use `fetch`:** When you want to see what changed on the remote before
deciding to merge (safer in collaborative settings).

**When to use `pull`:** When you trust the remote changes and want to update quickly.

**Award 1 point** if you explained that `fetch` only downloads while `pull` also merges.

### Answer 11

`set -euo pipefail` enables strict error handling:

- **`-e`** (errexit): Exit immediately if any command returns a non-zero exit status
- **`-u`** (nounset): Treat references to unset variables as errors
- **`-o pipefail`**: If any command in a pipeline fails, the whole pipeline's exit
  status is the failing command's status (not just the last command's)

Without this, Bash silently continues after errors -- dangerous in data pipelines.

**Award 1 point** if you correctly explained all three flags.

### Answer 13

Bugs in the script:

1. **Spaces around `=` in assignment:** `input_dir = $1` should be `input_dir=$1`
   (Bash interprets this as running a command called `input_dir` with arguments `=` and `$1`)

2. **Unquoted variables:** `$input_dir/*.bam` and `$file` should be `"$input_dir"/*.bam`
   and `"$file"` -- paths with spaces will break without quotes.

3. **Single brackets `[ ]` instead of `[[ ]]`:** While `[ ]` works, `[[ ]]` is safer
   and more portable in Bash. (This one is a style issue, not strictly a bug.)

4. **No `set -euo pipefail`:** If `samtools index` fails on one file, the script
   silently continues.

**Award 1 point** if you identified at least 2 actual bugs (the spaces around `=` and
unquoted variables are the critical ones).
