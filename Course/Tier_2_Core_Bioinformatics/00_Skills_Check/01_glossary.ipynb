## A

**ACMG (American College of Medical Genetics and Genomics)**
Organization that publishes guidelines for classifying genetic variants into five categories: pathogenic, likely pathogenic, variant of uncertain significance (VUS), likely benign, and benign. These classifications are central to clinical genomics reporting.
*Tier 3 > 11_Clinical_Genomics*

**Alignment (sequence alignment)**
The process of arranging two or more biological sequences (DNA, RNA, or protein) to identify regions of similarity. Similarity may reflect functional, structural, or evolutionary relationships. Pairwise alignment compares two sequences; multiple sequence alignment (MSA) handles three or more.
*Tier 2 > 03_Pairwise_Sequence_Alignment, 05_Multiple_Sequence_Alignment*

**ASCII (American Standard Code for Information Interchange)**
A 7-bit character encoding standard that maps 128 characters (letters, digits, control codes) to integer values 0--127. ASCII is the foundation for text-based bioinformatics file formats such as FASTA and FASTQ. Phred quality scores in FASTQ files are encoded as ASCII characters offset by 33.
*Tier 0 > 04_File_Encodings*

**ASV (Amplicon Sequence Variant)**
An exact DNA sequence obtained from high-throughput marker-gene analysis after removing sequencing errors, without the clustering step used for OTUs. ASVs provide single-nucleotide resolution of microbial community composition and are generated by tools such as DADA2.
*Tier 3 > 04_Microbial_Diversity*

---
## B

**BAM (Binary Alignment Map)**
The compressed binary version of a SAM file, storing read alignments against a reference genome. BAM files are indexed (`.bai`) for efficient random access and are the standard format for downstream analysis tools such as variant callers.
*Tier 3 > 01_NGS_Fundamentals, 02_Variant_Calling_and_SNP_Analysis*

**Bash**
The Bourne Again SHell, the default command-line interpreter on most Linux systems. Bash scripting enables automation of bioinformatics pipelines by chaining commands together with pipes, redirects, loops, and conditionals.
*Tier 0 > 03_Bash_Scripting*

**Beta diversity**
A measure of the difference in species composition between two communities or samples. Common beta diversity metrics include Bray-Curtis dissimilarity, UniFrac distance, and Jaccard index. Often visualized via PCoA or NMDS ordination plots.
*Tier 3 > 04_Microbial_Diversity*

**BLAST (Basic Local Alignment Search Tool)**
A heuristic algorithm for finding regions of local similarity between a query sequence and a database. BLAST comes in several variants: blastn (nucleotide vs. nucleotide), blastp (protein vs. protein), blastx (translated nucleotide vs. protein), tblastn (protein vs. translated nucleotide), and tblastx (translated nucleotide vs. translated nucleotide).
*Tier 2 > 04_BLAST_Searching*

**BLOSUM (BLOcks SUbstitution Matrix)**
A family of substitution matrices derived from observed amino acid substitutions in conserved protein blocks. The number (e.g., BLOSUM62) indicates the clustering threshold: higher numbers are designed for more closely related sequences. BLOSUM62 is the default for most BLAST protein searches.
*Tier 2 > 03_Pairwise_Sequence_Alignment, 04_BLAST_Searching*

---
## C

**CADD (Combined Annotation-Dependent Depletion)**
A tool that integrates multiple annotations into a single score (Phred-scaled) to predict the deleteriousness of genetic variants. A CADD score above 20 suggests a variant is among the top 1% most deleterious substitutions in the genome.
*Tier 3 > 11_Clinical_Genomics*

**Clade**
A group on a phylogenetic tree that includes a common ancestor and all of its descendants. A clade is monophyletic by definition. Identifying clades helps determine evolutionary relationships among organisms or gene families.
*Tier 2 > 06_Phylogenetics*

**Class (Python)**
A blueprint for creating objects that bundles data (attributes) and behavior (methods) together. In bioinformatics, BioPython's `Seq`, `SeqRecord`, and `SeqFeature` are important classes that model biological sequences and their annotations.
*Tier 1 > 13_Classes_and_OOP*

**ClinVar**
An NCBI database that archives the relationship between human genetic variants and clinical phenotypes, with supporting evidence. Submitters include clinical laboratories, research groups, and expert panels. Each variant is assigned a clinical significance classification.
*Tier 3 > 11_Clinical_Genomics*

**CNV (Copy Number Variation)**
A structural variant in which sections of the genome are repeated (duplicated) or deleted, resulting in a different number of copies compared to the reference. CNVs range from kilobases to megabases and can be detected from read-depth analysis of NGS data.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis, 11_Clinical_Genomics*

**Codon**
A triplet of three consecutive nucleotides in mRNA that specifies a particular amino acid or a stop signal during translation. There are 64 possible codons encoding 20 amino acids and 3 stop signals, making the code degenerate (redundant).
*Tier 2 > 02_BioPython_Essentials*

**Comprehension (Python)**
A concise syntax for creating lists, dictionaries, sets, or generators from iterables with optional filtering. For example, `[len(s) for s in sequences if len(s) > 100]` filters and transforms a list in one expression.
*Tier 1 > 10_Comprehensions*

**Conserved (sequence conservation)**
A nucleotide or amino acid position (or region) that has remained unchanged or nearly unchanged across species over evolutionary time, implying functional or structural importance. Conservation is revealed by multiple sequence alignment.
*Tier 2 > 05_Multiple_Sequence_Alignment, 10_Sequence_Motifs_and_Domains*

**Context manager (Python)**
An object that defines `__enter__` and `__exit__` methods, used with the `with` statement to manage resources such as file handles or database connections. Ensures proper cleanup even if an exception occurs.
*Tier 1 > 14_Decorators_and_Context_Managers*

**Contig**
A contiguous stretch of DNA sequence assembled from overlapping reads without gaps. In genome assembly, contigs are intermediate products that are later joined into scaffolds. The quality of an assembly can be summarized by metrics such as N50.
*Tier 3 > 01_NGS_Fundamentals*

**Coverage (sequencing depth)**
The average number of times each nucleotide in the genome is read during sequencing. For example, 30x coverage means each base is covered by approximately 30 reads on average. Higher coverage improves variant-calling confidence.
*Tier 3 > 01_NGS_Fundamentals, 02_Variant_Calling_and_SNP_Analysis*

**Cryo-EM (Cryo-Electron Microscopy)**
A structural biology technique that images flash-frozen biological molecules using an electron microscope, without the need for crystallization. Recent advances (the "resolution revolution") allow near-atomic resolution. Particularly useful for large complexes and membrane proteins.
*Tier 2 > 07_Protein_Structure; Tier 3 > 09_Molecular_Modeling_and_Docking*

---
## D

**DALI (Distance-matrix ALIgnment)**
A method and server for comparing protein structures based on distance matrices. DALI identifies structural neighbors even when sequence similarity is undetectable, making it valuable for detecting remote homology at the fold level.
*Tier 2 > 07_Protein_Structure*

**dbSNP**
The NCBI database of single nucleotide polymorphisms and other short genetic variations. Each variant is assigned an "rs" identifier (e.g., rs334 for sickle-cell variant). dbSNP is used for annotating VCF files and filtering known variants from novel ones.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis, 11_Clinical_Genomics*

**Decorator (Python)**
A function that wraps another function to modify or extend its behavior without changing its source code. Applied using the `@decorator` syntax. Common uses in bioinformatics include caching expensive computations (`@lru_cache`) and timing function execution.
*Tier 1 > 14_Decorators_and_Context_Managers*

**DESeq2**
An R/Bioconductor package for differential gene expression analysis from count-based RNA-seq data. DESeq2 uses a negative binomial model with shrinkage estimation for dispersion and fold change, providing robust statistical testing for low-replicate experiments.
*Tier 3 > 03_RNA_seq_Analysis*

**Dictionary (Python)**
A mutable, unordered collection of key-value pairs with O(1) average lookup time. Dictionaries are ubiquitous in bioinformatics for codon tables, amino acid property lookups, sequence ID-to-data mappings, and counting k-mer frequencies.
*Tier 1 > 09_Dictionaries_and_Sets*

**Docking (molecular docking)**
A computational method that predicts the preferred orientation of one molecule (typically a small-molecule ligand) when bound to another (typically a protein receptor). Used in structure-based drug design to screen compound libraries.
*Tier 3 > 09_Molecular_Modeling_and_Docking*

**Domain (protein domain)**
A conserved, independently folding structural and functional unit within a protein. Domains are often encoded by one or a few exons and can be found in multiple unrelated proteins. Databases such as Pfam and InterPro catalog known domains.
*Tier 2 > 10_Sequence_Motifs_and_Domains*

---
## E

**E-value (Expect value)**
The number of alignments with an equal or better score expected to occur by chance in a database of a given size. Lower E-values indicate more statistically significant hits. An E-value of 1e-5 means one would expect one hit this good by chance per 100,000 database searches.
*Tier 2 > 04_BLAST_Searching*

**Encoding (character encoding)**
A system that maps characters to numeric values for storage and transmission. Bioinformatics files typically use ASCII or UTF-8. Incorrect encoding assumptions are a common source of file-parsing errors.
*Tier 0 > 04_File_Encodings*

---
## F

**f-string (formatted string literal)**
A Python string prefixed with `f` that allows inline expression evaluation using `{expression}` syntax. Introduced in Python 3.6, f-strings are the preferred way to format output, e.g., `f"GC content: {gc:.2%}"`.
*Tier 1 > 04_Strings_and_Sequences*

**FASTA**
A plain-text format for representing nucleotide or amino acid sequences. Each entry has a header line starting with `>` followed by a description, and one or more lines of sequence. FASTA is the most common input format for BLAST and alignment tools.
*Tier 2 > 01_Biological_Databases, 02_BioPython_Essentials*

**FASTQ**
A text-based format for storing both nucleotide sequences and their per-base quality scores from high-throughput sequencing instruments. Each read occupies four lines: header (`@`), sequence, separator (`+`), and ASCII-encoded quality scores.
*Tier 3 > 01_NGS_Fundamentals*

**FDR (False Discovery Rate)**
The expected proportion of false positives among all results declared significant. In genomics, FDR correction (e.g., Benjamini-Hochberg procedure) is essential when performing thousands of simultaneous statistical tests, such as in differential expression or GWAS.
*Tier 3 > 06_Statistics_for_Bioinformatics*

**Fold change**
The ratio of expression levels between two conditions (e.g., treated vs. control). Typically reported on a log2 scale: a log2 fold change of 1 means the gene is expressed twice as much, while -1 means half as much. Central to differential expression analysis.
*Tier 3 > 03_RNA_seq_Analysis*

**Force field**
A set of mathematical functions and parameters used in molecular dynamics simulations to calculate the potential energy of a molecular system. Common force fields include AMBER, CHARMM, and OPLS. Parameters cover bonds, angles, dihedrals, and non-bonded interactions.
*Tier 3 > 09_Molecular_Modeling_and_Docking*

---
## G

**GATK (Genome Analysis Toolkit)**
A widely used software suite developed by the Broad Institute for variant discovery in high-throughput sequencing data. The GATK Best Practices pipeline includes steps for read mapping, duplicate marking, base quality recalibration, and variant calling (HaplotypeCaller).
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis*

**GenBank**
The NIH/NCBI annotated nucleotide sequence database, one of the three members of the International Nucleotide Sequence Database Collaboration (INSDC). GenBank records contain sequence data plus rich biological annotations (features, references, taxonomy). The GenBank flat-file format uses features such as `gene`, `CDS`, `mRNA`, and `exon`.
*Tier 2 > 01_Biological_Databases, 02_BioPython_Essentials*

**Generator (Python)**
A function that uses `yield` instead of `return` to produce a series of values lazily (on demand) rather than storing them all in memory. Generators are ideal for processing large FASTA/FASTQ files one record at a time.
*Tier 1 > 11_Iterators_and_Generators*

**GFF (General Feature Format)**
A tab-delimited text format for describing genes and other features on a genomic sequence. GFF3 is the current standard, with nine columns: seqid, source, type, start, end, score, strand, phase, and attributes. Used by genome browsers and annotation pipelines.
*Tier 2 > 01_Biological_Databases, 12_Comparative_Genomics*

**Git**
A distributed version control system for tracking changes in source code and other text files. Essential for reproducible bioinformatics: every analysis script, pipeline, and notebook should be under version control.
*Tier 0 > 02_Git_Version_Control*

**gnomAD (Genome Aggregation Database)**
A resource that aggregates exome and genome sequencing data from large-scale studies to provide allele frequency information across diverse populations. Used to filter common variants and identify rare disease-causing mutations.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis, 11_Clinical_Genomics*

**GO (Gene Ontology)**
A structured, controlled vocabulary of terms describing gene product functions, organized into three domains: Biological Process, Molecular Function, and Cellular Component. GO term enrichment analysis identifies overrepresented functions in a gene set.
*Tier 2 > 11_Gene_Ontology_and_Pathways*

**GROMACS**
A high-performance open-source molecular dynamics simulation package optimized for biomolecular systems (proteins, lipids, nucleic acids). Widely used for studying protein folding, ligand binding, and membrane dynamics.
*Tier 3 > 09_Molecular_Modeling_and_Docking*

**GWAS (Genome-Wide Association Study)**
A study design that scans the genomes of many individuals to find genetic variants associated with a trait or disease. Results are typically displayed as Manhattan plots, with significance thresholds corrected for millions of tests (p < 5e-8).
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis, 11_Clinical_Genomics*

---
## H

**Homolog**
A gene or protein related to another by shared ancestry. Homologs may be orthologs (diverged by speciation) or paralogs (diverged by gene duplication). Homology is binary -- sequences are either homologous or not -- whereas similarity is quantitative.
*Tier 2 > 04_BLAST_Searching, 06_Phylogenetics*

**HSP (High-Scoring Segment Pair)**
The fundamental unit of a BLAST result: a pair of subsequences (one from the query, one from a database hit) that can be aligned without gaps above a given score threshold. Multiple HSPs may exist between the same query-subject pair.
*Tier 2 > 04_BLAST_Searching*

**Hydrophobicity**
A physicochemical property describing the tendency of an amino acid side chain to avoid water. Hydrophobicity scales (e.g., Kyte-Doolittle) are used to predict transmembrane regions, protein-interior residues, and surface accessibility. A hydrophobicity plot along a protein sequence reveals potential membrane-spanning helices.
*Tier 2 > 07_Protein_Structure, 10_Sequence_Motifs_and_Domains*

---
## I

**Indel (insertion/deletion)**
A type of genetic variant in which one or more nucleotides are inserted into or deleted from a genome relative to the reference. Indels in coding regions that are not multiples of three cause frameshifts.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis*

**InterPro**
An integrated database that combines protein signatures (motifs, domains, families) from multiple member databases including Pfam, PROSITE, PRINTS, and CDD into a unified classification. InterProScan assigns InterPro entries to a query protein sequence.
*Tier 2 > 10_Sequence_Motifs_and_Domains*

**Iterator (Python)**
An object that implements the `__iter__` and `__next__` methods, producing items one at a time. BioPython's `SeqIO.parse()` returns an iterator of `SeqRecord` objects, allowing memory-efficient processing of files with millions of sequences.
*Tier 1 > 11_Iterators_and_Generators*

---
## K

**K-mer**
A subsequence of length k extracted from a biological sequence. K-mer frequencies are used in genome assembly (de Bruijn graphs), sequence classification, species identification, and error correction. For DNA with an alphabet of 4, there are 4^k possible k-mers.
*Tier 2 > 10_Sequence_Motifs_and_Domains, 12_Comparative_Genomics*

**Kaplan-Meier (estimator)**
A non-parametric statistic used to estimate the survival function from time-to-event data, accounting for censored observations. In clinical genomics, Kaplan-Meier curves are used to compare survival between patient groups (e.g., by genotype or treatment arm).
*Tier 3 > 11_Clinical_Genomics*

---
## L

**Lambda (Python)**
An anonymous, single-expression function defined with the `lambda` keyword. Useful as a quick inline function for sorting or filtering, e.g., `sorted(records, key=lambda r: len(r.seq))`.
*Tier 1 > 06_Functions*

**List (Python)**
A mutable, ordered collection that can hold items of any type. Lists support indexing, slicing, appending, and iteration. One of the most frequently used data structures for storing sequences of results, parsed records, or feature annotations.
*Tier 1 > 08_Lists_and_Tuples*

---
## M

**MAF (Minor Allele Frequency)**
The frequency of the second most common allele at a genetic locus in a given population. Variants with MAF < 1% are typically classified as rare. MAF is used to filter variants in GWAS and clinical genomics pipelines.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis, 11_Clinical_Genomics*

**Manhattan plot**
A scatter plot used in GWAS where each point represents a genetic variant, with chromosomal position on the x-axis and the negative log10 of its p-value on the y-axis. Significant associations appear as towers rising above the genome-wide significance threshold.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis, 06_Statistics_for_Bioinformatics*

**MD (Molecular Dynamics)**
A computational simulation method that models the physical movements of atoms and molecules over time by numerically integrating Newton's equations of motion. MD simulations reveal protein dynamics, folding pathways, and ligand binding mechanisms.
*Tier 3 > 09_Molecular_Modeling_and_Docking*

**Motif (sequence motif)**
A short, recurring pattern in DNA, RNA, or protein sequences that has (or is presumed to have) biological significance, such as a transcription factor binding site or an enzyme active-site signature. Motifs can be represented as consensus sequences, regular expressions, or position weight matrices.
*Tier 2 > 10_Sequence_Motifs_and_Domains; Tier 3 > 05_Promoter_and_Regulatory_Analysis*

**MSA (Multiple Sequence Alignment)**
An alignment of three or more biological sequences, arranged so that homologous positions are in the same column. MSAs are used to identify conserved regions, build phylogenetic trees, and construct sequence profiles. Popular tools include MAFFT, MUSCLE, Clustal Omega, and T-Coffee.
*Tier 2 > 05_Multiple_Sequence_Alignment*

---
## N

**N50**
A widely used assembly quality metric: the length of the shortest contig (or scaffold) such that 50% of the total assembly length is contained in contigs of that length or longer. Higher N50 values indicate a more contiguous assembly.
*Tier 3 > 01_NGS_Fundamentals*

**NCBI (National Center for Biotechnology Information)**
A division of the U.S. National Library of Medicine that hosts major biological databases including GenBank, RefSeq, PubMed, dbSNP, ClinVar, and SRA. NCBI provides the Entrez programming utilities (E-utilities) for programmatic access.
*Tier 2 > 01_Biological_Databases*

**Needleman-Wunsch (algorithm)**
A dynamic programming algorithm for global pairwise sequence alignment. It finds the optimal alignment across the entire length of both sequences by filling a scoring matrix and performing a traceback. Time and space complexity are both O(mn).
*Tier 2 > 03_Pairwise_Sequence_Alignment*

**NGS (Next-Generation Sequencing)**
High-throughput sequencing technologies (Illumina, PacBio, Oxford Nanopore, etc.) that generate millions to billions of reads in a single run. NGS has enabled genome sequencing, RNA-seq, ChIP-seq, metagenomics, and many other applications at scale.
*Tier 3 > 01_NGS_Fundamentals*

**NMR (Nuclear Magnetic Resonance)**
A spectroscopic technique used to determine the three-dimensional structure of biological molecules in solution. NMR provides information on dynamics and conformational flexibility that crystallography cannot, but is limited to smaller proteins (typically < 40 kDa).
*Tier 2 > 07_Protein_Structure; Tier 3 > 09_Molecular_Modeling_and_Docking*

---
## O

**ORF (Open Reading Frame)**
A continuous stretch of codons that begins with a start codon (ATG) and ends at a stop codon (TAA, TAG, or TGA), with no internal stop codons. A DNA sequence has six reading frames (three on each strand). Finding ORFs is the first step in gene prediction.
*Tier 2 > 02_BioPython_Essentials, 12_Comparative_Genomics*

**Ortholog**
Genes in different species that evolved from a common ancestral gene through speciation. Orthologs typically retain the same function. For example, the insulin gene in humans and the insulin gene in mice are orthologs.
*Tier 2 > 06_Phylogenetics, 12_Comparative_Genomics*

**OTU (Operational Taxonomic Unit)**
A cluster of similar sequences (typically 97% identity for 16S rRNA) used as a proxy for microbial species in community analysis. OTUs have been partially replaced by ASVs, which offer higher resolution without arbitrary clustering thresholds.
*Tier 3 > 04_Microbial_Diversity*

---
## P

**PAM (Point Accepted Mutation)**
A family of amino acid substitution matrices based on an evolutionary model of accepted point mutations. PAM1 represents 1% divergence; higher matrices (e.g., PAM250) are extrapolated for more divergent sequences. Note that the numbering is opposite to BLOSUM: higher PAM = more distant.
*Tier 2 > 03_Pairwise_Sequence_Alignment*

**Paralog**
Genes related by duplication within a genome. After duplication, paralogs may diverge in function (neofunctionalization), divide the ancestral function (subfunctionalization), or become nonfunctional (pseudogenization). Example: the alpha- and beta-globin genes.
*Tier 2 > 06_Phylogenetics, 12_Comparative_Genomics*

**PCA (Principal Component Analysis)**
A dimensionality-reduction technique that projects high-dimensional data onto the directions of greatest variance (principal components). In bioinformatics, PCA is used to visualize population structure from genotype data, cluster RNA-seq samples, and detect batch effects.
*Tier 3 > 06_Statistics_for_Bioinformatics, 07_Machine_Learning_for_Biology*

**PDB (Protein Data Bank)**
The global archive for experimentally determined three-dimensional structures of biological macromolecules. Structures are deposited with a four-character accession code (e.g., 1HHO for hemoglobin). The PDB file format stores atomic coordinates, B-factors, and metadata.
*Tier 2 > 07_Protein_Structure*

**Pfam**
A database of protein families, each represented by a curated seed alignment and a profile hidden Markov model (HMM). Now part of InterPro, Pfam classifies proteins into families and domains and is widely used for functional annotation.
*Tier 2 > 10_Sequence_Motifs_and_Domains*

**Phred (quality score)**
A logarithmic quality score assigned to each base call in sequencing data. Defined as Q = -10 * log10(P), where P is the probability of an incorrect base call. Q30 means a 1 in 1,000 chance of error (99.9% accuracy).
*Tier 3 > 01_NGS_Fundamentals*

**Phylogeny (phylogenetic tree)**
A branching diagram representing the inferred evolutionary relationships among organisms, genes, or other entities. Trees can be rooted (showing the direction of evolution) or unrooted. Common construction methods include neighbor-joining, maximum likelihood, and Bayesian inference.
*Tier 2 > 06_Phylogenetics*

**Pipe (Unix pipe)**
The `|` operator in Unix/Linux that connects the standard output of one command to the standard input of the next, enabling command chaining. For example, `samtools view file.bam | cut -f3 | sort | uniq -c` extracts and counts reference names from a BAM file.
*Tier 0 > 01_Linux_Basics, 03_Bash_Scripting*

**PROSITE**
A database of protein families and domains represented as patterns (regular expressions) and profiles. PROSITE patterns describe short functional motifs such as enzyme active sites, binding sites, and post-translational modification sites.
*Tier 2 > 10_Sequence_Motifs_and_Domains*

**PRS (Polygenic Risk Score)**
A single numerical value that summarizes an individual's genetic predisposition to a trait or disease, calculated as a weighted sum of risk allele counts across many loci. Weights are typically derived from GWAS effect sizes.
*Tier 3 > 11_Clinical_Genomics*

**PSI-BLAST (Position-Specific Iterated BLAST)**
An iterative variant of BLAST that builds a position-specific scoring matrix (PSSM) from significant hits in each round and uses it to search again, improving sensitivity for detecting remote homologs. Typically converges after 3--5 iterations.
*Tier 2 > 04_BLAST_Searching*

**PWM (Position Weight Matrix)**
A matrix of log-likelihood scores representing the frequency of each nucleotide (or amino acid) at each position of a sequence motif. PWMs are used to scan sequences for transcription factor binding sites or other short regulatory motifs.
*Tier 2 > 10_Sequence_Motifs_and_Domains; Tier 3 > 05_Promoter_and_Regulatory_Analysis*

---
## Q

**QQ plot (Quantile-Quantile plot)**
A diagnostic plot that compares observed p-value quantiles against expected quantiles under the null hypothesis. In GWAS, a well-calibrated study shows points along the diagonal, with true associations deviating upward at the tail. Systematic deviation (genomic inflation) suggests population stratification or other confounders.
*Tier 3 > 06_Statistics_for_Bioinformatics*

---
## R

**Ramachandran plot**
A scatter plot of backbone dihedral angles phi and psi for each residue in a protein structure. Certain angle combinations are sterically allowed (e.g., alpha-helix, beta-sheet), while others cause atomic clashes. Ramachandran plots are used to validate the quality of experimentally determined or predicted structures.
*Tier 2 > 07_Protein_Structure*

**Rarefaction**
A technique for estimating species richness at a standardized sequencing depth by randomly subsampling reads. Rarefaction curves plot the number of observed species (or OTUs/ASVs) as a function of sampling effort, showing whether diversity has been adequately captured.
*Tier 3 > 04_Microbial_Diversity*

**Redirect (Unix redirect)**
Operators (`>`, `>>`, `<`, `2>`) that redirect standard input, output, or error streams to or from files. For example, `blastn -query seq.fa -db nt > results.txt` saves BLAST output to a file instead of printing it to the terminal.
*Tier 0 > 01_Linux_Basics, 03_Bash_Scripting*

**Regex (regular expression)**
A pattern-matching language for searching and manipulating text. In bioinformatics, regex is used to find sequence motifs (e.g., restriction enzyme sites), parse file headers, validate input formats, and extract fields from structured text.
*Tier 1 > 12_Regular_Expressions*

**Repository (Git repository)**
A directory tracked by Git that contains project files and the complete history of their changes. Repositories can be local or remote (e.g., on GitHub). Each repository has branches, commits, and tags.
*Tier 0 > 02_Git_Version_Control*

**RMSD (Root-Mean-Square Deviation)**
A measure of the average distance between corresponding atoms of two superimposed structures, expressed in Angstroms. RMSD is the standard metric for comparing protein structures: values below 2 A typically indicate high structural similarity.
*Tier 2 > 07_Protein_Structure*

**RNA-seq**
A sequencing-based method for quantifying and profiling the transcriptome. RNA molecules are converted to cDNA, fragmented, and sequenced. Computational analysis maps reads to a reference, quantifies gene expression, and identifies differentially expressed genes.
*Tier 3 > 03_RNA_seq_Analysis*

**ROC (Receiver Operating Characteristic) curve**
A plot of the true positive rate (sensitivity) versus the false positive rate (1 - specificity) at various classification thresholds. The area under the ROC curve (AUC) summarizes classifier performance; AUC = 0.5 is random, AUC = 1.0 is perfect.
*Tier 3 > 07_Machine_Learning_for_Biology*

---
## S

**SAM (Sequence Alignment/Map)**
A tab-delimited text format for storing read alignments against a reference sequence. Each alignment record has 11 mandatory fields including read name, FLAG bits, reference position, mapping quality, and CIGAR string. BAM is its binary, compressed equivalent.
*Tier 3 > 01_NGS_Fundamentals*

**Sanger sequencing**
The chain-termination method of DNA sequencing developed by Frederick Sanger, using dideoxynucleotides to generate fragments of varying lengths. Sanger sequencing produces reads of 700--1000 bp with high accuracy and remains the gold standard for validating individual variants.
*Tier 2 > 09_Chromatogram_Analysis*

**Scaffold**
An ordered and oriented set of contigs separated by gaps of estimated size. Scaffolds are constructed using paired-end or mate-pair reads that span the gaps. Scaffolding improves the contiguity of a genome assembly beyond what contigs alone provide.
*Tier 3 > 01_NGS_Fundamentals*

**SeqIO (BioPython)**
The BioPython module for reading and writing biological sequence files in various formats (FASTA, GenBank, FASTQ, etc.). `SeqIO.parse()` returns an iterator of `SeqRecord` objects; `SeqIO.read()` is for single-record files; `SeqIO.write()` outputs records.
*Tier 2 > 02_BioPython_Essentials*

**Shell**
A program that provides a command-line interface to the operating system. The shell interprets user commands, manages processes, and supports scripting. Common shells include Bash, Zsh, and Fish.
*Tier 0 > 01_Linux_Basics, 03_Bash_Scripting*

**Smith-Waterman (algorithm)**
A dynamic programming algorithm for local pairwise sequence alignment. Unlike Needleman-Wunsch, it allows the alignment to start and end anywhere in either sequence by resetting negative scores to zero. Returns the highest-scoring local alignment.
*Tier 2 > 03_Pairwise_Sequence_Alignment*

**Substitution matrix**
A scoring matrix that assigns a value to each possible pair of residues (nucleotides or amino acids) in a sequence alignment. Positive values indicate favorable substitutions (e.g., conservative amino acid changes); negative values penalize unlikely ones. PAM and BLOSUM are the two major families for proteins.
*Tier 2 > 03_Pairwise_Sequence_Alignment*

**SVM (Support Vector Machine)**
A supervised machine learning algorithm that finds the optimal hyperplane separating two classes in feature space. SVMs are used in bioinformatics for tasks such as classifying protein subcellular localization, predicting disease variants, and distinguishing signal peptides from transmembrane domains.
*Tier 3 > 07_Machine_Learning_for_Biology*

**Synteny**
The conservation of gene order and content on chromosomes between species. Synteny analysis reveals large-scale genomic rearrangements (inversions, translocations, duplications) and helps identify orthologous regions. Visualized using dot plots or synteny block diagrams.
*Tier 2 > 12_Comparative_Genomics*

---
## T

**t-SNE (t-distributed Stochastic Neighbor Embedding)**
A nonlinear dimensionality-reduction technique well suited for visualizing high-dimensional data in 2D or 3D. Commonly used for single-cell RNA-seq data, population genetics, and any dataset where cluster structure needs to be revealed. Sensitive to perplexity hyperparameter.
*Tier 3 > 07_Machine_Learning_for_Biology*

**TMB (Tumor Mutational Burden)**
The total number of somatic mutations per megabase of tumor genome. High TMB (typically > 10 mutations/Mb) is a biomarker for response to immune checkpoint inhibitor therapy, as more mutations may produce more neoantigens.
*Tier 3 > 11_Clinical_Genomics*

**TPM (Transcripts Per Million)**
A normalized unit of gene expression in RNA-seq. TPM first normalizes each gene's read count by gene length, then scales so that the total across all genes sums to one million. Unlike RPKM/FPKM, TPM values are directly comparable between samples.
*Tier 3 > 03_RNA_seq_Analysis*

**Tuple (Python)**
An immutable ordered sequence of elements. Tuples are used when data should not change after creation, such as genomic coordinates `(chromosome, start, end)` or dictionary keys composed of multiple values.
*Tier 1 > 08_Lists_and_Tuples*

---
## U

**UMAP (Uniform Manifold Approximation and Projection)**
A dimensionality-reduction algorithm that preserves both local and global data structure better than t-SNE, with faster runtime. Increasingly used for visualizing single-cell datasets, population structure, and protein sequence space.
*Tier 3 > 07_Machine_Learning_for_Biology*

**UniProt (Universal Protein Resource)**
The comprehensive protein sequence and annotation database, consisting of two sections: SwissProt (manually reviewed, high quality) and TrEMBL (automatically annotated). UniProt entries include function, domain structure, post-translational modifications, variants, and cross-references.
*Tier 2 > 01_Biological_Databases*

**UTF-8 (Unicode Transformation Format, 8-bit)**
A variable-length character encoding that is backward-compatible with ASCII and can represent every character in the Unicode standard. UTF-8 is the dominant encoding on the web and in modern bioinformatics tools. ASCII characters occupy one byte; other characters use two to four bytes.
*Tier 0 > 04_File_Encodings*

---
## V

**VCF (Variant Call Format)**
A tab-delimited text format for reporting genomic variants (SNPs, indels, structural variants) and their genotypes across samples. A VCF file has a header section (lines starting with `##`) describing the fields, a column header line (`#CHROM`), and data lines. VCF is the standard output of variant-calling pipelines.
*Tier 3 > 02_Variant_Calling_and_SNP_Analysis*

**Volcano plot**
A scatter plot used in differential expression analysis where each point is a gene, with log2 fold change on the x-axis and negative log10 p-value on the y-axis. Genes that are both statistically significant and have large fold changes appear in the upper left and upper right corners.
*Tier 3 > 03_RNA_seq_Analysis, 06_Statistics_for_Bioinformatics*
