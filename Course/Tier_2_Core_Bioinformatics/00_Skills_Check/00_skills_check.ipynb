## Question 4: Substitution Matrices

What is the difference between BLOSUM62 and PAM250? Which one would you use for aligning **closely related** protein sequences?

````python
# Your answer here:
q4_answer = """

"""
````

````python
# --- Answer ---
print("BLOSUM (BLOcks SUbstitution Matrix):")
print("  - Derived from observed substitutions in conserved protein blocks")
print("  - Higher number = more closely related sequences (BLOSUM80 > BLOSUM45)")
print("  - BLOSUM62 is the default for most BLAST searches")
print()
print("PAM (Point Accepted Mutation):")
print("  - Based on evolutionary models of amino acid replacement")
print("  - Higher number = more divergent sequences (PAM250 > PAM30)")
print("  - PAM250 is good for distantly related proteins")
print()
print("For CLOSELY related proteins: use BLOSUM80 (or BLOSUM62).")
print("Note the reverse numbering: high BLOSUM = close; high PAM = distant.")
````

Write a BioPython Entrez query that searches the `protein` database for all human (Homo sapiens) insulin sequences and retrieves the first result in GenBank format.

````python
# Your answer here:
````

````python
# --- Answer ---
print("""from Bio import Entrez, SeqIO

Entrez.email = "your.email@example.com"

# Step 1: Search
handle = Entrez.esearch(db="protein", term="insulin[Gene] AND Homo sapiens[Organism]", retmax=1)
record = Entrez.read(handle)
handle.close()

# Step 2: Fetch the first result in GenBank format
protein_id = record["IdList"][0]
handle = Entrez.efetch(db="protein", id=protein_id, rettype="gp", retmode="text")
seq_record = SeqIO.read(handle, "genbank")
handle.close()

print(f"ID: {seq_record.id}")
print(f"Description: {seq_record.description}")
print(f"Sequence: {seq_record.seq[:50]}...")""")
print()
print("Key points:")
print("  - Always set Entrez.email (NCBI requirement)")
print("  - esearch() returns IDs; efetch() downloads records")
print("  - For protein GenBank format, use rettype='gp'")
print("  - Use field tags like [Gene], [Organism] to refine searches")
````

## Question 15: Complete BioPython Workflow

Describe (in pseudocode or real code) a workflow that:
1. Searches NCBI for a gene of interest
2. Downloads its mRNA sequence
3. Extracts the CDS
4. Translates it to protein
5. Calculates the molecular weight

Which BioPython modules would you need?

````python
# Your answer here:
````

````python
# --- Answer ---
print("""from Bio import Entrez, SeqIO
from Bio.SeqUtils import molecular_weight

Entrez.email = "your.email@example.com"

# 1. Search NCBI nucleotide database
handle = Entrez.esearch(db="nucleotide", term="BRCA1[Gene] AND Homo sapiens[Organism] AND RefSeq[Filter]", retmax=1)
results = Entrez.read(handle)
handle.close()

# 2. Download in GenBank format (to get annotations/features)
acc_id = results["IdList"][0]
handle = Entrez.efetch(db="nucleotide", id=acc_id, rettype="gb", retmode="text")
record = SeqIO.read(handle, "genbank")
handle.close()

# 3. Extract CDS
for feature in record.features:
    if feature.type == "CDS":
        cds_seq = feature.location.extract(record.seq)
        break

# 4. Translate to protein
protein = cds_seq.translate(to_stop=True)

# 5. Calculate molecular weight
mw = molecular_weight(protein, seq_type="protein")
print(f"Protein length: {len(protein)} aa")
print(f"Molecular weight: {mw:.1f} Da")""")
print()
print("Modules needed:")
print("  Bio.Entrez - NCBI database access")
print("  Bio.SeqIO - reading sequence records")
print("  Bio.Seq - sequence manipulation (translate)")
print("  Bio.SeqUtils - molecular weight calculation")
print("  Bio.SeqFeature - feature extraction (implicitly via SeqIO)")
````
