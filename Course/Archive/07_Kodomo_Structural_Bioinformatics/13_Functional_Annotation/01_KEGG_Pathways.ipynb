````python
# Example pathway data structure
EXAMPLE_PATHWAYS = {
    'hsa00010': {
        'name': 'Glycolysis / Gluconeogenesis',
        'category': 'Metabolism',
        'subcategory': 'Carbohydrate metabolism',
        'genes': ['HK1', 'HK2', 'GCK', 'GPI', 'PFKL', 'PFKM', 'ALDOA', 
                  'TPI1', 'GAPDH', 'PGK1', 'PGAM1', 'ENO1', 'PKM', 'LDHA']
    },
    'hsa00020': {
        'name': 'Citrate cycle (TCA cycle)',
        'category': 'Metabolism',
        'subcategory': 'Carbohydrate metabolism',
        'genes': ['CS', 'ACO1', 'ACO2', 'IDH1', 'IDH2', 'OGDH', 
                  'SUCLA2', 'SUCLG1', 'SDHA', 'FH', 'MDH1', 'MDH2']
    },
    'hsa04110': {
        'name': 'Cell cycle',
        'category': 'Cellular Processes',
        'subcategory': 'Cell growth and death',
        'genes': ['CDK1', 'CDK2', 'CDK4', 'CDK6', 'CCNA1', 'CCNB1', 
                  'CCND1', 'CCNE1', 'RB1', 'E2F1', 'TP53', 'CDKN1A']
    },
    'hsa04210': {
        'name': 'Apoptosis',
        'category': 'Cellular Processes',
        'subcategory': 'Cell growth and death',
        'genes': ['BAX', 'BAK1', 'BCL2', 'BCL2L1', 'CASP3', 'CASP8', 
                  'CASP9', 'CYCS', 'APAF1', 'FAS', 'FASLG', 'TP53']
    }
}

def get_genes_in_pathway(pathway_id):
    """Get genes in a KEGG pathway."""
    pathway = EXAMPLE_PATHWAYS.get(pathway_id)
    if pathway:
        return pathway['genes']
    return []

def find_pathways_for_gene(gene_symbol):
    """Find all pathways containing a gene."""
    pathways = []
    for pathway_id, info in EXAMPLE_PATHWAYS.items():
        if gene_symbol.upper() in [g.upper() for g in info['genes']]:
            pathways.append((pathway_id, info['name']))
    return pathways

# Example
gene = 'TP53'
pathways = find_pathways_for_gene(gene)
print(f"\nPathways containing {gene}:")
for pid, name in pathways:
    print(f"  {pid}: {name}")
````

## Pathway Enrichment Analysis

```
┌─────────────────────────────────────────────────────────────────────────┐
│                    PATHWAY ENRICHMENT                                   │
├─────────────────────────────────────────────────────────────────────────┤
│                                                                         │
│   Question: Are my genes enriched in specific pathways?                 │
│                                                                         │
│   Statistical test: Fisher's exact test / Hypergeometric test           │
│                                                                         │
│   ┌────────────────────┬──────────────┬──────────────┬────────┐         │
│   │                    │ In pathway   │ Not in path  │ Total  │         │
│   ├────────────────────┼──────────────┼──────────────┼────────┤         │
│   │ In my gene list    │     k        │    n-k       │   n    │         │
│   │ Not in my list     │    K-k       │   N-K-n+k    │  N-n   │         │
│   ├────────────────────┼──────────────┼──────────────┼────────┤         │
│   │ Total              │     K        │    N-K       │   N    │         │
│   └────────────────────┴──────────────┴──────────────┴────────┘         │
│                                                                         │
│   k = genes in both my list AND pathway                                 │
│   n = total genes in my list                                            │
│   K = total genes in pathway                                            │
│   N = total genes in genome/background                                  │
│                                                                         │
│   P-value = probability of seeing k or more overlapping genes           │
│             by chance                                                   │
│                                                                         │
│   Multiple testing correction: Benjamini-Hochberg FDR                   │
│                                                                         │
└─────────────────────────────────────────────────────────────────────────┘
```

````python
from scipy import stats
import numpy as np

def pathway_enrichment(gene_list, pathway_db, background_size=20000):
    """
    Perform pathway enrichment analysis.
    
    Parameters:
        gene_list: List of gene symbols
        pathway_db: Dictionary of pathway_id → pathway info with 'genes' key
        background_size: Total genes in genome
    
    Returns:
        List of (pathway_id, name, p_value, fold_enrichment, overlapping_genes)
    """
    gene_set = set(g.upper() for g in gene_list)
    n = len(gene_set)
    N = background_size
    
    results = []
    
    for pathway_id, info in pathway_db.items():
        pathway_genes = set(g.upper() for g in info['genes'])
        K = len(pathway_genes)
        
        # Overlap
        overlap = gene_set & pathway_genes
        k = len(overlap)
        
        if k == 0:
            continue
        
        # Hypergeometric test (Fisher's exact)
        # P(X >= k) where X ~ Hypergeometric(N, K, n)
        p_value = stats.hypergeom.sf(k - 1, N, K, n)
        
        # Fold enrichment
        expected = n * K / N
        fold = k / expected if expected > 0 else 0
        
        results.append({
            'pathway_id': pathway_id,
            'name': info['name'],
            'p_value': p_value,
            'fold_enrichment': fold,
            'overlap': k,
            'pathway_size': K,
            'genes': list(overlap)
        })
    
    # Sort by p-value
    results.sort(key=lambda x: x['p_value'])
    
    # FDR correction (Benjamini-Hochberg)
    n_tests = len(results)
    for i, r in enumerate(results):
        r['fdr'] = min(r['p_value'] * n_tests / (i + 1), 1.0)
    
    return results

# Example: test with differentially expressed genes
de_genes = ['HK1', 'HK2', 'PFKL', 'GAPDH', 'ENO1', 'PKM',  # Glycolysis
            'TP53', 'BAX', 'CASP3', 'BCL2',  # Apoptosis
            'BRCA1', 'ATM', 'CHEK2']  # DNA repair (not in our DB)

results = pathway_enrichment(de_genes, EXAMPLE_PATHWAYS)

print("Pathway Enrichment Results:")
print("=" * 70)
for r in results:
    print(f"\n{r['pathway_id']}: {r['name']}")
    print(f"  P-value: {r['p_value']:.2e}, FDR: {r['fdr']:.2e}")
    print(f"  Fold enrichment: {r['fold_enrichment']:.1f}x")
    print(f"  Overlap: {r['overlap']}/{r['pathway_size']} genes")
    print(f"  Genes: {', '.join(r['genes'])}")
````
