
## Asymmetric Unit

The **asymmetric unit** is the smallest portion of the crystal that can generate the entire unit cell through symmetry operations.

```
                    UNIT CELL with 4-fold symmetry
                    
              ┌─────────────────────────────────┐
              │               │                 │
              │    ╔═══╗     │     ╔═══╗       │
              │    ║ P ║     │     ║ P ║       │
              │    ╚═══╝     │     ╚═══╝       │
              │               │        ↑        │
              │───────────────┼─────────────────│
              │               │                 │
              │    ╔═══╗     │     ╔═══╗       │
              │    ║ P ║     │     ║ P ║       │  ← Asymmetric
              │    ╚═══╝     │     ╚═══╝       │     Unit (AU)
              │        ↑      │                 │
              └─────────────────────────────────┘
                       │
                  Asymmetric
                  Unit (AU)
              
    4 copies of AU → complete unit cell (via rotation symmetry)
```

````python
# Parse unit cell from PDB file

def parse_cryst1(pdb_file):
    """
    Extract unit cell parameters from PDB CRYST1 record.
    
    CRYST1 format:
    CRYST1   a      b      c     alpha  beta  gamma  spacegroup  Z
    Columns: 7-15  16-24  25-33  34-40  41-47  48-54  56-66      67-70
    """
    with open(pdb_file, 'r') as f:
        for line in f:
            if line.startswith('CRYST1'):
                a = float(line[6:15])
                b = float(line[15:24])
                c = float(line[24:33])
                alpha = float(line[33:40])
                beta = float(line[40:47])
                gamma = float(line[47:54])
                spacegroup = line[55:66].strip()
                z = int(line[66:70]) if len(line) > 66 else None
                
                return {
                    'a': a, 'b': b, 'c': c,
                    'alpha': alpha, 'beta': beta, 'gamma': gamma,
                    'spacegroup': spacegroup,
                    'z': z
                }
    return None

# Example CRYST1 line
example = "CRYST1   40.960   18.650   22.520  90.00  90.77  90.00 P 1 21 1      2"
print("Example CRYST1 record:")
print(example)
print("\nParsed values:")
print(f"  Cell dimensions: a={40.960:.2f} Å, b={18.650:.2f} Å, c={22.520:.2f} Å")
print(f"  Angles: α={90.00}°, β={90.77}°, γ={90.00}°")
print(f"  Space group: P 1 21 1 (monoclinic)")
print(f"  Z (molecules per cell): 2")
````

## Space Groups

Space groups describe the symmetry of a crystal. There are **230 space groups** total, but only **65** are compatible with proteins (chiral molecules).

### Most Common Space Groups for Proteins

| Space Group | System | % of PDB |
|-------------|--------|----------|
| P 21 21 21 | Orthorhombic | ~24% |
| P 21 | Monoclinic | ~15% |
| C 2 2 21 | Orthorhombic | ~8% |
| P 1 21 1 | Monoclinic | ~7% |
| P 43 21 2 | Tetragonal | ~5% |

````python
# Apply symmetry operations
import numpy as np

def apply_rotation_matrix(coords, rotation_matrix, translation):
    """
    Apply crystallographic symmetry operation.
    
    new_coords = R * coords + t
    """
    return np.dot(rotation_matrix, coords) + translation

# Example: 2-fold rotation around Y axis
rotation_2fold_y = np.array([
    [-1, 0, 0],
    [0, 1, 0],
    [0, 0, -1]
])

translation = np.array([0.5, 0.5, 0])  # Half cell translation

original_atom = np.array([10.0, 5.0, 8.0])
symmetric_atom = apply_rotation_matrix(original_atom, rotation_2fold_y, translation)

print(f"Original atom position: {original_atom}")
print(f"After 2-fold rotation + translation: {symmetric_atom}")
````
