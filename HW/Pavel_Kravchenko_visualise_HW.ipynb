Описание столбцов
- Ensembl ID - идентификатор гена в <a href=https://www.ensembl.org/index.html> Ensembl </a>
- Uniprot IDs - идентификаторы <a href=http://www.uniprot.org/> Uniprot</a>, соответствующие белковым продуктам гена
- Log10 Length - десятичный логарифм длины гена в нуклеотидах
- Symbol - официальный <a href=https://www.genenames.org/> символ </a> гена
- Log10 Degree - десятичный логарифм степени вершины гена в <a href=http://science.sciencemag.org/content/347/6224/1257601> интерактоме Menche et al, 2015 </a>
- Log10 B cell FPKM - десятичный логарифм <a href=https://statquest.org/2015/07/09/rpkm-fpkm-and-tpm-clearly-explained/> FPKM </a> в B клетках человека
- Log10 Neuron FPKM - десятичный логарифм FPKM в B клетках человека

Визуализируйте распределение длин, подпишите оси.  **(1 балл)** 

```python
def nan_to_mean(arr):
    nan_ind = np.isnan(arr)
    print(sum(nan_ind))
    arr[nan_ind] = np.mean(arr[~nan_ind])

length = df["Log10 Length"]
print(sum(np.isnan(length)))
nan_to_mean(length)
print(length)
```

```python

sns.distplot(length)
plt.title("Log10 Length distribution", size=18)
plt.xlabel("Length", size=15)
plt.ylabel("Density", size=15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.legend(["Density"], loc="upper right", fontsize=15)
```

Нарисуйте теплокарту, на которой по осям будут экспрессии в нейронах и B-клетках, чтобы оси были разбиты на 30 бинов, а цветом было показано количество строк в таблице, которые попадают в пересечение соответствующих бинов. Какой можно сделать вывод из этой теплокарты? ** (3 балла) **

```python
import altair as alt
```

```python
alt.data_transformers.enable('json')
alt.renderers.enable('notebook')


alt.Chart(df).mark_rect().encode(
    alt.X("Log10 B cell FPKM", bin=alt.Bin(maxbins=30)),
    alt.Y("Log10 Neuron FPKM", bin=alt.Bin(maxbins=30)),
    color='count()'
)
```

Можно  заметить, что экспрессии генов распределены нормально. Больше всего наблюдается слабо экспрессируемых генов. Также имеется симметричная область пониженной экспрессии посередине тепловой карты, что может говорить о низкой представленности определённой группы генов.
В целом, и в той, и в другой группе имеются непредставленные гены, сравнение по которым произвести нельзя (показаны белым)
