### Problem 2.1: Basic ORF Finder

```python
from Bio.Seq import Seq

class ORFFinder:
    """
    Find Open Reading Frames in DNA sequences.
    
    Based on the ORF_finder.ipynb project from source materials.
    """
    
    STOP_CODONS = ['TAA', 'TAG', 'TGA']
    START_CODON = 'ATG'
    
    def __init__(self, sequence, min_protein_length=100):
        """
        Initialize with a DNA sequence.
        
        Args:
            sequence: DNA sequence string or Bio.Seq object
            min_protein_length: Minimum protein length in amino acids
        """
        self.sequence = str(sequence).upper()
        self.min_protein_length = min_protein_length
        self.min_orf_length = min_protein_length * 3
        self.orfs = []
    
    def find_orfs_in_frame(self, frame):
        """
        Find all ORFs in a specific reading frame.
        
        Args:
            frame: 0, 1, or 2 for forward; -1, -2, -3 for reverse
        
        Returns:
            List of ORF dictionaries
        """
        orfs = []
        
        # Get sequence for this frame
        if frame >= 0:
            seq = self.sequence[frame:]
            strand = '+'
        else:
            # Reverse complement
            seq = str(Seq(self.sequence).reverse_complement())
            seq = seq[abs(frame) - 1:]
            strand = '-'
        
        # YOUR CODE HERE
        # Find all ORFs (ATG to stop codon)
        # Only include ORFs >= min_orf_length
        
        return orfs
    
    def find_all_orfs(self):
        """
        Find ORFs in all 6 reading frames.
        
        Returns:
            List of all ORF dictionaries
        """
        all_orfs = []
        
        # Forward frames (0, 1, 2)
        for frame in range(3):
            all_orfs.extend(self.find_orfs_in_frame(frame))
        
        # Reverse frames (-1, -2, -3)
        for frame in [-1, -2, -3]:
            all_orfs.extend(self.find_orfs_in_frame(frame))
        
        self.orfs = all_orfs
        return all_orfs

# Test
test_seq = "ATGAAAGCCCTGTAAATGGCCCCCTGAATGTTTTAA"
finder = ORFFinder(test_seq, min_protein_length=3)
orfs = finder.find_all_orfs()
print(f"Found {len(orfs)} ORFs")
for orf in orfs:
    print(orf)
```

## 🧬 Part 3: GC Content Landscape

Based on the GC_landscape project from source materials.

### Problem 3.1: GC Calculator

```python
class GCLandscape:
    """
    Calculate and visualize GC content landscape.
    
    Based on GC_landscape.ipynb from source materials.
    """
    
    def __init__(self, sequence):
        """
        Initialize with a DNA sequence.
        """
        self.sequence = str(sequence).upper()
        self.gc_values = None
        self.positions = None
    
    def calculate_gc_sliding_window(self, window_size=5000, step=5):
        """
        Calculate GC content in sliding windows.
        
        Args:
            window_size: Size of sliding window
            step: Step size between windows
        
        Returns:
            Tuple of (positions, gc_values)
        """
        positions = []
        gc_values = []
        
        for i in range(0, len(self.sequence) - window_size, step):
            window = self.sequence[i:i + window_size]
            gc = (window.count('G') + window.count('C')) / window_size * 100
            positions.append(i + window_size // 2)
            gc_values.append(round(gc, 3))
        
        self.positions = positions
        self.gc_values = gc_values
        
        return positions, gc_values
    
    def statistics(self):
        """
        Calculate GC landscape statistics.
        
        Returns:
            Dictionary with: mean, std, min, max, high_gc_regions, low_gc_regions
        """
        if self.gc_values is None:
            raise ValueError("Run calculate_gc_sliding_window first")
        
        gc_array = np.array(self.gc_values)
        mean_gc = np.mean(gc_array)
        std_gc = np.std(gc_array)
        
        # YOUR CODE HERE
        # Find high GC regions (> mean + std)
        # Find low GC regions (< mean - std)
        
        return {
            'mean': mean_gc,
            'std': std_gc,
            'min': np.min(gc_array),
            'max': np.max(gc_array),
            'high_gc_regions': None,  # YOUR CODE HERE
            'low_gc_regions': None  # YOUR CODE HERE
        }
    
    def plot(self, title="GC Content Landscape", show_mean=True, show_std=True):
        """
        Create GC landscape plot.
        
        Based on the visualization from GC_landscape.ipynb
        """
        if self.gc_values is None:
            raise ValueError("Run calculate_gc_sliding_window first")
        
        fig, ax = plt.subplots(figsize=(14, 5))
        
        # Plot GC content
        ax.fill_between(self.positions, self.gc_values, alpha=0.5, color='blue')
        ax.plot(self.positions, self.gc_values, color='blue', linewidth=0.5)
        
        # Statistics
        mean_gc = np.mean(self.gc_values)
        std_gc = np.std(self.gc_values)
        
        if show_mean:
            ax.axhline(mean_gc, color='red', linestyle='-', 
                      label=f'Mean: {mean_gc:.1f}%')
        
        if show_std:
            ax.axhline(mean_gc + std_gc, color='red', linestyle='--', alpha=0.5,
                      label=f'+1 SD: {mean_gc + std_gc:.1f}%')
            ax.axhline(mean_gc - std_gc, color='red', linestyle='--', alpha=0.5,
                      label=f'-1 SD: {mean_gc - std_gc:.1f}%')
        
        ax.set_xlabel('Position (bp)')
        ax.set_ylabel('GC Content (%)')
        ax.set_title(title)
        ax.legend()
        
        plt.tight_layout()
        return fig, ax

# Test
# Generate a sequence with varying GC content
np.random.seed(42)
low_gc = ''.join(np.random.choice(['A', 'T'], 5000))
high_gc = ''.join(np.random.choice(['G', 'C'], 5000))
mid_gc = ''.join(np.random.choice(['A', 'T', 'G', 'C'], 5000))

test_genome = low_gc + high_gc + mid_gc + low_gc + high_gc

gc = GCLandscape(test_genome)
gc.calculate_gc_sliding_window(window_size=500, step=50)
gc.plot(title="Test Genome GC Landscape")
print(gc.statistics())
```

## 🧬 Part 4: Complete Genome Analysis Pipeline

### Problem 4.1: Full Analysis Tool

Combine ORF finder and GC analysis into a complete tool.

```python
class GenomeAnalyzer:
    """
    Complete genome analysis pipeline.
    
    Combines:
    - Basic sequence statistics
    - ORF finding
    - Codon usage analysis
    - GC content landscape
    """
    
    def __init__(self, fasta_file):
        """
        Load genome from FASTA file.
        """
        record = SeqIO.read(fasta_file, 'fasta')
        self.sequence = str(record.seq)
        self.name = record.id
        self.description = record.description
        
        # Initialize analysis components
        self.orf_finder = None
        self.gc_landscape = None
        self.codon_analyzer = None
    
    def basic_stats(self):
        """
        Calculate basic sequence statistics.
        """
        nuc_counts = {'A': 0, 'T': 0, 'G': 0, 'C': 0}
        for nuc in self.sequence:
            if nuc in nuc_counts:
                nuc_counts[nuc] += 1
        
        total = sum(nuc_counts.values())
        gc_content = (nuc_counts['G'] + nuc_counts['C']) / total * 100
        
        print(f"\nSequence: {self.name}")
        print(f"Description: {self.description}")
        print(f"\nLength: {len(self.sequence):,} bp")
        print(f"\nNucleotide Composition:")
        for nuc, count in nuc_counts.items():
            print(f"  {nuc}: {count:,} ({count/total*100:.2f}%)")
        print(f"\nGC Content: {gc_content:.2f}%")
    
    def find_orfs(self, min_length=100):
        """
        Find all ORFs.
        """
        self.orf_finder = ORFFinder(self.sequence, min_length)
        orfs = self.orf_finder.find_all_orfs()
        
        print(f"\nORF Analysis (min length: {min_length} aa):")
        print(f"  Total ORFs found: {len(orfs)}")
        
        if orfs:
            lengths = [len(orf.get('protein', '')) for orf in orfs]
            print(f"  Mean protein length: {np.mean(lengths):.1f} aa")
            print(f"  Longest protein: {max(lengths)} aa")
        
        return orfs
    
    def analyze_gc(self, window_size=5000):
        """
        Perform GC landscape analysis.
        """
        self.gc_landscape = GCLandscape(self.sequence)
        self.gc_landscape.calculate_gc_sliding_window(window_size=window_size)
        stats = self.gc_landscape.statistics()
        
        print(f"\nGC Landscape Analysis (window: {window_size} bp):")
        print(f"  Mean GC: {stats['mean']:.2f}%")
        print(f"  SD: {stats['std']:.2f}%")
        print(f"  Range: {stats['min']:.2f}% - {stats['max']:.2f}%")
        
        return stats
    
    def full_report(self, output_prefix=None):
        """
        Generate complete analysis report with visualizations.
        """
        print("=" * 60)
        print("GENOME ANALYSIS REPORT")
        print("=" * 60)
        
        self.basic_stats()
        self.find_orfs()
        self.analyze_gc()
        
        # Create visualization
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        
        # Plot 1: GC landscape
        if self.gc_landscape:
            axes[0, 0].fill_between(self.gc_landscape.positions, 
                                    self.gc_landscape.gc_values, 
                                    alpha=0.5)
            axes[0, 0].set_title('GC Content Landscape')
            axes[0, 0].set_xlabel('Position (bp)')
            axes[0, 0].set_ylabel('GC %')
        
        # Plot 2: Protein length distribution
        if self.orf_finder and self.orf_finder.orfs:
            lengths = [len(orf.get('protein', '')) for orf in self.orf_finder.orfs]
            axes[0, 1].hist(lengths, bins=50, edgecolor='black')
            axes[0, 1].set_title('Protein Length Distribution')
            axes[0, 1].set_xlabel('Length (aa)')
            axes[0, 1].set_ylabel('Count')
        
        # Plot 3: ORF density
        # YOUR CODE HERE
        axes[1, 0].set_title('ORF Density')
        
        # Plot 4: Nucleotide composition
        nucs = ['A', 'T', 'G', 'C']
        counts = [self.sequence.count(n) for n in nucs]
        axes[1, 1].bar(nucs, counts, color=['green', 'red', 'yellow', 'blue'])
        axes[1, 1].set_title('Nucleotide Composition')
        axes[1, 1].set_ylabel('Count')
        
        plt.suptitle(f'Genome Analysis: {self.name}', fontsize=14)
        plt.tight_layout()
        
        if output_prefix:
            plt.savefig(f'{output_prefix}_analysis.png', dpi=150)
        
        plt.show()
        
        print("\n" + "=" * 60)
        print("ANALYSIS COMPLETE")
        print("=" * 60)
```
