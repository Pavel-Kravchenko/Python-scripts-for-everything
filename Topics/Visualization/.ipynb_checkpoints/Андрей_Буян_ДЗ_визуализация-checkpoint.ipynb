## Данные

Вам предоставлен датасет, похожий на использовавшийся на лекции, но слегка модифицированный. Каждая строка таблицы соответствует гену. <BR>
Описание столбцов
- Ensembl ID - идентификатор гена в <a href=https://www.ensembl.org/index.html> Ensembl </a>
- Uniprot IDs - идентификаторы <a href=http://www.uniprot.org/> Uniprot</a>, соответствующие белковым продуктам гена
- Log10 Length - десятичный логарифм длины гена в нуклеотидах
- Symbol - официальный <a href=https://www.genenames.org/> символ </a> гена
- Log10 Degree - десятичный логарифм степени вершины гена в <a href=http://science.sciencemag.org/content/347/6224/1257601> интерактоме Menche et al, 2015 </a>
- Log10 B cell FPKM - десятичный логарифм <a href=https://statquest.org/2015/07/09/rpkm-fpkm-and-tpm-clearly-explained/> FPKM </a> в B клетках человека
- Log10 Neuron FPKM - десятичный логарифм FPKM в B клетках человека

```python
df = pd.read_csv("primary.tsv", sep="\t")
```

```python
df["ProtNumber"] = df["Uniprot IDs"].str.split().str.len()
```

```python
df["Log10 ProtNumber"] = np.log10(df["ProtNumber"])
```

Настройте стиль seaborn и выберите себе цветовую схему ** (0.5 балла) **

```python
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("husl")
```

```python
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = [10, 6] 
mpl.rcParams['font.size'] = 15
mpl.rcParams['legend.fontsize'] = 15
```

### Все картинки должны быть 10 дюймов в ширину и 6 в высоту, шрифт подписей и отметок по осям, а также легенды - 15 размера

Визуализируйте распределение длин, подпишите оси.  **(1 балл)** 

```python
import matplotlib.pyplot as plt

df['Log10 Length'] = df['Log10 Length'].fillna(value=np.mean(df['Log10 Length']))
sns.distplot(df['Log10 Length'])
plt.title('Gene distribution by length')
plt.ylabel('Density')
plt.show()
```

Сравните одномерные распределения экспрессии для B-клеток и нейронов на одном графике, подпишите оси, добавьте легенду ** (2 балла) **.

```python
sns.distplot(df['Log10 B cell FPKM'], bins = 30)
sns.distplot(df['Log10 Neuron FPKM'], bins = 30)
plt.title('Genes distribution by length')
plt.xlabel('Log10 cell FPKM')
plt.ylabel('Density')
plt.legend(["Log10 B cell FPKM", "Log10 Neuron FPKM"], loc="upper center")
plt.show()
```

Сравните распределения (или упрощённые представления распределений с помощью ящиков с усами) экспрессий в нейронах и B-клетках на каждой хромосоме на одном графике ** (3 балла) ** 

```python
exp_df = pd.DataFrame()
exp_df["Log10 FPKM"] = pd.concat([df["Log10 B cell FPKM"], df["Log10 Neuron FPKM"]])
exp_df["Chromosome location"] = pd.concat([df["Chromosome"], df["Chromosome"]])
exp_df["Cell type"] = ["B cells" for i in range(len(df))] + ["Neurons" for i in range(len(df))]

sns.boxplot("Cell type", "Log10 FPKM", hue="Chromosome location", data=exp_df)
plt.legend(prop={'size': 9})
plt.show()
```

Нарисуйте теплокарту, на которой по осям будут экспрессии в нейронах и B-клетках, чтобы оси были разбиты на 30 бинов, а цветом было показано количество строк в таблице, которые попадают в пересечение соответствующих бинов. Какой можно сделать вывод из этой теплокарты? ** (3 балла) **

```python
import altair as alt
```

```python
alt.renderers.enable('notebook')


alt.Chart(df).mark_rect().encode(
    alt.X("Log10 B cell FPKM", bin=alt.Bin(maxbins=30)),
    alt.Y("Log10 Neuron FPKM", bin=alt.Bin(maxbins=30)),
    color='count()'
)
```

Из данной теплокарты можно заключить, что большинство генов почти не экспрессируется в B-клетках и нейронах. Также можно заметить нормальное распределение на других участках (светлое пятно посередине - локальный макисмум экспресии). Еще одно наблюдение - достоверной разницы по профилям экспрессии генов (в целом, не различая сами гены) в нейронах и B-клетках нет, иначе светлые участки отклонялись бы от диагонали.
