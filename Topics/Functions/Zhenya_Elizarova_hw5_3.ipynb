```python
import random

l1 = ['error']
for i in range(9):
    l1.append(random.randint(10,100))
in_file = open('in.txt','w')
for i in range(10):
    in_file.write(str(random.choice(l1)) + '\n')
in_file.close()


with open("in.txt") as file:
    ll = [int(row.strip()) for row in file if row.strip() != 'error' ]   
average = (sum(ll) / len(ll))
out = open('out.txt','w')
for i in range(len(ll)):
    out.write(str(ll[i] - average) + '\n')
out.close()

```

**D** Напишите функцию, считающую сумму чисел Фибоначчи, меньших 10000000, которые являются квадратами натуральных чисел. Использовать генераторы 

```python
import math
from itertools import takewhile

def fibonacci_gen():
    f0, f1 = 0, 1
    while True:
        if (math.sqrt(f0) - int(math.sqrt(f0))) == 0:
            yield f0
        f0, f1 = f1, f0 + f1
        
fib_gen = fibonacci_gen()       
al=list(takewhile(lambda x : x < 10000, fib_gen))
print(sum(al))
```

**E** Напишите функцию, считающую сумму чисел [трибоначчи](), делящихся на 7. Использовать генераторы

```python
def tribonacci_gen():
    f0, f1,f2 = 0, 0, 1
    while True:
        if f0 % 7 == 0:
            yield f0
        f0, f1, f2 = f1, f2, f0 + f1 + f2
        
            
ll=list((tribonacci_gen()))        
print(sum(ll))        
```
