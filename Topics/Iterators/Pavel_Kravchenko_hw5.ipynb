```python
from math import sqrt
from itertools import takewhile
from functools import reduce


def my_func():   
    def fibonacci_gen():
        f0, f1 = 0, 1
        while True:
            if sqrt(f0).is_integer() == True:
                yield f0
            f0, f1 = f1, f0 + f1      
    summ = reduce(lambda x,y : x + y, list(takewhile(lambda x : x < 10000000, fibonacci_gen())))
    return(summ)


print(my_func())
```
